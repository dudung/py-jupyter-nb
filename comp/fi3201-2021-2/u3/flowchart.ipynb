## flowchart
Langkah-langkah dengan persamaan yang disampaikan sebelumnya dapat disajikan dalam bentuk diagram alir berikut

Kode diagram alirnya dalam Mermaid adalah sebagai berikut

```mermaid
flowchart TD
  %% define elements
  A([ Mulai ])
  B([ Selesai ])
  %% connect elements
  A --> B
```
