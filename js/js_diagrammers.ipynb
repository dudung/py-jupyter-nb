## flowhart.js
How to use [flowchar.js](http://flowchart.js.org/) in `nb-js-diagrammers`.

```python
%%flowchart_magic -h 100

st=>start: Start
e=>end: End
op1=>operation: Generate
op2=>parallel: Evaluate
st(right)->op1(right)->op2
op2(path1, top)->op1
op2(path2, right)->e
```

## mermaid.js
How to use [Mermaid.js](https://mermaid-js.github.io/mermaid/#/) in `nb-js-digrammers`.

```python
%%mermaid_magic -h 400

flowchart TD
    A[Start] --> B{Is it?};
    B -->|Yes| C[OK];
    C --> D[Rethink];
    D --> B;
    B ---->|No| E[End];
```
