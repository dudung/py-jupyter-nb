# list_change
Value of a `list` can be changed in several ways.

## index

```python
data = ['x', 'y', 'z', 'a', 'b', 'c']
```

```python
# change 'z' to 'r'
data[2] = 'r';
print(data)
```

```python
# change 'a' to 'w'
data[-3] = 'w';
print(data)
```

## slice operator

```python
data = ['x', 'y', 'z', 'a', 'b', 'c', 'i', 'j', 'k', 'p', 'q']
```

```python
# access 'y', 'z', 'a'
d = data[1:4]
print(d)
```

```python
# access 'x', 'y', 'z'
d = data[:3]
print(d)
```

```python
# access 'k', p'
d = data[-3:-1]
print(d)
```

```python
# access  p', 'q'
d = data[-2:]
print(d)
```

```python
# access all
d = data[:]
print(d)
```

## in operator

```python
data = ['x', 'y', 'z', 'a', 'b', 'c', 'i', 'j', 'k', 'p', 'q']
```

```python
# check existence of 'g'
d = 'g' in data
print(d)
```

```python
# check existence of 'i'
d = 'i' in data
print(d)
```
