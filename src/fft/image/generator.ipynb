# generator
Generate chess board-like image from number of rectangles in x and y directions.

```python
from PIL import Image

def create_png_image_chess_board(width, height, numx, numy):
    im = Image.new('RGB', (width, height))
    pix = im.load()
    for y in range(height):
        for x in range(width):
            if y < height / 2:
                pix[x, y] = (0, 0, 0)
            else:
                pix[x, y] = (255, 255, 255)


    return im
```

```python
def binary_chess_board(width, height, numrex, numrey):
    rows = []
    y = True
    x = True
    for j in range(height):
        if j % (height / numrey) == 0:
            y = not y
        cols = []
        for i in range(width):
            if i % (width / numrex) == 0:
                x = not x
            val = int((x + y) % 2)
            cols.append(val)
        rows.append(cols)
    return rows

def test_beinary_chess_board():
    m = binary_chess_board(20, 20, 2, 10)
    for j in m:
        print(j)

test_beinary_chess_board()
```
