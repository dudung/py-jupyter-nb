# plot_wave_curve
Plot sine and cosine wave

```python
import numpy as np

def sine_wave(freq, sample_rate, duration):
    t = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    omega = 2 * np.pi * freq
    y = np.sin(omega * t)
    return t, y

def cosine_wave(freq, sample_rate, duration):
    t = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    omega = 2 * np.pi * freq
    y = np.cos(omega * t)
    return t, y
```
