# create_kernel_without_venv
Create Jupyter Notebook kernel without Python virtual environment.

## check available kernels


```python
!jupyter kernelspec list
```

## create new kernel

```python
!python -m ipykernel install --user --name othervenv --display-name "Python 3 (othervenv)"
```

Kernel &rightarrow; Restart

```python
!jupyter kernelspec list
```

**Note**: But not yet available in the menu.


1. Kernel &rightarrow; Shutdown.
2. Save notebook.
3. Close it.
4. Reopen it.

**Result**: New kernel now is available on dropdown menu.
