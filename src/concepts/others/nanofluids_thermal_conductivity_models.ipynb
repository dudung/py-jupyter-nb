```python
# EMT (effective medium theory) model
import matplotlib.pyplot as plt
import numpy as np

xmin = 0
xmax = 1
Nx = 100
x = np.linspace(xmin, xmax, Nx)

k0 = 1
k1 = 1
k2 = 2
k3 = 3
k4 = 4

y1 = ((k1 + 2 * k0 + 2 * (k1 - k0) * x) / (k1 + 2 * k0 - 1 * (k1 - k0) * x)) * k0
y2 = ((k2 + 2 * k0 + 2 * (k2 - k0) * x) / (k2 + 2 * k0 - 1 * (k2 - k0) * x)) * k0
y3 = ((k3 + 2 * k0 + 2 * (k3 - k0) * x) / (k3 + 2 * k0 - 1 * (k3 - k0) * x)) * k0
y4 = ((k4 + 2 * k0 + 2 * (k4 - k0) * x) / (k4 + 2 * k0 - 1 * (k4 - k0) * x)) * k0

fig = plt.figure()
plt.plot(x , y1, 'k', x, y2, 'b', x, y3, 'g', x, y4, 'r')

plt.legend(["k1 = 1", "k1 = 2", "k1 = 3", "k1 = 3"])

plt.xlabel("Ï•")
plt.ylabel("keff")

plt.show()
```
