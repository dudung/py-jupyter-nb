# for print list
Print a list using for [1] and range [2] or directly [3].

```python
# define a list
x = [2, 0, 2, 2, 1, 1, 1, 5]
```

```python
# print using range
N = len(x)
for i in range(N):
    print(x[i])
```

```python
# print using list
for i in x:
    print(i)
```

```python
# print the list
print(x)
```

```python
# print using * operator and separated with comma
print(*x, sep=',')
```

```python
# print using * operator and each in newline
print(*x, sep='\n')
```
