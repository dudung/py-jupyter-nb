# kinematics_xva_graphs_2
Plot kinematics graphs using Python `matplotlib` module.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

## position
Suppose there is a function of position

\begin{equation}\tag{1}
x(t) = 3t^3 - 6t + 1,
\end{equation}

where $t$ is time.

```python
# define x(t)
def x(t):
    xx = 3*t*t*t - 6*t + 1
    return xx
```

## velocity
Function of velocity is as follow

\begin{equation}\tag{2}
v(t) = 9t^2 - 6,
\end{equation}

which is obtaned by deriving Eqn (1) with respect to time $t$.

```python
# define v(t)
def v(t):
    vv = 9*t*t - 6
    return vv
```

## acceleration
Function of acceleration is as follow

\begin{equation}\tag{3}
a(t) = 18t,
\end{equation}

which is obtaned by deriving Eqn ([2](#eqn2)) with respect to time $t$.

```python
# define a(t)
def a(t):
    aa = 18*t
    return aa
```

```python
# create data
tbeg = 0
tend = 10
N = 100
dt = (tend - tbeg) / N

ttt = []
xxx = []
vvv = []
aaa = []

for i in range(0, N + 1):
    t = tbeg + i * dt
    tt = t
    xx = x(t)
    vv = v(t)
    aa = a(t)
    
    ttt.append(tt)
    xxx.append(xx)
    vvv.append(vv)
    aaa.append(aa)
```

```python
# plot data of x(t)
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$t$")
ax.set_ylabel("$x$")
plt.plot(ttt, xxx)

# draw horizontal line
line = plt.Line2D((tbeg, tend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()
```

```python
# plot data of v(t)
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$t$")
ax.set_ylabel("$v$")
plt.plot(ttt, vvv)

# draw horizontal line
line = plt.Line2D((tbeg, tend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()
```

```python
# plot data of a(t)
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$t$")
ax.set_ylabel("$a$")
plt.plot(ttt, aaa)

# draw horizontal line
line = plt.Line2D((tbeg, tend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()
```
