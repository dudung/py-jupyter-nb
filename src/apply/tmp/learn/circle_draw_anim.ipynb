```python
import matplotlib.pyplot as plt

fig = plt.figure()

c1 = plt.Circle((0, 0), 5, fc='blue', ec='red', lw=4)
c2 = plt.Circle((7.5, 2.5), 2.5, fc='green', ec='black', lw=2, ls='dotted')

ax = plt.gca()
ax.set_aspect(1)
ax.add_patch(c1)
ax.add_patch(c2)

plt.xlim(-20, 20)
plt.ylim(-10, 10)
plt.xticks(range(-20, 25, 5))
plt.yticks(range(-10, 15, 5))
plt.grid()
plt.show()
```
