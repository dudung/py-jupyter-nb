```python
from matplotlib import pyplot as plt 
import numpy as np 
from matplotlib.animation import FuncAnimation 

fig = plt.figure(figsize=(4, 4))
axis = plt.axes(xlim =(-2, 2), ylim =(-2, 2)) 
plt.grid()

line1, = axis.plot([], [], 'ro')
line2, = axis.plot([], [], 'bo', facecolors='none') 

def init(): 
 line1.set_data([], []) 
 line2.set_data([], []) 
 return _

def animate(i): 
 x = np.linspace(-2, 2, 10) 
 y = np.sin(2 * np.pi * (x - 0.01 * i)) 
 line1.set_data(x, y)
 line2.set_data(y, x)
 return _

anim = FuncAnimation(fig, animate, init_func = init, 
     frames = 160, interval = 20, blit = True) 
anim.save('test.mp4', writer = 'ffmpeg', fps = 30) 
```
