```python
from lesslines.math.interpolation import Linear as il
import matplotlib.pyplot as plt
```

```python
x = [0, 2, 4, 6, 8, 10]
y = [0, 2, 4, 6, 8, 10]

lin_interp = il(x, y)
xx = [(i * 0.25 + 3.5) for i in range(21)]
yy = lin_interp.interpolate(xx)

plt.plot(x, y, 'o', label='data')
plt.plot(xx, yy, '--', label='interpolation')
plt.legend()
plt.show()
```

```python
x = [0.5 * i for i in range(21)]
y = [20 - (i-5)**2 for i in x]

lin_interp = il(x, y)
xx = [(1.25 + i * 0.2) for i in range(31)]
yy = lin_interp.interpolate(xx)

plt.plot(x, y, 'o', label='data')
plt.plot(xx, yy, '--', label='interpolation')
plt.legend()
plt.show()
```
