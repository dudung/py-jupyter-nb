```python
from lesslines.data.generated import parametric as prm
import matplotlib.pyplot as plt
import math
```

```python
x, y1 = prm.sin(num_points=21)

plt.plot(x, y1, 'o--')
plt.show()
```

```python
x, y2 = prm.sin(initial_phase=math.pi/2, num_points=21)

plt.plot(x, y2, 'o--')
plt.show()
```

```python
plt.plot(y1, y2, 'o--')
plt.gca().set_aspect('equal')
plt.show()
```
