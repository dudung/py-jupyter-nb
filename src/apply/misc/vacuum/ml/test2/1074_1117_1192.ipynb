# 5. Model Machine Learning
## Kami memakai model ML:
1. Logistic Regression : ogistic Regression adalah metode statistik yang digunakan untuk memodelkan hubungan antara satu atau lebih variabel independen (fitur) dengan variabel dependen biner (target), dalam hal ini, apakah seorang pengunjung akan melakukan pembelian atau tidak pada toko belanja online. Dalam konteks ini, Logistic Regression akan memprediksi probabilitas bahwa pengunjung tertentu akan melakukan pembelian berdasarkan fitur-fitur yang diamati, seperti jumlah halaman yang dilihat, durasi kunjungan, dan lainnya. Dengan demikian, model ini membantu kita untuk memahami faktor-faktor yang mempengaruhi keputusan pembelian pengunjung.
2. K-Nearest Neighbors (KNN) : K-Nearest Neighbors (KNN) adalah algoritma pembelajaran mesin yang sederhana dan intuitif. Dalam konteks ini, KNN digunakan untuk mengklasifikasikan pengunjung toko belanja online sebagai pembeli atau tidak berdasarkan fitur-fitur yang diamati. Ide dasar KNN adalah bahwa pengunjung yang memiliki fitur-fitur serupa cenderung memiliki perilaku pembelian yang mirip. Algoritma ini mengklasifikasikan setiap pengunjung berdasarkan mayoritas dari k tetangga terdekatnya dalam ruang fitur. Dengan menggunakan KNN, kita dapat dengan cepat mengidentifikasi pola-pola dalam data yang dapat membantu dalam memprediksi perilaku pembelian pengunjung.
3. Decision Tree : Decision Tree adalah model prediktif yang menggunakan struktur pohon keputusan untuk memprediksi apakah seorang pengunjung akan melakukan pembelian atau tidak. Setiap simpul dalam pohon keputusan mewakili tes pada atribut, cabang merepresentasikan hasil tes, dan daun mewakili label kelas atau nilai target. Dalam konteks toko belanja online, Decision Tree dapat membantu kita memahami aturan-aturan yang kompleks dalam data yang mempengaruhi keputusan pembelian pengunjung. Dengan demikian, model ini memberikan wawasan yang mudah diinterpretasi tentang faktor-faktor yang paling memengaruhi transaksi pembelian.
4. Random Forest : Random Forest adalah metode ensemble yang menggabungkan beberapa pohon keputusan untuk meningkatkan akurasi prediksi dan mengurangi overfitting. Dalam hal ini, Random Forest digunakan untuk memprediksi kemungkinan pembelian oleh pengunjung toko belanja online. Model ini bekerja dengan membuat banyak pohon keputusan pada data pelatihan secara acak, dan kemudian menggabungkan prediksi dari setiap pohon untuk memperoleh prediksi akhir. Dengan menggunakan Random Forest, kita dapat mengatasi masalah overfitting yang sering terjadi pada Decision Tree tunggal, sehingga menghasilkan model prediktif yang lebih robust dan akurat.

# Modeling

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, precision_score, f1_score, recall_score, confusion_matrix, roc_auc_score
from sklearn.model_selection import cross_validate
```

```python
def evaluation(model):
    y_pred = model.predict(Xtest)
    y_pred_train = model.predict(Xtrain_oversampling)
    y_pred_proba = model.predict_proba(Xtest)
    y_pred_proba_train = model.predict_proba(Xtrain_oversampling)
    
    print('Classification Report:\n', classification_report(ytest, y_pred))
    print('Confusion Matrix:\n', confusion_matrix(ytest, y_pred))
    print('Precision Score:\n', precision_score(ytest, y_pred))
    print('Recall Score:\n', recall_score(ytest, y_pred))
    print('f1 Score:\n', f1_score(ytest, y_pred))
    print('\n')
    print('Accuracy Test:\n', accuracy_score(ytest, y_pred))
    print('Accuracy Train:\n', accuracy_score(ytrain_oversampling, y_pred_train))
    print('\n')
    print('AUC Score Test:\n', roc_auc_score(ytest, y_pred_proba[:,1]))
    print('AUC Score Train:\n', roc_auc_score(ytrain_oversampling, y_pred_proba_train[:,1]))
    
    score = cross_validate(model, df_feature, df_target, cv=5, scoring='roc_auc', return_train_score=True)
    print('roc_auc(crossval train):'+ str(score['train_score'].mean()))
    print('roc_auc(crossval test):'+ str(score['test_score'].mean()))
```

```python
model_lr = LogisticRegression(solver='newton-cg',C=0.00075) #'newton-cg','sag', 'saga'
model_lr.fit(Xtrain_oversampling, ytrain_oversampling)
evaluation(model_lr)
```

```python
y_pred = model_lr.predict(Xtest)
y_pred
```

```python
# Tuning Hyperparameter
```

```python
from sklearn.model_selection import RandomizedSearchCV

p = ['l2']
c = [float(x) for x in np.linspace(0.0001, 0.001, 100)]
hyperparameters = dict(penalty=p, C=c)

rs_lr = RandomizedSearchCV(model_lr, hyperparameters, scoring='roc_auc', random_state=42, cv=5)
rs_lr.fit(Xtrain_oversampling, ytrain_oversampling)
evaluation(rs_lr)
```

```python
# Analyzing Learning Curve
```

```python
def draw_learningcurve(param_values):
    train_scores = []
    test_scores = []
    
    for i in param_values:
        model_lr = LogisticRegression(penalty='l2', C=i)
        model_lr.fit(Xtrain_oversampling, ytrain_oversampling)
        
        #eval on train
        y_pred_train_proba = model_lr.predict_proba(Xtrain_oversampling)
        train_eval = roc_auc_score(ytrain_oversampling, y_pred_train_proba[:,1])
        train_scores.append(train_eval)
        
        #eval on test
        y_pred_proba = model_lr.predict_proba(Xtest)
        test_eval = roc_auc_score(ytest, y_pred_proba[:,1])
        test_scores.append(test_eval)
        
        print('param values: ' + str(i) + '; train: ' + str(train_eval) + '; test: '+ str(test_eval))
        
    plt.plot(param_values, train_scores, label='Train')
    plt.plot(param_values, test_scores, label='Test')
    plt.xlabel('Parameter')
    plt.ylabel('AUC')
    plt.title('Learning Curve Log Regression')
    plt.legend()
    plt.show()
    
param_values = [float(x) for x in np.linspace(0.00068, 0.00075, 100)]
draw_learningcurve(param_values)
```
