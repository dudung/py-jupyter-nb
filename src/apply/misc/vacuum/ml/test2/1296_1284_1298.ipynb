```python
import pandas as pd
hr = pd.read_csv('HR.csv')
col_names = hr.columns.tolist()
print("Column names:")
print(col_names)
print("\nSample data:")
hr.head()
```

```python
hr=hr.rename(columns = {'sales':'department'})
```

```python
hr.shape
```

```python
hr['department'].unique()
```

```python
import numpy as np
hr['department']=np.where(hr['department'] =='support', 'technical', hr['department'])
hr['department']=np.where(hr['department'] =='IT', 'technical', hr['department'])
```

```python
cat_vars=['department','salary']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(hr[var], prefix=var)
    hr1=hr.join(cat_list)
    hr=hr1
```

```python
hr.drop(hr.columns[[8, 9]], axis=1, inplace=True)
hr.columns.values
```

```python
hr_vars=hr.columns.values.tolist()
y=['left']
X=[i for i in hr_vars if i not in y]
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
rfe = RFE(model,n_features_to_select=10)
rfe = rfe.fit(hr[X], hr[y])
print(rfe.support_)
print(rfe.ranking_)
```

```python
cols=['satisfaction_level', 'last_evaluation', 'time_spend_company', 'Work_accident', 'promotion_last_5years', 
      'department_RandD', 'department_hr', 'department_management', 'salary_high', 'salary_low'] 
X=hr[cols]
y=hr['left']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print('Logistic regression accuracy: {:.3f}'.format(accuracy_score(y_test, logreg.predict(X_test))))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
print('Random Forest Accuracy: {:.3f}'.format(accuracy_score(y_test, rf.predict(X_test))))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, rf.predict(X_test)))
```

```python
import matplotlib.pyplot as plt
y_pred = rf.predict(X_test)
from sklearn.metrics import confusion_matrix
import seaborn as sns

forest_cm = confusion_matrix(y_test, y_pred, labels=[1, 0])  # Use y_test for ground truth

sns.heatmap(forest_cm, annot=True, fmt='.2f', xticklabels=["Left", "Stayed"], 
            yticklabels=["Left", "Stayed"], cmap="coolwarm")  # Optional colormap
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Random Forest Confusion Matrix')
plt.show()  # Explicitly call plt.show()
```

```python
logreg_y_pred = logreg.predict(X_test)
from sklearn.metrics import confusion_matrix
import seaborn as sns

logreg_cm = confusion_matrix(y_test, logreg_y_pred, labels=[1, 0])  # Only three arguments

sns.heatmap(logreg_cm, annot=True, fmt='.2f', xticklabels=["Left", "Stayed"], 
            yticklabels=["Left", "Stayed"], cmap="coolwarm")  # Optional colormap
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Logistic Regression Confusion Matrix')
plt.show()  # Explicitly call plt.show()
```

```python
#employee turnover prediction curve

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
rf_roc_auc = roc_auc_score(y_test, rf.predict(X_test))
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, rf.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot(rf_fpr, rf_tpr, label='Random Forest (area = %0.2f)' % rf_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.show()
```

```python
hr = hr.rename(columns={'sales': 'department'})
#kode ini menggunakan fungsi rename dari library pandas untuk mengubah pada dataframe hr
```

```python
#Baris kode ini menggunakan metode unique() pada Series hr['department'] untuk mendapatkan daftar nilai unik pada kolom "department" dalam dataframe hr.
hr['department'].unique()
```

```python
import numpy as np

# Import library NumPy
import numpy as np

# Mengubah nilai 'support' menjadi 'technical' pada kolom "department"
hr['department'] = np.where(hr['department'] == 'support', 'technical', hr['department'])

# Mengubah nilai 'IT' menjadi 'technical' pada kolom "department" (setelah perubahan sebelumnya)
hr['department'] = np.where(hr['department'] == 'IT', 'technical', hr['department'])
```

```python
cat_vars=['department','salary']
for var in cat_vars:
    cat_list='var'+'_'+var
    cat_list = pd.get_dummies(hr[var], prefix=var)
    hr1=hr.join(cat_list)
    hr=hr1 
     
    #Baris kode ini mendefinisikan daftar bernama cat_vars yang berisi nama-nama variabel kategori dalam dataframe hr. Dalam contoh ini, variabel kategori adalah "department" dan "salary".
    #Loop for ini mengiterasi setiap variabel kategori dalam daftar cat_vars.
    #var in cat_vars: Mengiterasi setiap elemen (nama variabel) dalam daftar cat_vars.
    #cat_list = 'var_' + var: Membuat nama baru untuk kolom dummy dengan format "var_" diikuti nama variabel kategori.
    #cat_list = pd.get_dummies(hr[var], prefix=var): Mengubah variabel kategori hr[var] menjadi kolom dummy menggunakan fungsi pd.get_dummies.
    #hr[var]: Mengakses kolom var dari dataframe hr.
    #prefix=var: Menambahkan awalan var pada nama kolom dummy yang dihasilkan.
    #Kolom dummy yang dihasilkan akan berisi satu kolom untuk setiap kategori unik dalam variabel kategori hr[var].
    #hr1 = hr.join(cat_list): Menggabungkan kolom dummy cat_list ke dalam dataframe hr dengan menggunakan fungsi join.
    #hr.join(cat_list): Menggabungkan dataframe hr dengan dataframe cat_list berdasarkan indeks baris.
    #Kolom dummy baru akan ditambahkan ke dataframe hr di sebelah kanan kolom yang ada.
    #hr = hr1: Mengubah dataframe hr menjadi dataframe hr1 yang dihasilkan dari penggabungan. Ini memastikan bahwa dataframe hr selalu berisi hasil terbaru dari iterasi loop.
```

```python
hr.drop(hr.columns[[8, 9]], axis=1, inplace=True)
hr.columns.values

#Kode untuk Menghapus Kolom ke-8 dan ke-9 dari Dataframe hr dengan Komentar
```

```python
#mendapatkan daftar nama kolom
hr_vars=hr.columns.values.tolist()
y=['left'] #mendefinisikan kolom target
X=[i for i in hr_vars if i not in y] #memilih fitur
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
model = LogisticRegression() #Baris kode ini membuat objek model LogisticRegression
rfe = RFE(model,n_features_to_select=10)#Baris kode ini membuat objek RFE dengan parameter berikut: #model: Model LogisticRegression yang diinisialisasi sebelumnya. #n_features_to_select: Jumlah fitur yang ingin dipilih (dalam hal ini, 10)
rfe = rfe.fit(hr[X], hr[y])#Baris kode ini melatih RFE dengan data hr. #hr[X]: Menyediakan data fitur yang telah dipilih sebelumnya (dari variabel X). #hr[y]: Menyediakan data target (kolom y).
print(rfe.support_)
print(rfe.ranking_)
```

```python
cols=['satisfaction_level', 'last_evaluation', 'time_spend_company', 'Work_accident', 'promotion_last_5years', 
      'department_RandD', 'department_hr', 'department_management', 'salary_high', 'salary_low']
#Baris kode ini mendefinisikan daftar bernama cols yang berisi nama-nama fitur yang ingin Anda gunakan untuk analisis.
X=hr[cols] #Baris kode ini memilih fitur yang ditentukan dalam daftar cols dari dataframe hr.
y=hr['left'] #Baris kode ini memilih kolom target dari dataframe hr.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# Membagi data menjadi training set dan testing set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

# Membangun model Logistic Regression
logreg = LogisticRegression()

# Melatih model dengan training set
logreg.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
print('Logistic regression accuracy: {:.3f}'.format(accuracy_score(y_test, logreg.predict(X_test))))
#Baris kode ini mencetak skor akurasi dengan format yang mudah dibaca.
```

```python
from sklearn.ensemble import RandomForestClassifier

# Membangun model Random Forest Classifier
rf = RandomForestClassifier()

# Melatih model dengan training set
rf.fit(X_train, y_train)
```

```python
print('Random Forest Accuracy: {:.3f}'.format(accuracy_score(y_test, rf.predict(X_test))))
#Baris kode ini mencetak skor akurasi dengan format yang mudah dibaca
```

```python
from sklearn.metrics import classification_report #Baris kode ini mengimpor fungsi classification_report dari library sklearn.metrics.
print(classification_report(y_test, rf.predict(X_test)))
#memprediksi nilai target untuk data testing set (X_test) menggunakan model rf yang telah dilatih sebelumnya.
#mencetak laporan klasifikasi untuk mengevaluasi kinerja model Random Forest.
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

# Prediksi target variable untuk testing set menggunakan model terlatih
y_pred = rf.predict(X_test)

# Hitung confusion matrix
forest_cm = confusion_matrix(y_test, y_pred, labels=[1, 0])  # Gunakan y_test untuk ground truth

# Visualisasi confusion matrix menggunakan Seaborn
sns.heatmap(forest_cm, annot=True, fmt='.2f', 
            xticklabels=["Left", "Stayed"], yticklabels=["Left", "Stayed"], cmap="coolwarm")

# Menambahkan label dan judul pada plot
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Random Forest Confusion Matrix')

# Menampilkan plot
plt.show()
```

```python
print(classification_report(y_test, logreg.predict(X_test)))
##mencetak laporan klasifikasi untuk mengevaluasi kinerja model Logistic Regression.
```

```python
# Prediksi target variable untuk testing set menggunakan model terlatih
logreg_y_pred = logreg.predict(X_test)

# Hitung confusion matrix untuk model Regresi Logistik
logreg_cm = confusion_matrix(y_test, logreg_y_pred, labels=[1, 0])

# Visualisasi confusion matrix menggunakan Seaborn
sns.heatmap(logreg_cm, annot=True, fmt='.2f', 
            xticklabels=["Left", "Stayed"], yticklabels=["Left", "Stayed"], cmap="coolwarm")

# Menambahkan label dan judul pada plot
plt.ylabel('True class')
plt.xlabel('Predicted class')
plt.title('Logistic Regression Confusion Matrix')

# Menampilkan plot
plt.show()
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt

# Hitung ROC AUC score untuk model Regresi Logistik
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))

# Hitung FPR, TPR, dan thresholds untuk model Regresi Logistik
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:, 1])

# Hitung ROC AUC score untuk model Random Forest
rf_roc_auc = roc_auc_score(y_test, rf.predict(X_test))

# Hitung FPR, TPR, dan thresholds untuk model Random Forest
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, rf.predict_proba(X_test)[:, 1])

# Buat plot ROC
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot(rf_fpr, rf_tpr, label='Random Forest (area = %0.2f)' % rf_roc_auc)

# Tambahkan garis diagonal acak
plt.plot([0, 1], [0, 1], 'r--')

# Atur batas sumbu
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])

# Tambahkan label sumbu dan judul plot
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC)')

# Tambahkan legend
plt.legend(loc="lower right")

# Tampilkan plot
plt.show()
```

```python
import numpy as np

# Akses atribut feature_importances_ dari model Random Forest
importance = rf.feature_importances_

# Urutkan indeks fitur berdasarkan kepentingan (descending order)
feature_indexes_by_importance = importance.argsort()[::-1]  # Mendapat indeks terbesar terlebih dahulu

# Label fitur
feature_labels = np.array(['satisfaction_level', 'last_evaluation', 'time_spend_company', 'Work_accident', 'promotion_last_5years',
                           'department_RandD', 'department_hr', 'department_management', 'salary_high', 'salary_low'])

# Tampilkan fitur dan kepentingan (top 5 atau semua)
for i in range(len(feature_labels)):  # Looping semua fitur (sesuaikan dengan kebutuhan)
    index = feature_indexes_by_importance[i]
    print('{}-{:.2f}%'.format(feature_labels[index], (importance[index] * 100.0)))
```

Tindak Lanjut dari Employee Turnover Prediction Project Penerapan pada Perangkat Keras:
Implementasi Model pada Perangkat Keras: Ubah model prediksi turnover yang dibuat menggunakan software seperti Python ke dalam framework machine learning seperti TensorFlow atau PyTorch. Hal ini memungkinkan model untuk dijalankan pada perangkat keras seperti GPU atau TPU, yang dapat meningkatkan performa dan kecepatan prediksi.

Pengembangan Aplikasi Prediksi Turnover:Kembangkan aplikasi web atau mobile yang terintegrasi dengan model prediksi turnover. Aplikasi ini dapat digunakan oleh HRD atau manajer untuk memprediksi risiko turnover pada karyawan mereka dan mengambil langkah-langkah preventif untuk mencegahnya.

Dilanjutkan dalam Tugas Akhir:
1. Investigasi Faktor-Faktor Lain: Memperluas model prediksi turnover dengan memasukkan faktor-faktor lain yang mungkin berkontribusi pada turnover, seperti tingkat kepuasan kerja, keseimbangan kerja-kehidupan, dan peluang pengembangan karir.
2. Analisis Prediktor Individual:Menganalisis secara mendalam faktor-faktor prediktor turnover untuk setiap karyawan, sehingga dapat memberikan intervensi yang lebih personal dan efektif untuk mencegah turnover.
