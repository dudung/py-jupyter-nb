- Ekonomi: Polusi udara dapat menyebabkan hilangnya produktivitas, biaya kesehatan yang tinggi, dan kerusakan infrastruktur.

Menariknya Masalah Polusi Udara:
- **Kompleksitas**: Polusi udara adalah masalah yang kompleks dengan banyak faktor yang berkontribusi, seperti emisi dari kendaraan, industri, pembangkit listrik, dan pembakaran bahan bakar fosil. Memahami dan mengatasi polusi udara membutuhkan pendekatan yang komprehensif dan multi-sektoral.
- **Perkembangan teknologi**: Teknologi baru, seperti sensor dan analisis data, dapat membantu kita memantau polusi udara dengan lebih baik dan mengembangkan solusi yang lebih efektif.
- **Kesadaran publik**: Semakin banyak orang yang menyadari dampak polusi udara terhadap kesehatan dan lingkungan. Hal ini mendorong permintaan akan tindakan dan solusi yang lebih baik.

Kebutuhan untuk Menyelesaikan Masalah Polusi Udara:
- **Melindungi kesehatan manusia**: Polusi udara merupakan ancaman besar bagi kesehatan manusia. Kita perlu mengambil tindakan untuk mengurangi paparan polusi udara dan melindungi kesehatan masyarakat.
- **Menjaga lingkungan**: Polusi udara merusak lingkungan dan berkontribusi pada perubahan iklim. Kita perlu bertindak untuk melindungi planet ini dan memastikan masa depan yang berkelanjutan.
- **Meningkatkan ekonomi**: Polusi udara memiliki dampak negatif pada ekonomi. Mengurangi polusi udara dapat meningkatkan produktivitas, menurunkan biaya kesehatan, dan menciptakan lapangan kerja baru.

Jaringan saraf tiruan (JST), khususnya convolutional neural networks (CNN) dan recurrent neural networks (RNN), memiliki beberapa keunggulan sehingga sangat cocok untuk menyelesaikan permasalahan polusi udara:

**1. Kemampuan untuk Belajar dari Data Kompleks:**

Polusi udara merupakan permasalahan yang kompleks dengan banyak faktor yang saling terkait, seperti emisi polutan, kondisi meteorologi, dan topografi wilayah. JST, khususnya CNN dan RNN, mampu belajar dari data yang kompleks ini dan menemukan pola yang tidak mudah dilihat oleh manusia.

**2. Akurasi Prediksi yang Tinggi:**

CNN dan RNN telah terbukti mampu menghasilkan prediksi yang akurat tentang konsentrasi polutan udara di masa depan. Hal ini penting untuk membantu pemerintah dan organisasi lainnya dalam mengembangkan strategi untuk mengurangi polusi udara.

**3. Generalizability:**

CNN dan RNN dapat digeneralisasikan ke lokasi dan kondisi baru. Hal ini berarti bahwa model yang dilatih dengan data dari satu lokasi dapat digunakan untuk memprediksi konsentrasi polutan udara di lokasi lain.

**4. Skalabilitas:**

CNN dan RNN dapat dijalankan pada skala besar, yang membuatnya ideal untuk menganalisis data polusi udara dari seluruh dunia.

**5. Kemampuan untuk Memproses Data Spaso-Temporal:**

Polusi udara adalah fenomena yang dinamis yang berubah-ubah seiring waktu dan ruang. CNN dan RNN dapat memproses data spaso-temporal, yang memungkinkan mereka untuk menangkap pola dalam data polusi udara yang tidak dapat ditangkap oleh model machine learning tradisional.

```python
# Header
import pandas as pd
import numpy as np
import missingno as msno  # Import library for missing data visualization
from sklearn import svm  # Import library for Support Vector Machine

# Data Visualization Libraries
import matplotlib.pyplot as plt
import seaborn as sns
# Install chart_studio for interactive plots (if not already installed)
!pip install chart_studio
import chart_studio.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
import cufflinks  # Cufflinks for interactive plots in Jupyter Notebook
cufflinks.go_offline()
cufflinks.set_config_file(world_readable=True, theme='pearl')

# Preprocessing Libraries
from sklearn.preprocessing import StandardScaler, LabelEncoder
from collections import Counter  # Counter for counting elements in a collection

# Classification Libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Read the CSV data into a Pandas DataFrame
data = pd.read_csv("E:\Kuliah TK.3\Semester 6\Data Sains\Filedata Indeks Standar Pencemaran Udara (ISPU) Tahun 2021.csv")

# Convert 'tanggal' column to datetime format
data['tanggal'] = pd.to_datetime(data['tanggal'])

# Print the first few rows of the data to get an idea of what it looks like
print(data.head())

# Count the occurrences of each value in the 'pm10' column
data['pm10'].value_counts()

# Drop rows with missing values in 'pm10'
data = data.dropna(subset=['pm10'])

# Replace "---" with -1 in the 'pm10' column (assuming "---" represents missing values)
data['pm10'] = data['pm10'].replace('---', -1)

# Try converting 'pm10' column to numeric format (coerce to handle errors)
data['pm10'] = pd.to_numeric(data['pm10'], errors='coerce')

# Print general information about the DataFrame (data types, missing values, etc.)
print(data.info())

# Convert 'pm10', 'pm25', 'so2', 'co', 'o3', 'no2', and 'max' columns to integer data type
data['pm10'] = data['pm10'].astype(int)
data['pm25'] = data['pm25'].astype(int)
data['so2'] = data['so2'].astype(int)
data['co'] = data['co'].astype(int)
data['o3'] = data['o3'].astype(int)
data['no2'] = data['no2'].astype(int)
data['max'] = data['max'].astype(int)

# Print general information about the DataFrame after type conversion
print(data.info())

# Create a copy of the data DataFrame
df1 = data.copy()

# Fill missing values in df1 using the median of each column
df1['pm10'] = df1['pm10'].fillna(df1['pm10'].median())
df1['pm25'] = df1['pm25'].fillna(df1['pm25'].median())
df1['so2'] = df1['so2'].fillna(df1['so2'].median())
df1['co'] = df1['co'].fillna(df1['co'].median())
df1['o3'] = df1['o3'].fillna(df1['o3'].median())
df1['no2'] = df1['no2'].fillna(df1['no2'].median())
df1['max'] = df1['max'].fillna(df1['max'].median())

# Find categorical attributes (columns with object data type)
categorical_attributes = list(df1.select_dtypes(include=['object']).columns)
print("List of categorical attributes:", categorical_attributes)

# Create a LabelEncoder object
le = LabelEncoder()

# Encode categorical attributes ('pm10' and 'max' in this case)
df1['pm10'] = le.fit_transform(df1['pm10'].astype(str))
df1['max'] = le.fit_transform(df1['max'].astype(str))

#make a boxplot graphic
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV (replace 'your_data.csv' with your actual filename)
data = pd.read_csv("E:\Kuliah TK.3\Semester 6\Data Sains\Filedata Indeks Standar Pencemaran Udara (ISPU) Tahun 2021.csv")
df = pd.DataFrame(data)

# ... rest of your code using df
# Assuming you want boxplots for 'pm10', 'pm25', and 'so2'
selected_columns = ['pm10', 'pm25', 'so2', 'co', 'o3', 'no2', 'max']

df[selected_columns].boxplot()
plt.title('Boxplot of Air Pollution Parameters')
plt.xlabel('Parameters')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```
