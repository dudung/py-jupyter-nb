Menurut saya Metadata dan Data sama sama penting karena :
* Data sendiri merupakan suatu informasi yang diperlukan untuk membentuk suatu produk
* Didalam Data terdapat Metadata yang juga tidak kalah penting untuk memastikan keaslian suatu data atau membantu dalam membuat analysis sehingga dapat menggambil suatu keputusan yang lebih akurat
* Bila data diibaratkan sebagai foto makan terdapat metadata seperti Lokasi foto diambil,waktu foto diambil, serta berapa size dari foto tersebut, dll.

Manfaat Metadata:

* Penemuan data: penemuan data yang diinginkan menjadi mudah dan cepat.
* Pemahaman data: memahami konteks dan makna di balik data.
* Pengelolaan data: pengaturan dan pengelolaan data menjadi lebih efisien.
* Ketahanan data: memastikan keakuratan dan keaslian data.

Berikut ini Tiga Jenis Meta Data :

**Metadata Deskriptif (Descriptive Metadata):**

Berperan dalam mengidentifikasi dan menjelaskan sumber informasi.

Ini membantu pengguna dalam memahami esensi dan isi dari suatu data.

Contoh dari metadata deskriptif mencakup judul, penulis, tanggal penerbitan, kata kunci, ringkasan, dan subjek.

**Metadata Struktural (Structural Metadata):**

Fokus pada informasi tentang manajemen data itu sendiri.

Ini terkait dengan bagaimana data dibuat, dijaga, dan diakses.

Contoh dari metadata administratif mencakup informasi tentang pemilik data, tanggal pembuatan, hak akses, dan riwayat perubahan data.

**Metadata Administratif (Administrative Metadata):**

Menjelaskan bagaimana data disusun dan diatur.

Informasi ini penting untuk memahami hubungan antara data dan cara penggunaannya.

Contoh metadata struktural mencakup informasi tentang format file, ukuran data, dan struktur database.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:\\Users\\shakt\\Downloads\\data11.csv")

plt.figure(figsize=(10, 6))
data[['PROVINCE', 'POP10', 'POP20']].set_index('PROVINCE').plot(kind='bar')
plt.xlabel('Provinsi')
plt.ylabel('Populasi')
plt.title('Perbandingan Populasi antara 2010 dan 2020 berdasarkan Provinsi')
plt.show()

plt.figure(figsize=(10, 6))
plt.scatter(data['LONG'], data['LAT'], c=data['EI18'], cmap='viridis')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Hubungan Koordinat Geografis dengan Edukasi Index pada Tahun 2018')
plt.colorbar(label='EI 2018')
plt.show()

plt.figure(figsize=(10, 6))
data[['UR20', 'UR21', 'UR22', 'UR23']].boxplot()
plt.ylabel('Unemployment Rate')
plt.title('Distribusi Unemployment Rate pada Tahun 2020-2023')
plt.show()

plt.figure(figsize=(10, 6))
data[['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']].plot()
plt.xlabel('Tahun')
plt.ylabel('GDP Growth Rate')
plt.title('Tren Pertumbuhan GDP pada Tahun 2020-2023')
plt.legend(loc='upper left')
plt.show()

plt.figure(figsize=(10, 6))
data[['PROVINCE', 'MW22', 'MW23']].set_index('PROVINCE').plot(kind='bar')
plt.xlabel('Provinsi')
plt.ylabel('Minimum Wage')
plt.title('Perbandingan Minimum Wage antara 2022 dan 2023 berdasarkan Provinsi')
plt.show()
```
