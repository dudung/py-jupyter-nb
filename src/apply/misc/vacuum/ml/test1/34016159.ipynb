Pengelolaan informasi dan pengetahuan merupakan bagian yang sangat penting dalam sebuah organisasi maupun perusahaan. Maka dari itu, penting untuk sebuah perusahaan memiliki sistem manajemen atau pengelolaan kearsipan yang baik. 

Dalam mengelola informasi dan pengetahuan, dikenal konsep DIKW atau Data, Information, Knowledge, dan Wisdom.

Konsep DIKW ini berbentuk piramida, dimana Data menjadi bagian paling bawah dan Wisdom adalah bagian paling atas. Piramida ini kemudian mewakili hubungan antara data, informasi, pengetahuan, dan kebijaksanaan, dimana tiap tingkatan merupakan proses untuk menuju tingkatan yang lebih tinggi.

 Setiap tingkatan menjawab pertanyaan tentang data awal dan menambah nilai dari data tersebut. Semakin kaya data yang dimiliki, maka semakin banyak pengetahuan dan wawasan yang didapatkan, sehingga perusahaan Anda mampu menggunakan data untuk pengambilan keputusan yang lebih baik.

```python
Seorang Data Scientist adalah profesional yang mengubah data menjadi wawasan yang dapat ditindaklanjuti. Mereka menggunakan berbagai teknik matematika, statistik, dan pemrograman untuk menganalisis data, membangun model, dan menemukan pola yang dapat membantu organisasi membuat keputusan yang lebih baik

Tujuan utama Data Science adalah untuk mendapatkan nilai dari data. Data Scientist membantu organisasi untuk:
a. Memahami pelanggan mereka dengan lebih baik
b. Meningkatkan efisiensi operasi
c. Mengembangkan produk dan layanan baru
d. Membuat keputusan yang lebih terinformasi

Aspek Fundamental Data Scientist:
Keterampilan Matematika dan Statistik: Data Scientist harus memiliki pemahaman yang kuat tentang matematika dan statistik untuk dapat menganalisis data dan membangun model.
Keterampilan Pemrograman: Data Scientist harus mampu menggunakan bahasa pemrograman seperti Python, R, atau SQL untuk membersihkan data, membangun model, dan memvisualisasikan hasil.
Kemampuan Berkomunikasi: Data Scientist harus mampu berkomunikasi dengan baik secara lisan dan tertulis untuk menjelaskan temuan mereka kepada pemangku kepentingan.
Keterampilan Bisnis: Data Scientist harus memiliki pemahaman yang baik tentang bisnis untuk dapat mengidentifikasi masalah yang dapat diselesaikan dengan data.
Keingintahuan dan Kreativitas: Data Scientist harus memiliki rasa ingin tahu yang tinggi dan kreatif untuk dapat menemukan solusi inovatif untuk masalah yang kompleks.
    
Contoh Pekerjaan Data Scientist:
a. Menganalisis data penjualan untuk mengidentifikasi tren dan peluang
b. Membangun model untuk memprediksi churn pelanggan
c. Mengembangkan sistem rekomendasi untuk produk dan layanan
d. Mendeteksi penipuan dalam transaksi keuangan
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Downloads/data03.csv')
```

```python
data.head()
```

```python
#plot ini menunjukkan apakah ada hubungan antara tingkat pendidikan suatu provinsi dengan pertumbuhan ekonominya
data['Avg_EI'] = data[['EI10', 'EI11', 'EI12', 'EI13', 'EI14', 'EI15', 'EI16', 'EI17', 'EI18']].mean(axis=1)

data['Avg_GDP_Growth'] = data[['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']].mean(axis=1)

sns.scatterplot(x='Avg_EI', y='Avg_GDP_Growth', data=data)
plt.title('Education Index vs. GDP Growth Rate')
plt.xlabel('Average Education Index (2010-2018)')
plt.ylabel('Average GDP Growth Rate (2020-2023)')
plt.show()
```

```python
#chart ini dapat menampilkan perkembangan jumlah penduduk dari masing masing provinsi dari 2010 ke 2020
data['Pop_Growth'] = data['POP20'] - data['POP10']

sns.barplot(x='PROVINCE', y='Pop_Growth', data=data)
plt.title('Population Growth by Province (2010-2020)')
plt.xlabel('Province Name')
plt.ylabel('Population Growth')
plt.xticks(rotation=45) 
plt.show()
```

```python
#chart ini dapat memvisualkan jumlah pengangguran dari semua provinsi dari 2020 - 2023
years = ['UR20', 'UR21', 'UR22', 'UR23']
for province in data['PROVINCE'].unique():
    province_data = data[data['PROVINCE'] == province]
    plt.plot(years, province_data[years].values.ravel(), label=province)

plt.title('Unemployment Rate Trends (2020-2023)')
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.legend() #hapus perintah ini untuk melihat diagram lebih jelas
plt.show()
```
