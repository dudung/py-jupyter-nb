Data dan Metadata keduanya berguna dan saling melengkapi\
Metadata adalah data yang mengandung informasi tentang data yang lain\
Dengan adanya metadata,pekerjaan yang dilakukan akan jauh lebih mudah \
Metadata membantu kita dalam merangkum set set data,yang berisi:\
- kapan data dibuat?
- bagaimana data dibuat?
- tujuan data dibuat?
- siapa yang membuat data?
- ukuran data
- dll\
itu semua digunakan untuk mentracking, mengklasifikasi, maupun menganalisis data\
adapun jenis metadata sebagai berikut:
- Structural Metadata
yaitu metadata yang mengandung judul,banyak halaman,table of contents dan segala hal yang menyangkut struktur sebuah data.ex= table of contents
- Descriptive Metadata
yaitu metadata yang mengandung informasi basic yang dapat mendeskripsikan sebuah data seperti ekstensi file,ukuran,tanggal,perangkat yang digunakan,ex=digital photo metadata
- Definitional Metadata
yaitu data yang mengandung pengertian umum yang dapat membantu memahami arti atau meaning dari data tersebut.ex=meaning of data


## Nomor 3
Data Scientist adalah pekerjaan yang melibatkan menganalisis data untk dijadikan wawasan atau kearifan yang selanjutnya dijadikan bahan pertimbangan untuk membuat suatu keputusan.\
Pekerjaan Data Scientist melibatkan:
- Pemrosesan Data
- Eksplorasi Data
- Pemodelan Data
- Kesimpulan dan Penyajian Data\
\
Tujuan dari Data Scientist adalah untuk memanfaatkan informasi yang terkandung dalam data untuk mendukung pengambilan keputusan yang lebih baik. Ini bisa dilakukan dengan berbagai cara, termasuk mengidentifikasi peluang bisnis baru, meningkatkan efisiensi operasional, mengoptimalkan proses, memahami perilaku pelanggan, dan banyak lagi. Dengan memahami data secara mendalam, tujuan utama saya adalah membantu organisasi atau perusahaan membuat keputusan yang lebih cerdas dan berbasis bukti.\
\
Aspek Fundamental Data Science\
Pemahaman Data: Pemahaman tentang sumber, jenis, dan struktur data yang diperlukan untuk analisis
Pemrosesan Data: Proses membersihkan, memanipulasi, dan mempersiapkan data agar dapat digunakan untuk analisis
Statistika: Penggunaan metode statistika untuk menganalisis data, menarik kesimpulan, dan membuat prediksi
Machine Learning: Penerapan algoritma komputer yang memungkinkan sistem untuk belajar dari data, mengidentifikasi pola, dan membuat keputusan tanpa program yang ekspisit
Visualisasi Data: Representasi grafis dari data yang membantu memahami pola, tren, dan hubungan dalam data
Komunikasi dan Presentasi: Kemampuan untuk menyajikan temuan analisis secara jelas dan persuasif kepada pemangku kepentingan, baik secara lisan maupun tertulis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("data-05.csv")
data.head()
```

```python
plt.figure(figsize=(25,25))
plt.pie(data["AREA"], labels=data["PROVINCE"])
plt.title("Luas Wilayah Setiap Provinsi di Indonesia")
plt.show()
```

```python
plt.figure(figsize=(25,25))
plt.pie(data["LONG"], labels=data["PROVINCE"])
plt.title("Panjang Keliling Setiap Provinsi di Indonesia")
plt.show()
```

```python
aceh = data[data['PROVINCE'] == 'Aceh']
tahun = [2020, 2021, 2022, 2023]
data_provinsi = data[['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']].values[0]

plt.plot(tahun, data_provinsi, marker='o')
plt.title('Perkembangan GDPACMP Provinsi Aceh (2020-2023)')
plt.xlabel('Tahun')
plt.ylabel('GDPACMP')
plt.grid(True)
plt.xticks(tahun)  # Menampilkan label sumbu x dengan tahun yang ditentukan
plt.show()
```

Saya Menggunakan Pie Chart untuk menvisualisasikan luas wilayah dan panjang keliling Provinsi\
Lalu grafik dipakai untuk menvisualisasikan perkembangan data
