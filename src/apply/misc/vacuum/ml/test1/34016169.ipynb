```python
Header 2

Pada dasarnya data dan metadata merupakan sesuatu yang membutuhkan satu dengan yang lainnya, keduanya sama bermanfaat dan berguna.

Data adalah kumpulan informasi mentah yang belum diolah. Merupakan sumber informasi 
mengenai dasar tentang suatu topik atau fenomena. Serta menjadi bahan baku untuk analisis 
dan pengambilan keputusan. Data juga digunakan sebagai alat untuk pengembangan serta solusi baru.

Sedangkan metadata adalah informasi tentang data, yang memberikan konteks dan makna. Bermanfaat 
untuk membantu memahami data dengan lebih baik dan mempermudah pencarian data yang relevan dengan kebutuhan sehingga
pengelolaan data menjadi lebih efektif dan efisien.

Dengan tidak adanya metadata, maka data akan disajikan dengan maksud dan konteks data tidak jelas, sehingga sulit untuk 
dianalisis dan diinterpretasikan. Untuk menemukan data yang relevan membutuhkan waktu dan usaha yang ekstra sehingga
tidak efisien, karena tidak ada informasi tentang bagaimana data tersebut terstruktur dan diorganisir.

Berikut merupakan 3 jenis metadata dan penjelasan singkatnya:

Teknis
Merupakan suata metadata yang menjelaskan format, struktur, dan penyimpanan data. 
Contohnya ialah nama, ukuran, dan jenis file.

Administratif
Merupakan suatu metadata yang menjelaskan informasi administrasi terkait pengelolaan data. 
Contohnya ialah hak cipta dan sumber data.

Struktural:
Merupakan suatu metadata yang menjelaskan hubungan dan organisasi antar elemen data.
Contohnya ialah skema database dan hierarki file
```

```python
Header 3

Data scientist merupakan suatu profesi yang menganalisis, dan memodelkan data untuk mengidentifikasi pola, 
tren, dan hubungan tersembunyi agar diterjemahkan menjadi wawasan yang berguna dalam pengambilan keputusan
sehingga lebih efisien dan efektif serta memprediksi kejadian yang mungkin terjadi di masa depan

Data science merupakan suatu ilmu yang menggabungkan ilmu komputer, statistik, dan matematika yang berguna untuk
membantu membuat keputusan yang lebih tepat dan terukur dengan menganalisis data sehingga mengoptimalkan proses 
dan meningkatkan efisiensi operasional.

Berikut aspek fundamental dari data science
1. Pengumpulan data dari berbagai sumber seperti database dan internet.
2. Pembersihan dan persiapan data untuk analisis.
3. Penggunaaan berbagai teknik statistik untuk menganalisis data dan menemukan pola.
4. Visualisasi untuk mengkomunikasikan hasil analisis data kepada orang lain.
5. Penggunaan algoritma pembelajaran mesin untuk membangun model yang dapat memprediksi masa depan dan membuat keputusan.
```
