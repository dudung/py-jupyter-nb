```python
1. Piramida dibagi menjadi empat stage:

1.Data: merupakan fakta mentah dan terukur. Data tidak memiliki makna intrinsik dan membutuhkan interpretasi untuk menjadi berguna.
2.Informasi: merupakan data yang telah diolah dan diberi makna. Informasi dapat berupa fakta, statistik, atau berita.
3.Pengetahuan: merupakan pemahaman yang diperoleh dari informasi dan pengalaman. Pengetahuan dapat berupa konsep, teori, atau keterampilan.
4.Kearifan: merupakan kebijaksanaan yang diperoleh dari pengalaman dan pengetahuan yang mendalam. Kearifan dapat digunakan untuk membuat keputusan yang tepat dan menyelesaikan masalah.
#Keberagaman:
Gambar tersebut menunjukkan bahwa setiap tingkatan memiliki beragam jenis:
Data: Beragam dalam jenis (numerik, tekstual, gambar), format, dan sumber.
Informasi: Beragam dalam tingkat akurasi, keandalan, dan relevansi.
Pengetahuan: Beragam dalam domain, disiplin ilmu, dan perspektif.
Kearifan: Beragam dalam budaya, nilai, dan tradisi.
#Arah Waktu:
Gambar tersebut menunjukkan dua panah yang menghubungkan masa lalu dan masa depan dengan piramida:
Masa Lalu (Pengalaman): Pengalaman masa lalu menjadi sumber data, informasi, pengetahuan, dan kearifan. Pengalaman membentuk pemahaman dan membantu dalam pengambilan keputusan.
Masa Depan (Kebaruan): Data, informasi, pengetahuan, dan kearifan dapat diterapkan untuk menciptakan solusi dan peluang baru di masa depan.
```

```python
2.Secara umum, metadata lebih bermanfaat daripada data. Hal ini karena metadata menyediakan konteks dan informasi yang membantu dalam memahami dan menggunakan data.

Tanpa metadata, data mentah hanyalah sekumpulan fakta yang tidak berarti dan sulit untuk dianalisis. metadata, di sisi lain, memberikan informasi penting seperti:

Asal data: Di mana dan kapan data dikumpulkan?
Struktur data: Bagaimana data diorganisir?
Definisi data: Apa arti setiap nilai data?
Kualitas data: akurat data?

Manfaat Metadata:

Mempermudah pencarian dan pengorganisasian data: Metadata memungkinkan Anda untuk menandai dan mengelompokkan data dengan cara yang logis, sehingga memudahkan untuk menemukan data yang di butuhkan.
Meningkatkan kualitas data: Metadata membantu Anda melacak asal data dan memastikan bahwa data akurat dan konsisten.
Mempermudah kolaborasi: Metadata memungkinkan orang yang berbeda untuk memahami dan menggunakan data dengan cara yang sama.

Tiga Jenis Metadata:
Metadata Deskriptif: Memberikan informasi tentang isi data, seperti judul, penulis, dan kata kunci.
Metadata Teknis: Memberikan informasi tentang format data, struktur, dan lokasi penyimpanan.
Metadata Administratif: Memberikan informasi tentang pengelolaan data, seperti hak cipta, kepemilikan, dan riwayat versi.
```

```python
#5.Memuat data
import pandas as pd


data = pd.read_csv('data09.csv')
import matplotlib.pyplot as plt
```

```python
#Menghitung perubahan populasi
data['POP_CHANGE'] = data['POP20'] - data['POP10']
```

```python
#Membuat grafik batang perubahan populasi
plt.figure(figsize=(10, 6))
data.sort_values(by='POP_CHANGE', ascending=False, inplace=True)
plt.bar(data['PROVINCE'], data['POP_CHANGE'])
plt.xticks(rotation=90)  # Memutar label x-axis untuk keterbacaan
plt.xlabel('Province')
plt.ylabel('Change in Population')
plt.title('Population Change from 2010 to 2020')
plt.show()
```

```python
#Menghitung rata-rata indeks pendidikan per tahun
ei_columns = [f'EI{year}' for year in range(10, 19)]
data['AVG_EI'] = data[ei_columns].mean(axis=1)
```

```python
#Membuat grafik
plt.figure(figsize=(10, 6))
plt.plot(ei_columns, data[ei_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average Education Index')
plt.title('Average Education Index from 2010 to 2018')
plt.grid(True)
plt.show()
gdp_columns = ['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']
plt.figure(figsize=(10, 6))
plt.plot(gdp_columns, data[gdp_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average GDP Growth Rate')
plt.title('Average GDP Growth Rate at Current Market Price (2020-2023)')
plt.grid(True)
plt.show()
```
