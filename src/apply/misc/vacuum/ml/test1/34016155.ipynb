```python
## NOMOR 1 ##

import matplotlib.pyplot as plt
import numpy as np

# Mengatur tingkatan piramida
levels = ['Data', 'Informasi', 'Pengetahuan', 'Kearifan']

# Mengatur ukuran atau lebar tiap tingkat, semakin ke atas semakin sempit
sizes = [4, 3, 2, 1]

# Mengatur warna untuk tiap tingkat
colors = ['skyblue', 'lightgreen', 'orange', 'gold']

# Membuat figure dan axes
fig, ax = plt.subplots(figsize=(5, 7))

# Plotting piramida
for i in range(len(levels)):
    ax.barh(levels[i], sizes[i], color=colors[i], align='center')

# Menambahkan judul dan label
ax.set_xlabel('Keberagaman dan Tingkatan')
ax.set_title('Piramida Data ke Kearifan')

# Menampilkan plot
plt.show()
```

```python
Kode di atas akan menghasilkan sebuah visualisasi berbentuk piramida yang menunjukkan tingkatan dari Data ke Kearifan. 
Piramida ini dimulai dari Data sebagai lapisan paling lebar, 
menggambarkan bahwa di awal terdapat banyak data yang belum terstruktur atau memiliki makna khusus. 
Seiring pergerakan ke atas, lebar piramida mengecil yang menunjukkan proses penyaringan dan peningkatan 
nilai atau kegunaan dari data tersebut menjadi Informasi, kemudian menjadi Pengetahuan, dan pada akhirnya menjadi Kearifan yang paling sempit, 
menggambarkan keunikan dan kedalaman yang terbatas namun sangat berharga.

Konsep waktu dari masa lalu ke masa depan dalam konteks piramida ini menggambarkan 
proses transformasi data yang tidak terstruktur dan mentah menjadi kearifan yang berharga. 
Ini adalah perjalanan yang membutuhkan waktu, pengalaman, dan pemahaman mendalam untuk mengekstrak nilai dan insight 
yang berarti dari kumpulan data atau informasi.
```

```python
## NOMOR 2


Pertanyaan tentang mana yang lebih perlu antara data atau metadata tidak memiliki jawaban yang absolut 
karena keduanya memainkan peran yang sangat penting dalam konteks pengelolaan dan analisis informasi. 
Data adalah inti dari setiap analisis, keputusan, atau insight, sedangkan metadata adalah informasi yang menjelaskan atau 
memberikan konteks tentang data tersebut. Metadata memungkinkan data menjadi lebih bermanfaat, dapat diakses, dan mudah dimengerti.

META DATA PERLU KARENA :
Kontekstualisasi Data: Metadata memberikan informasi penting tentang data, seperti asal usul, tujuan, waktu penciptaan, penulis, dan 
parameter lain yang menjelaskan kondisi atau batasan data. Tanpa metadata, interpretasi data bisa menjadi ambigu atau salah.

Pengelolaan Data: Dengan adanya metadata, pengelolaan arsip data menjadi lebih efisien. 
Metadata memudahkan pencarian dan pengambilan data yang relevan dari kumpulan data yang besar atau kompleks.

Kualitas dan Keakuratan Data: Metadata bisa digunakan untuk melacak perubahan pada data dan menjaga versi yang berbeda, 
yang penting untuk memastikan keakuratan dan keandalan data.

Integrasi Data: Dalam sistem informasi di mana data berasal dari berbagai sumber, 
metadata membantu mengintegrasikan data tersebut dengan memberikan standar yang memungkinkan penggabungan yang koheren.

Keamanan dan Kepatuhan: Metadata dapat menyertakan informasi tentang hak akses, kebijakan privasi, dan 
kepatuhan terhadap regulasi yang berlaku, memastikan bahwa data digunakan dan dikelola dengan cara yang aman dan sesuai hukum.

3 Jenis Metadata
Descriptive Metadata (Metadata Deskriptif): Jenis metadata ini menjelaskan informasi untuk pencarian dan penemuan. 
Contohnya termasuk judul, abstrak, penulis, dan kata kunci. 
Fungsinya adalah untuk memudahkan identifikasi dan penemuan sumber data atau informasi.

Structural Metadata (Metadata Struktural): Metadata ini menjelaskan bagaimana suatu objek digital dibangun atau bagaimana bagian-bagiannya berhubungan.
Contohnya termasuk bagaimana halaman buku diatur atau bagaimana file multimedia dikompilasi. 
Fungsinya adalah untuk memudahkan navigasi dan penyajian data atau objek informasi.

Administrative Metadata (Metadata Administratif): Jenis metadata ini berkaitan dengan manajemen sumber informasi, termasuk informasi hak cipta, 
informasi lisensi, serta data tentang preservasi dan pemeliharaan file. Ini juga bisa mencakup metadata teknis yang berkaitan 
dengan jenis file, codec, atau software yang diperlukan untuk membuka file tersebut. 
Fungsinya adalah untuk memudahkan pengelolaan hak akses, pemeliharaan, dan preservasi sumber informasi.

baik data maupun metadata sama-sama krusial. Metadata memperkaya nilai data dengan menambahkan lapisan informasi yang 
membuat data tersebut lebih mudah dimengerti, dikelola, dan digunakan secara efektif.
```

```python
## NOMOR 3 
data scientist ada lah profesional yang menggunakan keterampilan analisis, pengetahuan dan statistik dan kemampuan pemrograman
untuk mengumpulkan, dan menganalisis kumpulan data besar, tugas data Scientist seringkali meliputi pengolahan data mentah, 
pengembangan model prediktif, analisis tren data, dan komunikasi temuan kepada stakeholder non-teknis 
melalui visualisasi data dan laporan yang mudah dimengerti

ilmu Data Science bertujuan untuk membuat keputusan yang lebih informatif dan efektif 
dengan mengubah data mentah menjadi pemahaman, pengetahuan, atau produk yang bernilai. 
Hal ini dicapai melalui analisis data untuk mengidentifikasi pola, tren, dan hubungan yang tidak terlihat 
yang dapat membantu organisasi dalam memecahkan masalah, mengidentifikasi peluang baru, meningkatkan efisiensi operasional, 
dan memaksimalkan nilai dari data yang mereka miliki

adapun yang menjadi aspek fundamental dari data scientist adalah 

1. Statistika dan Probabilitas: Memahami konsep statistik dan probabilitas adalah krusial karena 
membantu dalam merumuskan hipotesis, melakukan uji hipotesis, dan menganalisis karakteristik data. 
Hal ini memungkinkan pengambilan kesimpulan yang berbasis pada data dan pembuatan model prediktif.

2. Pemrograman: Kemampuan pemrograman, terutama dalam bahasa seperti Python atau R, 
sangat penting untuk mengolah data, melakukan analisis eksploratif, membangun model machine learning, 
dan mengotomatisasi tugas-tugas pengolahan data.

3. Pengolahan dan Analisis Data: Meliputi kemampuan untuk mengumpulkan, membersihkan, 
dan memanipulasi data. Data cleaning dan preprocessing adalah langkah awal yang penting sebelum analisis lebih lanjut bisa dilakukan.

4. Machine Learning dan Model Prediktif: Menggunakan algoritma dan teknik machine learning untuk membuat model 
yang dapat memprediksi hasil masa depan atau mengungkap pola dalam data. Ini termasuk supervised learning, unsupervised learning, 
dan reinforcement learning.

5. Visualisasi Data: Kemampuan untuk menyajikan data dan temuan analisis secara visual yang efektif 
menggunakan alat seperti Matplotlib, Seaborn, atau Tableau. Visualisasi membantu dalam mempresentasikan insight kompleks dalam 
format yang lebih mudah dipahami.

6. Pemikiran Kritis dan Analitis: Kemampuan untuk mendekati masalah dengan pemikiran kritis, 
merumuskan pertanyaan yang relevan, dan menganalisis hasil untuk mendapatkan insight yang actionable.

7. Komunikasi dan Kolaborasi: Kemampuan untuk menyampaikan temuan dan rekomendasi kepada audiens non-teknis 
serta bekerja sama dengan tim lintas disiplin ilmu untuk menerapkan solusi berbasis data.
```

```python
## NOMOR 4

# Penggunaan virtual environment dalam Python adalah praktek yang sangat direkomendasikan, 
terutama ketika bekerja pada proyek yang kompleks atau ketika ada kebutuhan untuk menjaga konsistensi 
dan isolasi dependensi proyek. Virtual environment memungkinkan developer untuk membuat lingkungan yang terisolasi untuk setiap proyek, 
dimana masing-masing dapat memiliki versi library dan package yang berbeda, tanpa mengganggu proyek lain atau sistem operasi utama. 
Hal ini sangat berguna untuk:
##
1. Mengelola Dependensi: Setiap proyek Python mungkin memerlukan versi library yang berbeda. 
Dengan menggunakan virtual environment, Anda dapat menginstal versi spesifik dari sebuah library yang diperlukan oleh proyek tertentu 
tanpa mempengaruhi proyek lain yang mungkin memerlukan versi yang berbeda.

2. Konsistensi Lingkungan: Virtual environment membantu memastikan bahwa semua pengembang yang bekerja pada proyek 
menggunakan versi library dan interpreter Python yang sama, sehingga mengurangi "it works on my machine" syndrome 
Hal ini memudahkan debugging dan pengembangan karena semua anggota tim bekerja dalam kondisi yang serupa.

3. Isolasi: Proyek yang berjalan dalam virtual environment tidak akan mempengaruhi sistem operasi utama atau proyek lain. 
Ini berarti bahwa perubahan, penambahan, atau penghapusan library dalam satu virtual environment tidak akan mempengaruhi environment lain.

4. Kemudahan Deployment: Virtual environment memudahkan proses deployment karena developer dapat dengan mudah membuat file requirement 
yang mencantumkan semua dependensi proyek. Ini memastikan bahwa semua dependensi dapat diinstal secara otomatis di lingkungan produksi 
atau pengembangan lainnya.

5. Pengembangan yang Lebih Aman: Dengan mengisolasi proyek dari lingkungan global, 
virtual environment juga mengurangi risiko konflik antara package dan memastikan bahwa sistem operasi tetap aman 
dari library yang tidak diinginkan atau berpotensi berbahaya.##
```

```python
import pandas as pd

# Memuat data
data = pd.read_csv('data09.csv')
```

```python
import matplotlib.pyplot as plt

# Menghitung perubahan populasi
data['POP_CHANGE'] = data['POP20'] - data['POP10']

# Membuat grafik batang perubahan populasi
plt.figure(figsize=(10, 6))
data.sort_values(by='POP_CHANGE', ascending=False, inplace=True)
plt.bar(data['PROVINCE'], data['POP_CHANGE'])
plt.xticks(rotation=90)  # Memutar label x-axis untuk keterbacaan
plt.xlabel('Province')
plt.ylabel('Change in Population')
plt.title('Population Change from 2010 to 2020')
plt.show()
```

```python
# Menghitung rata-rata indeks pendidikan per tahun
ei_columns = [f'EI{year}' for year in range(10, 19)]  
data['AVG_EI'] = data[ei_columns].mean(axis=1)

# Membuat grafik
plt.figure(figsize=(10, 6))
plt.plot(ei_columns, data[ei_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average Education Index')
plt.title('Average Education Index from 2010 to 2018')
plt.grid(True)
plt.show()
```

```python
gdp_columns = ['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']
plt.figure(figsize=(10, 6))
plt.plot(gdp_columns, data[gdp_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average GDP Growth Rate')
plt.title('Average GDP Growth Rate at Current Market Price (2020-2023)')
plt.grid(True)
plt.show()
```
