```python
import matplotlib.pyplot as plt

# Label untuk setiap tingkat piramida
labels = ['Informasi', 'Data', 'Pengetahuan', 'Kearifan']

# Tinggi dan lebar setiap tingkat piramida
heights = [10, 20, 30, 40]
widths = [0.8] * len(labels)  # Lebar batang

# Warna untuk setiap tingkat piramida
colors = ['#77dd77', '#33cc33', '#009900', '#006600']

# Plot batang untuk setiap tingkat piramida
fig, ax = plt.subplots(figsize=(10, 10))
for i in range(len(labels)):
    ax.bar(i, heights[i], width=widths[i], color=colors[i], linewidth=2)

# Set label sumbu
ax.set_xlabel('Tingkatan')
ax.set_ylabel('Tingkat Keberagaman')

# Set label sumbu x sesuai dengan label piramida
ax.set_xticks(range(len(labels)))
ax.set_xticklabels(labels)

# Set title
plt.title('Piramida DIKW')

# Tambahkan panah
arrow_x = [0.2, 0.8]
arrow_y = [6, 2]
for i in range(len(arrow_x)):
    ax.annotate('', xy=(arrow_x[i], arrow_y[i]), xytext=(arrow_x[i], arrow_y[i] - 1),
                 arrowprops=dict(arrowstyle='->', facecolor='black'))

# Tambahkan teks
ax.text(0, 7, 'Masa Lalu (Pengalaman)', color='black')
ax.text(3, 2, 'Masa Depan (Kebaruan)', color='black')

# Tampilkan plot
plt.show()
```

```python
Arah Waktu:

Masa Lalu (Pengalaman): Data dan informasi berasal dari pengalaman masa lalu.
    Contoh: data suhu di masa lalu digunakan untuk membuat ramalan cuaca.
Masa Depan (Kebaruan): Pengetahuan dan kearifan digunakan untuk mengarahkan masa depan.
    Contoh: pengetahuan tentang faktor yang mempengaruhi tinggi badan digunakan untuk merancang program latihan yang tepat.
```

```python
Data dan metadata memiliki peranan penting dalam berbagai bidang, namun dengan manfaat yang berbeda.
DATA
Bahan baku: Digunakan untuk analisis, pelaporan, dan pengambilan keputusan.
Sumber informasi: Memberikan wawasan tentang berbagai aspek.
Membangun pengetahuan: Menjadi fondasi untuk pengembangan ilmu pengetahuan.

METADATA
Informasi tentang data: Menjelaskan asal-usul, struktur, dan makna data.
Memudahkan pengelolaan data: Membantu menemukan, memahami, dan menggunakan data.
Meningkatkan kualitas data: Memastikan data akurat, konsisten, dan terdokumentasi.

Metadata penting dikarenakan:
Mempermudah pencarian data: Metadata membantu pengguna menemukan data yang mereka butuhkan dengan cepat dan mudah.
Meningkatkan pemahaman data: Metadata memberikan informasi tentang konteks dan makna data, sehingga pengguna dapat lebih memahami data tersebut.

Tiga Jenis Metadata dan Penjelasannya:

Metadata Deskriptif: Menjelaskan isi dan karakteristik data, seperti judul, penulis, tanggal publikasi, dan format data.
Metadata Struktural: Menjelaskan struktur dan organisasi data, seperti skema database, nama kolom, dan hubungan antar tabel.
Metadata Administratif: Menjelaskan informasi tentang pengelolaan data, seperti hak cipta, izin akses, dan riwayat versi data.
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create the DataFrame with actual province names (replace "..." with names)
data = pd.DataFrame({
    "Province": ["Aceh", "Sumatera Utara", "Sumatera Barat", "Papua Barat", "Papua"],
    "EI10": [63.42, 61.2, 61.06, 45.21, 42.45]
})

# Create the line plot
plt.plot(data["EI10"], data["Province"])

# Add labels and title
plt.xlabel("Education Index (EI)")
plt.ylabel("Province")
plt.title("Education Index per Province")

# Add annotations (optional)
for i in range(len(data)):
    plt.annotate(data["Province"][i], (data["EI10"][i], data["Province"][i]))

# Save the plot (optional)
plt.savefig("education_index.png")

# Display the plot
plt.show()
```
