```python
## NO 2
    Data dan metadata memiliki peran penting dalam berbagai bidang. Berikut adalah penjelasan singkat tentang kegunaan keduanya:
Data:
1.	Lebih berguna untuk analisis dan pengambilan keputusan: Data mentah dapat dianalisis untuk menghasilkan informasi dan wawasan yang berguna.
2.	Lebih mudah dipahami dan dimanipulasi: Data dapat diolah dan ditampilkan dalam berbagai format untuk memudahkan pemahaman dan penggunaan.
Metadata:
1.	Lebih berguna untuk pengorganisasian dan penemuan data: Metadata membantu dalam menemukan data yang relevan dengan cepat dan mudah.
2.	Meningkatkan interoperabilitas data: Metadata memungkinkan data dari berbagai sumber untuk diintegrasikan dan digunakan bersama.
Mengapa Perlu Ada Metadata?
1.	Metadata penting karena membantu dalam:
2.	Menemukan data: Metadata memberikan informasi tentang isi data, sehingga memudahkan untuk menemukan data yang relevan.
3.	Memahami data: Metadata memberikan informasi tentang konteks data, sehingga membantu dalam memahami makna data.
4.	Menggunakan data: Metadata memberikan informasi tentang cara menggunakan data, sehingga membantu dalam mengolah dan menganalisis data.
Jenis Metadata:
1.	Metadata Deskriptif: Memberikan informasi tentang isi data, seperti judul, penulis, dan tanggal publikasi.
2.	Metadata Struktural: Memberikan informasi tentang struktur data, seperti format file dan skema database.
3.	Metadata Administratif: Memberikan informasi tentang pengelolaan data, seperti hak akses dan informasi lisensi.
```

```python
## NO 4
Reprodusibilitas Program Python dengan Virtual Environment
Virtual environment di Python merupakan fitur yang memungkinkan Anda untuk membuat lingkungan Python terisolasi yang berisi pustaka dan paket spesifik yang diperlukan untuk menjalankan program Anda. Hal ini membantu memastikan reproduktivitas program, bahkan pada sistem yang berbeda, karena semua dependensi program diinstal dan dikonfigurasi dalam lingkungan virtualnya sendiri.
Manfaat menggunakan virtual environment untuk reproduktivitas:
1.	Menghilangkan konflik dependensi: Virtual environment memungkinkan Anda untuk menginstal versi spesifik pustaka dan paket tanpa memengaruhi instalasi Python global Anda. Hal ini membantu menghindari konflik dependensi yang dapat terjadi saat menjalankan program di sistem yang berbeda.
2.	Memastikan konsistensi: Virtual environment memastikan bahwa program Anda akan selalu berjalan dengan cara yang sama, apa pun sistem yang digunakannya, karena semua dependensi program dikonfigurasi secara identik dalam lingkungan virtual.
3.	Memudahkan kolaborasi: Virtual environment memungkinkan Anda untuk berbagi lingkungan virtual dengan kolaborator, sehingga mereka dapat menjalankan program Anda dengan konfigurasi yang sama.
Cara membuat virtual environment:
1.Pastikan Anda memiliki Python versi 3.3 atau lebih baru.
2.Instal paket venv dengan menjalankan perintah berikut di terminal:
pip install venv
3.Buat virtual environment baru dengan nama pilihan Anda:
python -m venv venv_nama_anda
4.Aktifkan virtual environment:
source venv_nama_anda/bin/activate
5.Instal pustaka dan paket yang diperlukan untuk program Anda di dalam virtual environment.
6.Deactivate virtual environment:
deactivate
7. program dalam virtual environment:
8.Aktifkan virtual environment yang sesuai.
9.Jalankan program Anda.
Kesimpulan:
Virtual environment adalah alat yang penting untuk memastikan reproduktivitas program Python yang kompleks. Dengan menggunakan virtual environment, Anda dapat memastikan bahwa program Anda akan selalu berjalan dengan cara yang sama, apa pun sistem yang digunakannya.
```

```python
import pandas as pd
# Memuat data
data = pd.read_csv('data07.csv')
```

```python
import matplotlib.pyplot as plt

##Menghitung perubahan populasi
data['POP_CHANGE'] = data['POP20'] - data['POP10']

##Membuat grafik batang perubahan populasi
plt.figure(figsize=(10, 6))
data.sort_values(by='POP_CHANGE', ascending=False, inplace=True)
plt.bar(data['PROVINCE'], data['POP_CHANGE'])
plt.xticks(rotation=90)  # Memutar label x-axis untuk keterbacaan
plt.xlabel('Province')
plt.ylabel('Change in Population')
plt.title('Population Change from 2010 to 2020')
plt.show()
```

```python
ei_columns = [f'EI{year}' for year in range(10, 19)]  # Menghasilkan list EI10, EI11, ..., EI18
data['AVG_EI'] = data[ei_columns].mean(axis=1)


plt.figure(figsize=(10, 6))
plt.plot(ei_columns, data[ei_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average Education Index')
plt.title('Average Education Index from 2010 to 2018')
plt.grid(True)
plt.show()
```

```python
gdp_columns = ['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']
plt.figure(figsize=(10, 6))
plt.plot(gdp_columns, data[gdp_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average GDP Growth Rate')
plt.title('Average GDP Growth Rate at Current Market Price (2020-2023)')
plt.grid(True)
plt.show()
```
