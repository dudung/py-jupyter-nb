Tingkatan:
-	Data: Fakta mentah, terisolasi, dan tidak terstruktur. Contoh: tanggal lahir, suhu, penjualan harian.
-	Informasi: Data yang diolah, diorganisir, dan memiliki makna. Contoh: laporan penjualan bulanan, peta cuaca, berita terkini.
-	Pengetahuan: Pemahaman yang mendalam tentang informasi, pola, dan hubungan. Contoh: strategi pemasaran, teori ilmiah, keahlian di bidang tertentu.
-	Kearifan: Pengetahuan yang dipadukan dengan pengalaman, intuisi, dan kebijaksanaan. Contoh: pengambilan keputusan strategis, kepemimpinan yang efektif, solusi kreatif.

Keberagaman:
-	Sumber: Berasal dari berbagai tempat, seperti sensor, manusia, dokumen, dan internet.
-	Format: Teks, gambar, audio, video, dan data terstruktur.
-	Kualitas: Akurasi, keandalan, dan relevansi data dan informasi.

Arah Waktu:
-	Masa Lalu (Pengalaman): Kearifan dan pengetahuan dibangun berdasarkan akumulasi data dan informasi dari masa lalu.
-	Masa Depan (Kebaruan): Kearifan dan pengetahuan digunakan untuk memprediksi dan merencanakan masa depan, mendorong inovasi dan kemajuan.

Antara data dan metadata saling melengkapi dan sama-sama penting. Data lebih berguna untuk analisis dan pengambilan keputusan. Sedangkan Metadata lebih berguna untuk memahami, mengelola, dan menemukan data.

Tiga Jenis Metadata dan Penjelasan:

-Metadata Deskriptif:
Memberikan informasi tentang isi data, seperti judul, penulis, tanggal pembuatan, dan format file.
Contoh: metadata pada foto digital berisi informasi tentang kamera, tanggal pengambilan, dan lokasi.

-Metadata Struktural:
Menjelaskan struktur data, seperti tabel, kolom, dan hubungan antar data.
Contoh: metadata dalam database relasional berisi informasi tentang tabel, kolom, dan kunci utama.

-Metadata Administratif:
Menyimpan informasi tentang pengelolaan data, seperti hak akses, riwayat modifikasi, dan catatan keamanan.
Contoh: metadata pada file elektronik berisi informasi tentang siapa yang membuat file, kapan file terakhir diubah, dan siapa yang memiliki akses ke file

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data21 = pd.read_csv("Downloads/data21.csv")
```

```python
data21.head()
```

```python
# Pisahkan data menjadi baris
lines = data21.splitlines()

# Ekstrak nama kolom dari baris pertama
column_names = lines[1].split()[2:]  

# Buat DataFrame kosong
df = pd.DataFrame(columns=column_names)

# Isi DataFrame 
for line in lines[1:]:
    values = line.split()[2:]
    row = pd.Series(values, index=column_names)
    df = df.append(row, ignore_index=True)

# kolom untuk visual
columns_to_visualize = ['POP10', 'POP20']

# grafik garis untuk data populasi
plt.figure(figsize=(10, 6))  
for col in columns_to_visualize:
    plt.plot(df['PROVINCE'], df[col], label=col)
plt.xlabel('Provinsi')
plt.ylabel('Populasi')
plt.title('Perbandingan Populasi (2010 vs. 2020)')
plt.xticks(rotation=45, ha='right')  
plt.legend()  
plt.grid(True)
plt.tight_layout()
plt.show()
```
