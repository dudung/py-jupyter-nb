```python
## JAWABAN NOMOR 2
Data:
Lebih berguna secara langsung: Menyajikan informasi mentah yang dapat dianalisis dan diolah untuk menghasilkan pengetahuan dan wawasan.
Contoh: Data penjualan, data pelanggan, data sensor, dll.

Metadata:
Meningkatkan nilai data: Memberikan informasi kontekstual tentang data, membuatnya lebih mudah dipahami, diinterpretasikan, dan dimanfaatkan.
Contoh: Jenis data, asal data, format data, hak akses, dll.

    Alasan Perlunya Metadata:
1. Organisasi:
Membantu mengkategorikan, mengelompokkan, dan mencari data dengan lebih mudah.
Contoh: Metadata dapat digunakan untuk mengelompokkan data penjualan berdasarkan wilayah, produk, atau periode waktu.

2. Dokumentasi:
Menyimpan informasi penting tentang data, seperti sumber, metode pengumpulan, dan makna.
Contoh: Metadata dapat mendokumentasikan format file, unit pengukuran, dan definisi variabel dalam dataset.

3. Interoperabilitas:
Memastikan kompatibilitas dan pertukaran data yang mudah antar sistem dan aplikasi.
Contoh: Metadata dapat mendefinisikan standar data, struktur, dan API untuk integrasi data yang mulus.

Jenis Metadata:
1. Metadata Deskriptif:
Menjelaskan isi dan struktur data (judul, penulis, tanggal pembuatan).

2. Metadata Teknis:
Memberikan informasi tentang format data, pengaturan file, dan spesifikasi teknis.

3. Metadata Administratif:
Menyimpan informasi tentang kepemilikan data, hak akses, dan catatan pengelolaan.
```

```python
## JAWABAN NOMOR 4
Reprodusibilitas dalam pemrograman mengacu pada kemampuan untuk menghasilkan hasil yang sama dengan menjalankan program yang sama di lingkungan yang berbeda. Hal ini sangat penting dalam ilmu data dan pengembangan perangkat lunak, karena memungkinkan kolaborasi, verifikasi, dan pengujian ulang.

Virtual environment di Python adalah alat yang membantu mencapai reprodusibilitas dengan mengisolasi dependensi program dan mencegah konflik versi. Berikut penjelasannya:

1. Isolasi Dependensi:
Virtual environment membuat lingkungan Python mandiri dengan salinan dependensi programnya sendiri (library, package, dll.).
Hal ini mencegah konflik versi antara dependensi program dan dependensi global yang digunakan oleh proyek lain.
Isolasi ini memastikan bahwa program selalu berjalan dengan dependensi yang sama, terlepas dari lingkungan sistem.

2. Kontrol Versi Dependensi:
Virtual environment memungkinkan Anda untuk menentukan versi spesifik dari setiap dependensi program.
Versi ini disimpan dalam file "requirements.txt" yang dapat dilacak dengan sistem kontrol versi (seperti Git).
Hal ini memungkinkan Anda untuk mereproduksi lingkungan program dengan tepat di mana saja, dengan dependensi yang sama di setiap versi.

3. Manfaat Reprodusibilitas:
Memastikan bahwa program berjalan dengan benar di berbagai lingkungan.
Memudahkan kolaborasi dengan memungkinkan tim untuk bekerja pada proyek yang sama dengan dependensi yang konsisten.
Memungkinkan pengujian ulang dan verifikasi hasil program.
Mendukung dokumentasi dan pemeliharaan program di masa depan.

4. Penggunaan Virtual Environment:
Untuk membuat virtual environment, gunakan perintah python3 -m venv <nama_virtual_environment>.
Aktifkan virtual environment dengan perintah source <nama_virtual_environment>/bin/activate.
Instal dependensi program dengan perintah pip install -r requirements.txt.
Jalankan program dengan perintah python <nama_program.py>
```

```python
## JAWABAN NOMOR 5
import pandas as pd
data = pd.read_csv('data14.csv')
```

```python
import matplotlib.pyplot as plt

##Menghitung perubahan populasi
data['POP_CHANGE'] = data['POP20'] - data['POP10']

##Membuat grafik batang perubahan populasi
plt.figure(figsize=(10, 6))
data.sort_values(by='POP_CHANGE', ascending=False, inplace=True)
plt.bar(data['PROVINCE'], data['POP_CHANGE'])
plt.xticks(rotation=90)  # Memutar label x-axis untuk keterbacaan
plt.xlabel('Province')
plt.ylabel('Change in Population')
plt.title('Population Change from 2010 to 2020')
plt.show()
```

```python
ei_columns = [f'EI{year}' for year in range(10, 19)]  # Menghasilkan list EI10, EI11, ..., EI18
data['AVG_EI'] = data[ei_columns].mean(axis=1)


plt.figure(figsize=(10, 6))
plt.plot(ei_columns, data[ei_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average Education Index')
plt.title('Average Education Index from 2010 to 2018')
plt.grid(True)
plt.show()
```

```python
gdp_columns = ['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']
plt.figure(figsize=(10, 6))
plt.plot(gdp_columns, data[gdp_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average GDP Growth Rate')
plt.title('Average GDP Growth Rate at Current Market Price (2020-2023)')
plt.grid(True)
plt.show()
```
