# 2. dari sisi nilai kemanfaatannya manakah yang lebih berguna, data atau metadata? lalu mengapa perlu ada metadata? jelaskan dengan singkat. berikan setidaknya tiga jenis metadata dan penjelasan singkatnya

data merupakan fakta dan angka mentah yang belum diolah. sedangkan metadata merupakan informasi tambahan tentang data itu sendiri.
data dan metadata memiliki kegunaan yang berbeda tetapi saling melengkapi, data berguna untuk melakukan analisis dan pengambilan keputusan, metadata berguna untuk mengelola dan memahami data. metadata memberikan informasi tentang data, seperti siapa yang membuatnya, kapan dibuatnya, apa artinya, dan bagaimana menggunakannya.
macam macam metadata: 
1. metadata deskriptif, memberikan informasi tentang isi data seperti judul, penulis, tanggal pembuatan, dan kata kunci.
2. metadata struktural, memberikan informasi mengenai struktur data, seperti format file, skema database, dan hubungan antar data.
3. metadata administratif, memberikan informasi tentang pengelolaan data, seperti hak akses, kebijakan penggunaan, dan riwayat versi.

# 4. reprodusibilitas suatu program yang kompleks dan menggunakan banyak library dan package versi tertentu dapat dipastikan dengan memanfaatkan fitur virtual environment di python. jelaskan mengenai hal itu.

Virtual environment di Python adalah fitur yang membantu memastikan reprodusibilitas program dengan menciptakan lingkungan terisolasi untuk menjalankan program tersebut. reprodusibilitas dalam pemrograman mengacu pada kemampuan untuk menjalankan program  yang sama dan mendapatkan hasil yang sama di berbagai lingkungan. hal ini penting untuk program yang kompleks, terutama yang menggunakan banyak library dan package dengan versi tertentu.
#manfaat menggunakan virtual environment
1. virtual environment memungkinkan untuk menginstall library dan package dengan versi tertentu tanpa memengaruhi versi yang terinstal di sistem yang ada. hal ini dapat memastikan program anda akan selalu berjalan dengan versi library yang diharapkan.
2. Virtual environment memungkinkan Anda untuk membuat lingkungan yang berbeda untuk setiap proyek. Hal ini membantu menghindari konflik antar library dan package yang digunakan oleh proyek yang berbeda.
3.  Virtual environment memungkinkan Anda untuk membagikan program Anda dengan orang lain beserta lingkungan yang diperlukan untuk menjalankannya. Hal ini membantu memastikan orang lain dapat menjalankan program Anda dengan hasil yang sama.


# 5. perhatikan data yang diberikan di https://github.com/dudung/mk-data-sains/tree/main/data dalam folder sesuai dengan id anda. buat grafik untuk data-data tersebut sesuai dengan sifat datanya dan jelaskan mengapa jenis grafik tertentu digunakan untuk data tertentu.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data04.csv')
```

```python
#grafik perbahan populasi
plt.figure(figsize=(10, 6))
data.sort_values(by='POP20', ascending=False, inplace=True)
plt.bar(data['PROVINCE'], data['POP20'])
plt.xticks(rotation=90)
plt.xlabel('Province')
plt.ylabel('Change in Population')
plt.title('Population Change From 2010 to 2020')
plt.show()
```

```python
# Menghasilkan list EI10, EI11, ..., EI18
ei_columns = [f'EI{year}' for year in range(10, 19)]  
data['AVG_EI'] = data[ei_columns].mean(axis=1)


plt.figure(figsize=(10, 6))
plt.plot(ei_columns, data[ei_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average Education Index')
plt.title('Average Education Index from 2010 to 2018')
plt.grid(True)
plt.show()
```

```python
# Data luas wilayah
luas_wilayah = {
    "Jawa Barat": [36923, 36923, 36923, 36923],
    "Jawa Tengah": [32852, 32852, 32852, 32852],
    "Jawa Timur": [47922, 47922, 47922, 47922],
}

# Tahun
tahun = [2020, 2021, 2022, 2023]

# Buat grafik batang untuk setiap provinsi
for provinsi, data in luas_wilayah.items():
    plt.bar(tahun, data, label=provinsi)
    plt.title(f"Luas Wilayah di {provinsi}")
    plt.xlabel("Tahun")
    plt.ylabel("Luas Wilayah (km)")
    plt.show()

# Buat grafik batang dengan semua provinsi
plt.bar(tahun, list(luas_wilayah.values())[0], label="Jawa Barat")
plt.bar(tahun, list(luas_wilayah.values())[1], label="Jawa Tengah")
plt.bar(tahun, list(luas_wilayah.values())[2], label="Jawa Timur")
plt.title("Luas Wilayah di semua Provinsi")
plt.xlabel("Tahun")
plt.ylabel("Luas Wilayah (km)")
plt.legend()
plt.show()
```
