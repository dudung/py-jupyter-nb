```python
#Nilai Kemanfaatan Data dan Metadata :
#    Data merupakan bahan baku yang penting untuk berbagai keperluan, seperti analisis, statistik, dan pengambilan keputusan.

#    Metadata memberikan informasi tentang data, seperti asal-usulnya, strukturnya, makna dan konteksnya.

#    Namun, data tanpa metadata akan sulit untuk dipahami, diinterpretasikan, dan dimanfaatkan secara maksimal.

#    sehingga dapat disimpulkan bahwa metadata lebih berguna karena dengan adanya metadata kita dapat memahami, mengintepretasikan dan memanfaatkan data secara maksimal atau mudahnya metadata ada untuk menjelaskan data.

#metadata sangat diperlukan karena :
#    Meningkatkan Pemahaman: Metadata membantu pengguna memahami data dengan lebih mudah dan cepat. Contohnya, metadata dapat menjelaskan arti kolom dalam tabel data, format tanggal dan waktu, dan unit pengukuran yang digunakan.
#    Mempermudah Pencarian: Metadata memungkinkan pengguna untuk mencari data yang mereka butuhkan dengan lebih mudah dan efisien. Contohnya, metadata dapat digunakan untuk mencari file gambar berdasarkan tanggal pengambilan, lokasi, atau kategori.
#    Meningkatkan Interoperabilitas: Metadata memungkinkan data dari berbagai sumber untuk diintegrasikan dan dikombinasikan dengan lebih mudah. Contohnya, metadata dapat digunakan untuk menyelaraskan format data dari berbagai database.
#    Meningkatkan Keamanan dan Keandalan: Metadata dapat membantu dalam mengelola akses ke data dan memastikan integritasnya. Contohnya, metadata dapat digunakan untuk melacak siapa yang telah mengakses data dan kapan data tersebut dimodifikasi.
#    Mempermudah Pengarsipan dan Pelestarian: Metadata dapat membantu dalam mengarsipkan dan melestarikan data untuk jangka panjang. Contohnya, metadata dapat digunakan untuk mendokumentasikan sejarah dan konteks data.

#3 Jenis Metadata dan Contohnya:
#    Metadata Deskriptif: Memberikan informasi tentang konten data, seperti judul, penulis, tanggal pembuatan, dan subjek. Contohnya, metadata deskriptif untuk file gambar dapat menyertakan nama file, tanggal pengambilan, lokasi pengambilan, dan jenis kamera yang digunakan.
#    Metadata Struktural: Memberikan informasi tentang struktur data, seperti format file, skema database, dan hubungan antar tabel. Contohnya, metadata struktural untuk database dapat menyertakan nama tabel, nama kolom, tipe data, dan kunci utama.
#    Metadata Administratif: Memberikan informasi tentang pengelolaan data, seperti hak akses, keamanan, dan riwayat perubahan. Contohnya, metadata administratif untuk file data dapat menyertakan nama pengguna yang memiliki akses ke file, tanggal modifikasi terakhir, dan catatan perubahan yang dilakukan.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data24.csv")
print(df)
```

# Grafik Bar (Bar Chart) untuk membandingkan penduduk pada tahun 2010 dan 2020 untuk setiap provinsi

```python
plt.figure(figsize=(10, 6))
plt.bar(df['PROVINCE'], df['POP10'], label='Population 2010', alpha=0.7)
plt.bar(df['PROVINCE'], df['POP20'], label='Population 2020', alpha=0.7)
plt.xlabel('Province')
plt.ylabel('Population')
plt.title('Population Comparison Between 2010 and 2020')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# Grafik Pie (Pie Chart) untuk membandingkan proporsi penduduk pada tahun 2020 di setiap provinsi

```python
plt.figure(figsize=(10, 6))
plt.pie(df['POP20'], labels=df['PROVINCE'], autopct='%1.1f%%')
plt.title('Population Proportion in 2020 by Province')
plt.show()
```

# Grafik Line (Line Chart) untuk menampilkan tren daya yang dihasilkan dari pembangkit listrik tenaga air dari tahun 2022 hingga 2023

```python
plt.figure(figsize=(10, 6))
plt.plot(df['PROVINCE'], df['MW22'], marker='o', label='MW 2022')
plt.plot(df['PROVINCE'], df['MW23'], marker='o', label='MW 2023')
plt.xlabel('Province')
plt.ylabel('Hydro Power (MW)')
plt.title('Hydro Power Generation Trend (2022-2023)')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

# Grafik Scatter (Scatter Plot) untuk menunjukkan hubungan antara GDP per kapita dan tingkat pengangguran pada tahun 2022

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['GDPACMP22'], df['UR22'], color='blue')
plt.xlabel('GDP per Capita (2022)')
plt.ylabel('Unemployment Rate (2022)')
plt.title('Scatter Plot of GDP per Capita vs Unemployment Rate (2022)')
plt.show()
```

# Grafik Scatter untuk menampilkan lokasi geografis provinsi berdasarkan longitude dan latitude

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['LONG'], df['LAT'], color='red')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geographical Locations of Provinces')
plt.show()
```

# Grafik Line untuk menampilkan tren pertumbuhan rata-rata GDPACMP dari tahun 2020 hingga 2023

```python
plt.figure(figsize=(10, 6))
avg_gdp = df[['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']].mean(axis=1)
plt.plot(df['PROVINCE'], avg_gdp, marker='o', color='green')
plt.xlabel('Province')
plt.ylabel('Average GDPACMP Growth Rate')
plt.title('Average GDPACMP Growth Rate (2020-2023)')
plt.xticks(rotation=90)
plt.show()
```

```python
#Scatter Plot: Digunakan untuk menampilkan hubungan antara dua variabel numerik, seperti Longitude dan Latitude. Scatter plot akan menunjukkan pola persebaran geografis dari setiap provinsi.
#Bar Chart: Cocok digunakan untuk membandingkan kategori-kategori berbeda, misalnya untuk membandingkan Unemployment Rate (UR) dari setiap provinsi pada tahun yang berbeda.
#Line Chart: Berguna untuk melihat tren atau perubahan dari suatu variabel numerik seiring waktu, seperti Growth Rate of Gross Domestic Product (GDPACMP) dari tahun 2020 hingga 2023.
#pie chart: digunakan untuk menunjukkan visualisasi pada proporsi penduduk pada tahun 2020 di setiap provinsi.

#Pemilihan jenis grafik tergantung pada informasi yang ingin disampaikan dan bagaimana cara terbaik untuk mempresentasikannya agar mudah dipahami oleh pembaca.
#Misalnya, Scatter Plot atau Map Plot cocok untuk menunjukkan lokasi geografis provinsi, sedangkan Bar Chart atau Line Chart dapat digunakan untuk membandingkan angka-angka statistik seperti tingkat pengangguran atau pertumbuhan ekonomi antar provinsi.
```
