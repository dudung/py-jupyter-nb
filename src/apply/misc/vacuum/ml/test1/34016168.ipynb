```python
## jawaban nomor 4
#Kompleksitas Program dan Reprodusibilitas:

#Program Python yang kompleks sering kali melibatkan banyak library dan package. Hal ini dapat menimbulkan masalah reprodusibilitas, di mana program tidak dapat dijalankan dengan hasil yang sama di lingkungan yang berbeda.

#Solusi: Virtual Environment:

#Virtual environment di Python menyediakan solusi untuk masalah ini. Virtual environment adalah direktori terisolasi di mana program Python dapat dijalankan dengan dependensi yang independen dari instalasi Python global.

#Manfaat Virtual Environment:

#Menjaga Keterpisahan: Virtual environment memisahkan dependensi program dari instalasi Python global, sehingga perubahan pada instalasi global tidak memengaruhi program.
#Memastikan Reprodusibilitas: Program dapat dijalankan dengan hasil yang sama di berbagai lingkungan dengan membuat virtual environment dan menginstal dependensi yang sama di dalamnya.
#Mengelola Dependensi: Virtual environment memungkinkan Anda untuk mengontrol versi spesifik dari library dan package yang digunakan program, sehingga program tidak terpengaruh oleh perubahan pada versi global.
#Cara Kerja Virtual Environment:

#Membuat Virtual Environment: Gunakan perintah virtualenv untuk membuat direktori virtual environment.
#Aktivasi Virtual Environment: Aktifkan virtual environment dengan perintah source bin/activate.
#Instal Dependensi: Instal library dan package yang dibutuhkan program di dalam virtual environment.
#Jalankan Program: Jalankan program Python dari dalam virtual environment.
```

```python
import pandas as pd
data = pd.read_csv('data18.csv')
```

```python
import matplotlib.pyplot as plt

##Menghitung perubahan populasi
data['POP_CHANGE'] = data['POP20'] - data['POP10']

##Membuat grafik batang perubahan populasi
plt.figure(figsize=(10, 6))
data.sort_values(by='POP_CHANGE', ascending=False, inplace=True)
plt.bar(data['PROVINCE'], data['POP_CHANGE'])
plt.xticks(rotation=90)  # Memutar label x-axis untuk keterbacaan
plt.xlabel('Province')
plt.ylabel('Change in Population')
plt.title('Population Change from 2010 to 2020')
plt.show()
```

```python
ei_columns = [f'EI{year}' for year in range(10, 19)]  # Menghasilkan list EI10, EI11, ..., EI18
data['AVG_EI'] = data[ei_columns].mean(axis=1)


plt.figure(figsize=(10, 6))
plt.plot(ei_columns, data[ei_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average Education Index')
plt.title('Average Education Index from 2010 to 2018')
plt.grid(True)
plt.show()
```

```python
gdp_columns = ['GDPACMP20', 'GDPACMP21', 'GDPACMP22', 'GDPACMP23']
plt.figure(figsize=(10, 6))
plt.plot(gdp_columns, data[gdp_columns].mean(), marker='o')
plt.xlabel('Year')
plt.ylabel('Average GDP Growth Rate')
plt.title('Average GDP Growth Rate at Current Market Price (2020-2023)')
plt.grid(True)
plt.show()
```
