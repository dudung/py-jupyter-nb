```python
##1
#gambarkan dalam suatu piramida tingkatan dan keberagaman antara data,informasi,pengetahuan,dan kearifan,serta jelaskan arah waktu dari masa lalu (pengalaman) dan masa depan (kebaruan)

#Piramida Tingkatan dan Keberagaman Data, Informasi, Pengetahuan, dan Kearifan
#Berikut adalah representasi piramida tingkatan dan keberagaman antara data, informasi, pengetahuan, dan kearifan, beserta arah waktu dari masa lalu (pengalaman) dan masa depan (kebaruan):

#Tingkatan:
#-Data: Fakta dan observasi mentah yang belum diolah. (Contoh: suhu, curah hujan, penjualan)
#-Informasi: Data yang telah diolah dan memiliki makna. (Contoh: rata-rata suhu, tren penjualan, prediksi cuaca)
#-Pengetahuan: Pemahaman yang lebih dalam tentang informasi dan hubungannya. (Contoh: faktor yang mempengaruhi suhu, strategi meningkatkan penjualan, dampak perubahan iklim)
#-Kearifan: Kemampuan untuk menerapkan pengetahuan secara praktis dan bijaksana. (Contoh: kebijakan untuk mengelola sumber daya alam, solusi inovatif untuk masalah sosial)

#Arah Waktu:
#-Masa Lalu (Pengalaman): Berasal dari data dan informasi yang dikumpulkan di masa lalu.
#-Masa Depan (Kebaruan): Ditujukan untuk menghasilkan pengetahuan dan kearifan yang bermanfaat untuk masa depan.
```

```python
##4
# Reprodusibilitas suatu program yang kompleks dan menggunakan banyak library dan package versi tertentu dapat dipastikan dengan memanfaatkan fitur virtual environment di python. Jelaskan mengenai hal ini

#Jawaban :

#Virtual environment merupakan fitur penting dalam Python yang membantu memastikan reprodusibilitas program yang kompleks dengan banyak library dan package.
#Reprodusibilitas berarti program dapat dijalankan pada komputer lain dengan hasil yang sama, tanpa modifikasi. Hal ini penting untuk:
##-Memastikan keandalan dan konsistensi program.
##-Memudahkan kolaborasi dengan tim lain.
##-Mempermudah deployment program ke production environment.

#Virtual environment bekerja dengan:
#-Membuat folder terisolasi yang berisi library dan package yang dibutuhkan program.
#-Menginstal library dan package dengan versi spesifik di dalam virtual environment.
#-Mengisolasi program dan dependensi dari environment Python global.

#-Manfaat Virtual Environment:
#-Reprodusibilitas: Program dapat dijalankan pada komputer lain dengan hasil yang sama.
#-Keandalan: Program terhindar dari konflik library dan package yang berbeda versi.
#-Keteraturan: Memudahkan pengelolaan library dan package yang berbeda untuk proyek yang berbeda.
#-Keamanan: Meminimalisir risiko keamanan dan bug dari library dan package pihak ketiga.
```

```python
## 5 

import pandas as pd
import matplotlib.pyplot as plt

# Membaca data dari file CSV
data = pd.read_csv("data17.csv")

# Menghitung persentase perubahan populasi
data["Perubahan Populasi"] = (data["POP20"] - data["POP10"]) / data["POP10"] * 100

# Mengurutkan data berdasarkan persentase perubahan populasi
data = data.sort_values("Perubahan Populasi", ascending=False)

# Memvisualisasikan perbandingan populasi
plt.bar(data["PROVINCE"], data["POP10"], label="Populasi 2010")
plt.bar(data["PROVINCE"], data["POP20"], bottom=data["POP10"], label="Populasi 2020")
plt.xlabel("Provinsi")
plt.ylabel("Jumlah Penduduk")
plt.legend()
plt.show()

# Menampilkan tabel data
print(data.to_string())
```
