```python
# # 2

# keduanya sama penting nya, dari sisi kemanfaatannya tentu data lebih berguna, akan percuma jika kita memiliki metadata dari sesuatu yang tidak ada (data nya tidak kita miliki), 
# sejatinya metadata digunakan untuk mempermudah kita mengidentifikasi data tersebut
# data adalah informasi yang belum di olah, sekumpulan fakta dan statistik yang akan berguna setelah diolah menjadi informasi, data bisa berbentuk sebuah gambar, audio, angka, teks
# metadata adalah informasi tentang data itu sendiri, data about data, data yang digunakan untuk mempermudah mengatur data tersebut dalam managerial
# kita bisa menganalogikan data sebagai sebuah isi dan metadata sebagai label dari isi tersebut
# lalu mengapa perlu ada metadata? metadata diperlukan untuk beberapa hal antara lain:
  # untuk mengidentifikasi data
  # untuk menemukan data dengan cepat
  # untuk memahami data
  # untuk mempermudah pengelolaan data
  # untuk memastikan originalitas suatu data

# metadata terdiri dari beberapa jenis antara lain:
  # descriptive metadata : memberikan informasi tentang isi data / judul, penulis, genre,
  # Structural Metadata : memberikan informasi tentang struktur data / daftar isi, bab sub-bab, urutan presentasi
  # Administrative metadata : memberikan informasi tentang pengelolaan data / hak cipta, izin akses, tanggal dibuat dan diedit
```

```python
# # 5

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data08.csv")
print(df)
```

```python
# penduduk tahun 2010
y = df["PROVINCE"]
x = df["POP10"]

plt.bar(x, y)

plt.title("Populasi di Indonesia pada tahun 2010")
plt.xlabel("Populasi (Jiwa)")
plt.ylabel("Provinsi")

plt.show()
```

```python
# penduduk tahun 2020
y = df["PROVINCE"]
x = df["POP10"]

plt.bar(x, y, linewidth=1)
plt.xlim(0, 100000000)

plt.title("Populasi di Indonesia pada tahun 2020")
plt.xlabel("Populasi (Jiwa)")
plt.ylabel("Provinsi")

plt.show()
```
