```python
import random as rnd
import matplotlib.pyplot as plt
```

```python
N = 100
a = 2
b = 8
x = []
for i in range(N):
    j = rnd.randint(a, b)
    x.append(j)
    print(j, end=' ')
print('\n\n')

print(len(x))
```

```python
bins = (b - a) + 1

plt.figure(figsize=(4, 3))
plt.hist(x, range=[a, b + 1], bins=bins)
plt.xlabel('x')
plt.ylabel('N')
plt.xticks(range(a, b + 2))
plt.yticks(range(0, 26, 2))
plt.grid()
plt.show()
```

```python
def f(x):
    y = (x - 3)*(x - 5)*(x - 7)
    return y

xx = [*range(a, b + 1)]
y = [f(i) for i in xx]

plt.figure(figsize=(4, 3))
plt.plot(xx, y, 'o-')
plt.xticks(range(a, b + 2))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
rx = rnd.randint(a, b)
ry = f(rx)
xr = [a, b]
yr = [ry, ry]
c = '#cfc'
y2 = [0, 0]

plt.figure(figsize=(4, 3))
plt.plot(xx, y)
plt.plot(rx, ry, 'o')
plt.plot(xr, yr, '-')
if yr[0] > 0:
    plt.fill_between(xr, yr, y2, color=c)
else:
    plt.fill_between(xr, y2, yr, color=c)
plt.xticks(range(a, b + 2))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
c = '#cfc'

plt.figure(figsize=(5, 5))
for i in range(6):
    rx = rnd.randint(a, b)
    ry = f(rx)
    xr = [a, b]
    yr = [ry, ry]
    y2 = [0, 0]
    
    plt.subplot(2, 3, i+1)
    plt.plot(xx, y)
    plt.plot(rx, ry, 'o')
    plt.plot(xr, yr, '-')

    if yr[0] > 0:
        plt.fill_between(xr, yr, y2, color=c)
    else:
        plt.fill_between(xr, y2, yr, color=c)
    
    plt.xticks(range(a, b + 2))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
plt.show()
```

$$
\begin{array}{rcl}
\displaystyle \int_2^8 (x - 3)(x - 5)(x - 7) \ \ dx & = & \displaystyle \int_2^8 (x^2 - 8x + 15)(x - 7) \ \ dx \newline
& = & \displaystyle \int_2^8 (x^3 - 8x^2 + 15x) - (7x^2 - 56x + 105)\ \ dx \newline
& = & \displaystyle \int_2^8 (x^3 - 15x^2 + 71x - 105)\ \ dx \newline
& = & \displaystyle \left[ \tfrac14 x^4 - 5x^3 + 35\tfrac12 x^2 - 105x \right]_{x = 3}^{8} \newline
& = & \displaystyle (\tfrac14 \cdot 8^4 - 5 \cdot 8^3 + 35\tfrac12 \cdot 8^2 - 105 \cdot 8) \newline
&&- (\tfrac14 \cdot 2^4 - 5 \cdot 2^3 + 35\tfrac12 \cdot 2^2 - 105 \cdot 2) \newline
& = & (1024 - 2560 + 2272 - 840) - (4 - 40 + 142 - 210) \newline
& = & -104 - (-104)\newline
& = & 0
\end{array}
$$

```python
N2 = 1000
n = []
A = []
AA = 0
for i in range(1, N2 + 1):
    xi = rnd.randint(a, b)
    fi = f(xi)
    Ai = (b - a) * fi
    
    n.append(i)
    if i == 1:
        AA = Ai
    else:
        AA = (A[i-2] * (i-1) + Ai) / i
    A.append(AA)

plt.plot([0, N2], [0, 0], 'r-')
plt.plot(n, A, '-')
plt.xlabel('n')
plt.ylabel('A')
plt.grid()
plt.show()
```
