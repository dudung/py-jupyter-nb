```python
import random as rnd
import matplotlib.pyplot as plt
```

```python
N = 100
a = 3
b = 8
x = []
for i in range(N):
    j = rnd.randint(a, b)
    x.append(j)
    print(j, end=' ')
print('\n\n')

print(len(x))
```

```python
bins = (b - a) + 1

plt.figure(figsize=(4, 3))
plt.hist(x, range=[a, b + 1], bins=bins)
plt.xlabel('x')
plt.ylabel('N')
plt.xticks(range(a, b + 2))
plt.yticks(range(0, 26, 2))
plt.grid()
plt.show()
```

```python
def f(x):
    y = 4*(x - 3) + 2
    return y

y = [f(i) for i in x]

plt.figure(figsize=(4, 3))
plt.plot(x, y)
plt.xticks(range(a, b + 2))
plt.yticks(range(0, 26, 4))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
rx = rnd.randint(a, b)
ry = f(rx)
xr = [a, b]
yr = [ry, ry]

plt.figure(figsize=(4, 3))
plt.plot(x, y)
plt.plot(rx, ry, 'o')
plt.plot(xr, yr, '-')
plt.xticks(range(a, b + 2))
plt.yticks(range(0, 26, 4))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
