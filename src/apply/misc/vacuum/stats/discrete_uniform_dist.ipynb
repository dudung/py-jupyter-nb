```python
import statistics as stat
import random as rnd

N = 1000; a = 20; b = 60; x = []
for i in range(N):
    xi = rnd.randint(a, b)
    x.append(xi)

print('N =', len(x))
print('xmin =', min(x))
print('xmax =', max(x))
print('mean =', stat.mean(x))
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.hist(x, bins=10)
plt.xticks(np.arange(20, 64, 4))
plt.xlim([20, 60])
plt.grid()
plt.show()
```
