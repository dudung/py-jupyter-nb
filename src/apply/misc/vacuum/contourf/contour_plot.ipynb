```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(X**2 + Y**2)
```

```python
plt.figure(figsize=(6, 6))
plt.grid()
plt.contourf(X, Y, Z)
plt.colorbar()
```

```python
plt.figure(figsize=(6, 6))
plt.grid()
plt.contourf(X, Y, Z)
plt.colorbar()
```

```python
x = np.linspace(-np.pi, np.pi, 20)
y = np.linspace(-np.pi, np.pi, 20)
X, Y = np.meshgrid(x, y)
Z = np.sin(X)
```

```python
plt.figure(figsize=(6, 6))
plt.grid()
plt.contourf(X, Y, Z)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)
Z1 = np.sqrt((X-0.5)**2 + (Y-0.5)**2)

plt.figure(figsize=(6, 6))
plt.grid()
plt.contourf(X, Y, Z1)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)
Z2 = np.sqrt((X+0.5)**2 + (Y+0.5)**2)

plt.figure(figsize=(6, 6))
plt.grid()
plt.contourf(X, Y, Z2)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
Z = Z1 + Z2

plt.figure(figsize=(6, 6))
plt.grid()
plt.contourf(X, Y, Z)
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
