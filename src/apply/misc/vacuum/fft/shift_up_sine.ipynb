```python
import math
import numpy as np
import pandas as pd
import scipy as sc
import matplotlib.pyplot as plt
```

```python
N = 100
tbeg = 0
tend = 1
dt = (tend - tbeg) / N
t = [tbeg + i*dt for i in range(N)]
print(np.round(t, 2))
```

```python
A1 = 4
f1 = 2
w1 = 2 * math.pi * f1
b1 = 0
x1 = [A1 * math.sin(w1 * i) + b1 for i in t]

plt.plot(t, x1, 'tab:red')
plt.xlabel('t')
plt.ylabel('x1')
plt.grid()
plt.show()
```

```python
A2 = 4
f2 = 2
w2 = 2 * math.pi * f2
b2 = 2
x2 = [A2 * math.sin(w2 * i) + b2 for i in t]

plt.plot(t, x2, 'tab:blue')
plt.xlabel('t')
plt.ylabel('x2')
plt.grid()
plt.show()
```

```python
Af1 = np.abs(sc.fft.fft(x1))
ff1 = sc.fft.fftfreq(N, dt)

plt.xticks(ff1[0:N//5])
plt.stem(ff1[0:N//5], Af1[0:N//5],
  linefmt='tab:red',
  basefmt='tab:red', markerfmt='none')
plt.grid()
plt.xlabel('f')
plt.ylabel('x1')
plt.show()
```

```python
Af2 = np.abs(sc.fft.fft(x2))
ff2 = sc.fft.fftfreq(N, dt)

plt.xticks(ff2[0:N//5])
plt.stem(ff2[0:N//5], Af2[0:N//5],
  linefmt='tab:blue',
  basefmt='tab:blue', markerfmt='none')
plt.grid()
plt.xlabel('f')
plt.ylabel('x2')
plt.show()
```

```python
x3 = []
for i in range(N):
    xi = x1[i] + x2[i]
    x3.append(xi)

plt.plot(t, x3, 'tab:green')
plt.xlabel('t')
plt.ylabel('x3')
plt.grid()
plt.show()
```

```python
Af3 = np.abs(sc.fft.fft(x3))
ff3 = sc.fft.fftfreq(N, dt)

plt.xticks(ff3[0:N//5])
plt.stem(ff3[0:N//5], Af3[0:N//5],
  linefmt='tab:green',
  basefmt='tab:green', markerfmt='none')
plt.grid()
plt.xlabel('f')
plt.ylabel('x3')
plt.show()
```
