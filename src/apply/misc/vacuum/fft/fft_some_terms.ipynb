```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sc
```

```python
N = 1000
tbeg = 0
tend = 2
dt = (tend - tbeg) / N
t = [tbeg + i*dt for i in range(N)]

for i in t:
    print(f'{i:.3f}', end=', ')
```

```python
A1 = 2
f1 = 10
w1 = 2 * math.pi * f1 
x1 = [A1 * math.sin(w1 * i) for i in t]

plt.plot(t[:100], x1[:100], 'tab:red')
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

```python
A2 = 1
f2 = 20
w2 = 2 * math.pi * f2 
x2 = [A2 * math.sin(w2 * i) for i in t]

plt.plot(t[:100], x2[:100], 'tab:green')
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

```python
A3 = 0.5
f3 = 40
w3 = 2 * math.pi * f3 
x3 = [A3 * math.sin(w3 * i) for i in t]

plt.plot(t[:100], x3[:100], 'tab:blue')
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

```python
x = []
for i in range(N):
    xi = x1[i] + x2[i] + x3[i]
    x.append(xi)

plt.plot(t[:100], x[:100], 'tab:orange')
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.plot(t[:100], x1[:100], 'tab:red')
plt.ylabel('x')
plt.subplot(2, 2, 2)
plt.plot(t[:100], x2[:100], 'tab:green')
plt.subplot(2, 2, 3)
plt.plot(t[:100], x3[:100], 'tab:blue')
plt.xlabel('t'); plt.ylabel('x')
plt.subplot(2, 2, 4)
plt.plot(t[:100], x[:100], 'tab:orange')
plt.xlabel('t')
plt.show()
```

```python
y = np.abs(sc.fft.fft(x))
f = sc.fft.fftfreq(N, dt)

plt.subplot(2, 1, 1)
plt.stem(f[0:N//2], y[0:N//2],
  basefmt='tab:blue', markerfmt='none')

plt.subplot(2, 1, 2)
plt.stem(f[0:N//10], y[0:N//10],
  basefmt='tab:blue', markerfmt='none')
plt.xlabel('f')
plt.show()
```
