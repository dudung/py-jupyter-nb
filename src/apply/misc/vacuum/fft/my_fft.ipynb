```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
tbeg = 0
tend = 2
N = 40
dt = (tend - tbeg) / N
print('dt =', dt)
print()

t = np.linspace(tbeg, tend, N, endpoint=False)
print('t =')
print(t)
```

```python
A = 2
f = 1
omega = 2 * math.pi * f
x = [1 + A * math.sin(omega * i) for i in t]
print(np.round(x, 3))
```

```python
plt.plot(t, x, 'o')
plt.grid()
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```

```python
yre = []

for f in range(N):
    omega1 = 2 * math.pi * f / N
    sre = 0
    
    for id, el in enumerate(x):
        re = el * math.cos(id * omega1)
        sre += re
    
    yre.append(sre)
    
print(yre)
```

```python
yim = []

for f in range(N):
    omega1 = 2 * math.pi * f / N
    sim = 0
    
    for id, el in enumerate(x):
        im = -el * math.sin(id * omega1)
        sim += im
    
    yim.append(sim)
    
print(np.round(yim))
```

```python
yabs = []
for i in range(N):
    abs = math.sqrt(yre[i]**2 + yim[i]**2) / N
    yabs.append(abs)

print(np.round(yabs))
```

```python
srate = 1 / dt
T = N / srate
freq = [i/T for i in range(N)]
print(freq)
```

```python
if N // 2 == N /2:
    M = round(N / 2 - 1)
else:
    M = round((N - 1) / 2)

plt.stem(freq[0:M], yabs[0:M], markerfmt='')
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.show()
```
