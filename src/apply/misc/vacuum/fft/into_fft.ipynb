```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
A = 1
T = 2; omega = 2 * math.pi / T
tbeg = 0; tend = 2;
N = 10; dt = (tend - tbeg) / N
x1 = []
t1 = []
for i in range(N):
    ti = tbeg + i * dt
    t1.append(ti)
    xi = A * math.sin(omega * ti)
    x1.append(xi)
tticks = t1.copy();
tticks.append(t1[-1] + dt)
print(x1)
```

```python
plt.scatter(t1, x1)
plt.xticks(tticks)
plt.xlim([tbeg, tend])
plt.grid()
plt.show()
```
