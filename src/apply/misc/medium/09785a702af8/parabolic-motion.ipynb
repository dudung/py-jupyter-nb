```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# time
ti = 0
tf = 8
t = np.arange(ti, tf + 1, 0.5)

# non-uniform liear motion in vertical direction
ay = -10
v0y = 40
y0 = 0
y = y0 + v0y *t + 0.5 * ay * t**2

# uniform linear motion in horzontal direction
ax = 0
v0x = 30
x0 = 0
x = x0 + v0x *t + 0.5 * ax * t**2
```

```python
plt.scatter(t, x, label='motion along x', marker='o', color='tab:blue')
plt.xlabel('t (s)');
plt.ylabel('x (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(t, y, label='free fall', marker='v', color='tab:red')
plt.xlabel('t (s)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, label='parabolic motion', marker='s', color='tab:green')
plt.xlabel('x (m)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2,2,1)
plt.scatter(t, y, label='free fall', marker='v', color='tab:red')
plt.xlabel('t (s)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()

plt.subplot(2,2,4)
plt.scatter(x, t, label='motion along x', marker='o', color='tab:blue')
plt.ylabel('t (s)');
plt.xlabel('x (m)');
plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.scatter(x, y, label='parabolic motion', marker='s', color='tab:green')
plt.grid()
plt.legend()

plt.show()
```
