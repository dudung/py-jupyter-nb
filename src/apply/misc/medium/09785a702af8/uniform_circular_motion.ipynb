```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# time
ti = 0
tf = 2
dt = 0.1
t = np.arange(ti, tf + dt, dt)

# radius, period, frequency, angular velocity
R = 2
T = 2
f = 1 / T
omega = 2 * np.pi * f

# simple harmonic motion in x direction
x = R * np.cos(omega * t)

# uniform linear motion in horzontal direction
y = R * np.sin(omega * t)
```

```python
plt.scatter(t, x, label='motion along x', marker='o', color='tab:blue')
plt.xlabel('t (s)');
plt.ylabel('x (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(t, y, label='free fall', marker='v', color='tab:red')
plt.xlabel('t (s)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.scatter(x, y, label='parabolic motion', marker='s', color='tab:green')
plt.xlabel('x (m)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 8))

plt.subplot(2,2,1)
plt.scatter(t, y, label='free fall', marker='v', color='tab:red')
plt.xlabel('t (s)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()

plt.subplot(2,2,4)
plt.scatter(x, t, label='motion along x', marker='o', color='tab:blue')
plt.ylabel('t (s)');
plt.xlabel('x (m)');
plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.scatter(x, y, label='parabolic motion', marker='s', color='tab:green')
plt.grid()
plt.legend()

plt.show()
```
