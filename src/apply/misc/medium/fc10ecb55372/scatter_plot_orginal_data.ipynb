```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf("ibnas_c04_20_new.txt")
print(df.head())
print(df.tail())
```

```python
plt.plot(df['DATE'], df['X'], '.')
plt.xlabel('t')
plt.ylabel('x1')
plt.xlim([47500,60500])
plt.ylim([-1.5,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['DATE'], df['Y'], '.')
plt.xlabel('t')
plt.ylabel('x2')
plt.xlim([47500,60500])
plt.ylim([-1.5,1.5])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(df['X'], df['Y'], '.')
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim([-1.5,1.5])
plt.ylim([-1.5,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['X'], df['sigmaX'], '.')
plt.xlabel('x1')
plt.ylabel('σ1')
plt.xlim([-1.5,1.5])
plt.ylim([0,5])
plt.grid()
plt.show()
```

```python
plt.plot(df['Y'], df['sigmaY'], '.')
plt.xlabel('x2')
plt.ylabel('σ2')
plt.xlim([-1.5,1.5])
plt.ylim([0,5])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(df['sigmaX'], df['sigmaY'], '.')
plt.xlabel('σ1')
plt.ylabel('σ2')
plt.xlim([0,5])
plt.ylim([0,5])
plt.grid()
plt.show()
```

```python
import math
N = len(df)
r = []
q = []
for i in range(N):
    xi = df['X'][i]
    yi = df['Y'][i]
    
    ri = math.sqrt(xi**2 + yi**2)
    r.append(ri)
    
    qi = math.atan2(yi,xi)
    q.append(qi)
```

```python
plt.plot(q, r, '.')
plt.xlabel('θ')
plt.ylabel('r')
plt.xlim([-3.5,3.5])
plt.ylim([0,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['DATE'], r, '.')
plt.xlabel('t')
plt.ylabel('r')
plt.xlim([47500,60500])
plt.ylim([0,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['DATE'], q, '.')
plt.xlabel('t')
plt.ylabel('θ')
plt.xlim([47500,60500])
plt.ylim([-3.5,3.5])
plt.grid()
plt.show()
```
