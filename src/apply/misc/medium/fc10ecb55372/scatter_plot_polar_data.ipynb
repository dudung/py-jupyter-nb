```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf("ibnas_c04_20_new.txt")
print(df.head())
print(df.tail())
```

```python
import math
N = len(df)
r = []
q = []
for i in range(N):
    xi = df['X'][i]
    yi = df['Y'][i]
    
    ri = math.sqrt(xi**2 + yi**2)
    r.append(ri)
    
    qi = math.atan2(yi,xi)
    q.append(qi)
```

```python
plt.plot(q, r, '.')
plt.xlabel('θ')
plt.ylabel('r')
plt.xlim([-3.5,3.5])
plt.ylim([0,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['DATE'], r, '.')
plt.xlabel('t')
plt.ylabel('r')
plt.xlim([47500,60500])
plt.ylim([0,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['DATE'], q, '.')
plt.xlabel('t')
plt.ylabel('θ')
plt.xlim([47500,60500])
plt.ylim([-3.5,3.5])
plt.grid()
plt.show()
```

```python
plt.hist(r, bins=100)
plt.xlabel('r')
plt.ylabel('N')
plt.xlim([0,1.5])
plt.ylim([0,700])
plt.grid()
plt.show()
```
