```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_fwf("ibnas_c04_20_new.txt")
print(df.head())
print(df.tail())
```

```python
plt.plot(df['DATE'], df['X'], '.')
plt.xlabel('t')
plt.ylabel('x1')
plt.xlim([47500,60500])
plt.ylim([-1.5,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['DATE'], df['Y'], '.')
plt.xlabel('t')
plt.ylabel('x2')
plt.xlim([47500,60500])
plt.ylim([-1.5,1.5])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(df['X'], df['Y'], '.')
plt.xlabel('x1')
plt.ylabel('x2')
plt.xlim([-1.5,1.5])
plt.ylim([-1.5,1.5])
plt.grid()
plt.show()
```

```python
plt.plot(df['X'], df['sigmaX'], '.')
plt.xlabel('x1')
plt.ylabel('1')
plt.xlim([-1.5,1.5])
plt.ylim([0,5])
plt.grid()
plt.show()
```

```python
plt.plot(df['Y'], df['sigmaY'], '.')
plt.xlabel('x2')
plt.ylabel('2')
plt.xlim([-1.5,1.5])
plt.ylim([0,5])
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(4.8, 4.8))
plt.plot(df['sigmaX'], df['sigmaY'], '.')
plt.xlabel('1')
plt.ylabel('2')
plt.xlim([0,5])
plt.ylim([0,5])
plt.grid()
plt.show()
```

```python
plt.hist(df['Y'], bins=100)
plt.xlabel('x2')
plt.ylabel('N')
plt.xlim([-1.5,1.5])
plt.ylim([0,700])
plt.grid()
plt.show()
```
