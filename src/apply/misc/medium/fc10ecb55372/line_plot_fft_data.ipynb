```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import rfft, rfftfreq
```

```python
df = pd.read_fwf("ibnas_c04_20_new.txt")
print(df.head())
print(df.tail())
```

```python
import math
N = len(df)
x1 = []
x2 = []
r = []
q = []
for i in range(N):
    xi = df['X'][i]
    yi = df['Y'][i]
    
    x1.append(xi)
    x2.append(yi)
    
    ri = math.sqrt(xi**2 + yi**2)
    r.append(ri)
    
    qi = math.atan2(yi,xi)
    q.append(qi)
```

```python
y1 = np.abs(rfft(x1))
#plt.yscale('log')
plt.plot(y1, 'r-', y1, '.b')
plt.xlabel('f')
plt.ylabel('x1')
plt.xlim([-50,1000])
plt.ylim([0,1000])
plt.grid()
plt.show()
```

```python
y2 = np.abs(rfft(x2))
#plt.yscale('log')
plt.plot(y2, 'r-', y2, '.b')
plt.xlabel('f')
plt.ylabel('x2')
plt.xlim([-50,1000])
plt.ylim([0,1000])
plt.grid()
plt.show()
```

```python
y3 = np.abs(rfft(r))
#plt.yscale('log')
plt.plot(y3, 'r-', y3, '.b')
plt.xlabel('f')
plt.ylabel('r')
plt.xlim([-50,1000])
plt.ylim([0,4000])
plt.grid()
plt.show()
```

```python
y4 = np.abs(rfft(q))
#plt.yscale('log')
plt.plot(y4, 'r-', y4, '.b')
plt.xlabel('f')
plt.ylabel('θ')
plt.xlim([-50,1000])
plt.ylim([0,6000])
plt.grid()
plt.show()
```
