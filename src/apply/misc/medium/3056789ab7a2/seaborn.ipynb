```python
#Import Libraries
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

#Import Dataset
penguins = sns.load_dataset('penguins')

# Pairplot with hues
sns.pairplot(data=penguins, hue='species');
```

```python
#Simple Pairplot
sns.pairplot(data=penguins, corner=True);
```
