```python
# parameters
x0 = [5, 0, -5]
v = [-1, 0, 1]
ti = 0
tf = 10
t = [*range(ti, tf+1)]
```

```python
# datasets
xx = []
for i in x0:
    for j in v:
        x = []
        for k in t:
            l = i + j * k
            x.append(l)
        xx.append(x)
```

```python
# graph with suplots
import matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(9, 6))
for i, x in enumerate(xx):
    plt.subplot(len(x0), len(v), i+1)
    plt.xticks([*range(0, tf+1, 2)])
    plt.grid()
    j = i % len(v)
    k = i // len(v)
    text = f'x₀={x0[k]}, v={v[j]}'
    plt.plot(t, x, '.--', label=text)
    if i==3:
        plt.ylabel('x')
    if i==7:
        plt.xlabel('t')
    plt.legend()

plt.show()
```
