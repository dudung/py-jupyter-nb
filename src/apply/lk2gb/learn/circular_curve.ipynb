```python
import numpy as np
import random as rnd
import matplotlib.pyplot as plt
```

```python
mat = np.zeros((100, 100))
for i in range(100):
    for j in range(100):
        dx2 = (j - 50)**2
        dy2 = (i - 50)**2
        dr2 = dx2 + dy2
        if 1560 < dr2 and dr2 < 1640:
            mat[i][j] = 1
```
