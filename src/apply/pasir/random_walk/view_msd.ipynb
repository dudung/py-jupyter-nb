```python
import ipynb.fs.full.walk as walk
import matplotlib.pyplot as plt

N = 100; ndir = 4; rep = 100
dirs = walk.discrete_dirs(ndir, N)

ac = walk.estimated_autocorrelation(dirs)

print('max autocorrelation [1:n/2] ='
      , f'{max(ac[1:N//2]):0.4f}')

msd = walk.msd_shuffle_dirs(dirs, rep)

plt.figure(figsize=(4,4))
plt.plot(msd, '-')
plt.xlabel('i'); plt.ylabel('s')
plt.grid(); plt.show()
plt.show()
```

```python
import ipynb.fs.full.walk as walk
import matplotlib.pyplot as plt

N = 100; ndir = 3; rep = 10000
dirs = walk.discrete_dirs(ndir, N)

ac = walk.estimated_autocorrelation(dirs)

print('max autocorrelation [1:n/2] ='
      , f'{max(ac[1:N//2]):0.4f}')

msd = walk.mad_shuffle_dirs(dirs, rep)

plt.figure(figsize=(4,4))
plt.plot(msd, '.')
plt.xlabel('i'); plt.ylabel('s')
plt.grid(); plt.show()
plt.show()
```

```python
import ipynb.fs.full.walk as walk
import matplotlib.pyplot as plt
import random as rnd

N = 100; ndir = 3
dirs = walk.discrete_dirs(ndir, N)

plt.figure(figsize=(6,6))
for i in range(9):
    rnd.shuffle(dirs)
    ac = walk.estimated_autocorrelation(dirs)
    acs = f'{max(ac[1:N//2]):0.4f}'

    plt.subplot(3, 3, i+1)
    plt.title(acs)
    msd = walk.mad_shuffle_dirs(dirs, rep)
    plt.plot(msd, '.')
    if i < 8:
        plt.axis('off')
plt.show()
```
