# view walk with step variations

```python
import ipynb.fs.full.walk as walk
import matplotlib.pyplot as plt

N = 10; D = 4
num = walk.discrete_dirs(D, N)
x, y = walk.move_dirs(0, 0, 1, num)

plt.figure(figsize=(4,4))
plt.plot(x, y, '-')
plt.plot(x[0], y[0], 'r>')
plt.plot(x[-1], y[-1], 'ro')
plt.gca().set_aspect('equal')
plt.legend(['', 'start', 'stop'])
plt.xlabel('x'); plt.ylabel('y')
plt.grid(); plt.show()
```
