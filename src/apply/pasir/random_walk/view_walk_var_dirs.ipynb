# view walk with direction variations

```python
import ipynb.fs.full.walk as walk
import matplotlib.pyplot as plt

N = 100; D = 1
num = walk.discrete_dirs(D, N)
x, y = walk.move_dirs(0, 0, 1, num)

plt.figure(figsize=(4,4))
plt.plot(x, y, '-')
plt.plot(x[0], y[0], 'g>')
plt.plot(x[-1], y[-1], 'ro')
plt.gca().set_aspect('equal')
plt.legend(['', 'start', 'stop'])
plt.xlabel('x'); plt.ylabel('y')
plt.grid(); plt.show()
```
