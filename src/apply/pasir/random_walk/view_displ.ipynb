# view displacement

```python
import ipynb.fs.full.walk as walk
import matplotlib.pyplot as plt

N = 100; D = 4
num = walk.discrete_dirs(D, N)
x, y = walk.move_dirs(0, 0, 1, num)
s = walk.displ_from_coords(x, y)

plt.figure(figsize=(4,4))
plt.plot(s, '-')
plt.xlabel('i'); plt.ylabel('s')
plt.grid(); plt.show()
```
