# view available datasets

```python
import matplotlib.pyplot as plt
import ipynb.fs.full.datasets as ds

styles = ['s','o']
colors = ['b', 'r']
plt.figure(figsize=(4, 4))
x = ds.dataset(1)

for i in range(len(x[0])):
    xx = x[0][i]
    yy = x[1][i]
    cc = colors[x[2][i]]
    mm = styles[x[2][i]]
    plt.scatter(xx, yy, c=cc, marker=mm)

plt.grid()
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```
