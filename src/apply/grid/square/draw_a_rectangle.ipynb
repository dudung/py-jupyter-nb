# draw_a_rectangle
Mix sinewaves as beats and save as wav file (see full feature in [nbviewer](https://nbviewer.org/github/dudung/py-jupyter-nb/blob/main/src/fft/beats.ipynb))

```python
# import necessary libraries
import numpy as np
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
from IPython.display import Audio
```

```python
# define default parameters
sample_rate = 44100 # Hz
int16max = 32767 # 16-bit int -32768 -- 32767
duration = 5 # s
```

```python
# create 1st signal
freq = 400 # Hz
fname = 'data/f1_400hz.wav'
_, y = gen.sine_wave(freq, sample_rate, duration)

# normalize amplitude for wav file
yn = np.int16((y / y.max()) * int16max)

# write to file with 44100 Hz playback rate
write(fname, sample_rate, yn)

# show player
Audio(fname, autoplay=False)
```

```python
# create 2nd signal
freq = 403 # Hz
fname = 'data/f2_403hz.wav'
_, y = gen.sine_wave(freq, sample_rate, duration)

# normalize amplitude for wav file
yn = np.int16((y / y.max()) * int16max)

# write to file with 44100 Hz playback rate
write(fname, sample_rate, yn)

# show player
Audio(fname, autoplay=False)
```

```python
# mix signals -- information + noise
freq1 = 400 # Hz
freq2 = 403 # Hz
fname = 'data/fmix_f1_f2.wav'
_, y1 = gen.sine_wave(freq1, sample_rate, duration)
_, y2 = gen.sine_wave(freq2, sample_rate, duration)
y = y1 + y2

# normalize amplitude for wav file
yn = np.int16((y / y.max()) * int16max)

# write to file with 44100 Hz playback rate
write(fname, sample_rate, yn)

# show player
Audio(fname, autoplay=False)
```
