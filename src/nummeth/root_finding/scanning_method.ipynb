# scanning_method
Find a root in an interval using scanning method

```python
import math

def f(x):
    y = math.sin(x - 0.26 * math.pi) - 0.41
    return y

x = [i * 0.03  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

xroot = [1, 2]
yroot = [f(i) for i in xroot]

print("Root existance between", xroot[0], "and", xroot[1], ":", end=' ')
if yroot[0] * yroot[1] < 0:
    print(True)
else:
    print(False)
```

```python
N = 1
xbeg = xroot[0]
xend = xroot[1]
dx = (xend - xbeg) / N


sign = 0
for i in range(N + 1):
    xi = xbeg + i * dx
    xi = round(xi * N) / N
    
    yi = f(xi)
    if yi > 0:
        cur_sign = 1
    else:
        cur_sign = -1
    
    if i == 0:
        sign = cur_sign
    
    if cur_sign * sign < 0:
        xleft = xi - dx
        xright = xi
        break

print("Root exists between", xleft, "and", xright)
```
