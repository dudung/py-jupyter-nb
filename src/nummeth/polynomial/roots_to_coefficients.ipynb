# roots_to_coefficients
Relation between roots and polynomial coefficients
<br>Refs: [Polynomial Roots and Coefficients](https://www.youtube.com/watch?v=7xCJrTc366k)

## cubic equation

$$\tag{1}
f(x) = c_3 x^3 + c_2 x^2 + c_1 x + c_0
$$

```python
def f(x, c, roots=[]):
    denom = 1
    if len(roots) > 0:
        for i in roots:
            if i != x:
                denom *= (x - i)
            else:
                denom *= (x - i + 1E-30)
    
    y = c[3] * x * x * x
    y += c[2] * x * x
    y += c[1] * x
    y += c[0]
    y /= denom
    return y
```

## false position method
$$\tag{2}
x_{n+2} = \frac{x_n f(x_{n+1}) - x_{n+1} f(x_n)}{f(x_{n+1}) - f(x_n)}
$$


```python
def false_position(xinit, func, coefs, roots, eps=1E-20, iter=200):
    xr = xinit.copy()
    i = 2
    dx = 10
    
    while dx > eps and i < iter:
        x0 = xr[i-2]
        f0 = func(x0, coefs, roots) 

        x1 = xr[i-1]
        f1 = func(x1, coefs, roots)
                
        x2 =  (x0 * f1 - x1 * f0) / (f1 - f0)
        f2 = func(x2, coefs, roots)

        xr.append(x2)
        dx = abs(f2)
        
        i += 1
    
    root = xr[-1]
    return root
```

## roots
Modify quadratic function with obtained root(s) to get other root(s)

```python
coefs = [-80, 66, -15, 1]
print("cubic function =", coefs)

xinit = [0, 9]
print("initial values =", xinit)

roots = []
root = false_position(xinit, f, coefs, roots)
roots.append(root)
root = false_position(xinit, f, coefs, roots)
roots.append(root)
root = false_position(xinit, f, coefs, roots)
roots.append(root)
print("roots = ", roots)
```

## graphical method

```python
# create lists for data points
x = [i/10 for i in range(0, 101)]
y = [f(i, coefs) for i in x]

# create approximated roots based on previous information
xroot = [2, 5, 8]
yroot = [f(i, coefs) for i in xroot]

# plot data points
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("y")
plt.plot(x, y, 'b-', xroot, yroot, 'ro')
plt.show()
```
