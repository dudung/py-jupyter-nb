# write_file_sheets
Write excel file using pandas to some sheets

```python
import os
if os.path.exists("data/xy_data.xlsx"):
    os.remove("data/xy_data.xlsx")
    print("File is deleted")
else:
    print("File does not exist")
```

```python
import pandas as pd

# read excel file as dataframe
df_in = pd.read_excel("data/xy_data.xlsx")
print(df_in)
```

```python
import pandas as pd

# create lists
headers = ['x', 'y']
col_0 = [0, 1, 2, 3, 4, 5]
col_1 = [8, 4, 3, 3, 4, 8]

# create dataframe from previous lists
df_out = pd.DataFrame(list(zip(col_0, col_1)), columns=headers)
print(df_out)

# write dataframe to excel file
df_out.to_excel(
    "data/xy_data.xlsx",
    sheet_name="sheet_xy",
    index=False
)
```
