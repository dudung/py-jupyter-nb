# 3d_scatter
Read excel file using pandas and plot 3d scatter chart using matplotlib

## read dataframe from file

```python
import pandas as pd
df = pd.read_excel('data/xy_scatter.xlsx')
print(df)
```

## get available columns header

```python
headers = df.columns.tolist()
print(headers)
```

## get x as list

```python
x = df['x'].tolist()
print(x)
```

## get quadratic data

```python
yp2 = df['yp2'].tolist()
print(yp2)
```

## plot quadratic function

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, yp2, 'b-')
plt.legend(['quadratic'])
plt.show()
```

## get cubic data

```python
yp3 = df['yp3'].tolist()
print(yp3)
```

## plot cubic function

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, yp3, 'r--')
plt.legend(['cubic'])
plt.show()
```

## plot quadratic and cubic functions

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, yp2, 'b-', x, yp3, 'r--')
plt.legend(['quadratic', 'cubic'])
plt.show()
```
