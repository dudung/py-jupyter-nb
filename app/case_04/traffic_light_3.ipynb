simulate state change of traffic light and display in old fashion way (txt)

## way=1, lane=1, light=0
```
--------------------------------------------------------------------------------
  -->
--------------------------------------------------------------------------------
```

````python
# state 0 move, 1 prepare to stop, 2 stop
stateG = 0
stateY = 1
stateR = 2

# time is s
G = 50
Y = 5
R = 35

# create sequence
seqG = [stateG] * G
seqY = [stateY] * Y
seqR = [stateR] * R

# join all sequences
seq = seqG + seqY + seqR
print(seq)
````

````python
import time as tm
# time delay in s
dt = 0.1
t = 0
for i in seq:
    if i == 0:
        print("t = ", t, end=' ', sep='')
        print("Jalan     ", end='\r')
        tm.sleep(dt)
    elif i == 1:
        print("t = ", t, end=' ', sep='')
        print("Hati-hati ", end='\r')
        tm.sleep(dt)
    else:
        print("t = ", t, end=' ', sep='')
        print("Berhenti  ", end='\r')
        tm.sleep(dt)
    t += 1
````
