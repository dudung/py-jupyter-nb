## multiple series

```python
x1 = [0, 1, 2, 3, 4]
y1 = [0, 1, 2, 3, 4]

x2 = [4, 5, 6, 7]
y2 = [4, 4, 4, 4]

x3 = [7, 8, 9, 10]
y3 = [4, 5, 7, 10]

print('S1\nx1 =', x1, '\ny1 =', y1, '\n')
print('S2\nx2 =', x2, '\ny2 =', y2, '\n')
print('S3\nx3 =', x3, '\ny3 =', y3, '\n')
```

```python
import matplotlib.pyplot as plt

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)

plt.show()
```
