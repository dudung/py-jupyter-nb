## multiple series

```python
x1 = [0, 1, 2, 3, 4]
y1 = [0, 1, 2, 3, 4]

x2 = [4, 5, 6, 7]
y2 = [4, 4, 4, 4]

x3 = [7, 8, 9, 10]
y3 = [4, 5, 7, 10]

print('S1\nx1 =', x1, '\ny1 =', y1, '\n')
print('S2\nx2 =', x2, '\ny2 =', y2, '\n')
print('S3\nx3 =', x3, '\ny3 =', y3, '\n')
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)

labels = [
    'linear',
    'constant',
    'quadratic'
]
plt.legend(labels)

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.scatter(x3, y3)

labels = [
    'linear',
    'constant',
    'quadratic'
]
plt.legend(labels)

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, y1, color='r')
plt.scatter(x2, y2, color='g')
plt.scatter(x3, y3, color='b')

labels = [
    'linear',
    'constant',
    'quadratic'
]
plt.legend(labels)

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, y1, marker='v')
plt.scatter(x2, y2, marker='o')
plt.scatter(x3, y3, marker='s')

labels = [
    'linear',
    'constant',
    'quadratic'
]
plt.legend(labels)

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(x1, y1, c='b', marker='s')
plt.scatter(x2, y2, c='g', marker='v')
plt.scatter(x3, y3, c='r', marker='^')

labels = [
    'linear',
    'constant',
    'quadratic'
]
plt.legend(labels)

plt.grid()
plt.show()
```
