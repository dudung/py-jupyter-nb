## function

```python
def linear(x, a, b):
    y = []
    for i in x:
        y.append(a + b * i)
    return y

x = [*range(0, 11)]
y = linear(x, 2, 1)
print('x:', x)
print('y:', y)
```

```python
import matplotlib.pyplot as plt

x = [*range(0, 11)]
y = linear(x, 4, -1)

plt.plot(x, y)
plt.show()
```

```python
def quadrat(x, a, b, c):
    y = []
    for i in x:
        y.append(a * i**2 + b * i + c)
    return y

x = [*range(0, 11)]
y = quadrat(x, 1, -10, 21)
print('x:', x)
print('y:', y)
```

```python
import matplotlib.pyplot as plt

x = [*range(0, 11)]
y = quadrat(x, 1, -13, 40)

plt.plot(x, y)
plt.show()
```
