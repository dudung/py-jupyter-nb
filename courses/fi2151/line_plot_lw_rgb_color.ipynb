# line width and rgb color

```python
import matplotlib.pyplot as plt
x = [*range(-5, 6)]
y0 = [1 * i**2 for i in x]
y1 = [2 * i**2 for i in x]
y2 = [4 * i**2 for i in x]
y3 = [8 * i**2 for i in x]
y4 = [16 * i**2 for i in x]
plt.plot(x, y0)
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.ylim([0, 100])
plt.legend([1, 2, 4, 8, 16])
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [*range(-5, 6)]
y0 = [1 * i**2 for i in x]
y1 = [2 * i**2 for i in x]
y2 = [4 * i**2 for i in x]
y3 = [8 * i**2 for i in x]
y4 = [16 * i**2 for i in x]
plt.plot(x, y0, '#000')
plt.plot(x, y1, '#444')
plt.plot(x, y2, '#888')
plt.plot(x, y3, '#aaa')
plt.plot(x, y4, '#ccc')
plt.ylim([0, 100])
plt.legend([1, 2, 4, 8, 16])
plt.grid()
plt.show()
```
