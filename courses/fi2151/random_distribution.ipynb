# random distribution

## uniform

```python
import random as rnd

xmin = 20
xmax = 30
N = 16

for i in range(N):
    r = rnd.uniform(xmin, xmax)
    r = round(r, 4)
    print(r)
```

```python
import random as rnd

xmin = 20
xmax = 30
x = []
N = 16

for i in range(N):
    r = rnd.uniform(xmin, xmax)
    r = round(r, 4)
    x.append(r)

for i in x:
    print(i)
```

```python
import random as rnd

def datauniform(xmin, xmax, N):
    xx = []
    for ii in range(N):
        rr = rnd.uniform(xmin, xmax)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = datauniform(20, 30, 16)

for i in x:
    print(i)
```

```python
import matplotlib.pyplot as plt

xmin = 20; xmax = 30; N = 16
x = datauniform(xmin, xmax, N)

plt.hist(x, bins=10, range=(xmin, xmax))
plt.xticks(range(xmin,xmax+1))
plt.xlim(xmin, xmax)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```

## triangular()

```python
import random as rnd

def datatriangular(low, high, mode, N):
    xx = []
    for ii in range(N):
        rr = rnd.triangular(low, high, mode)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = datatriangular(10, 20, 15, 16)

for i in x:
    print(i)
```

```python
import matplotlib.pyplot as plt

low = 10; high = 20; mode = 15; N = 16
x = datatriangular(low, high, mode, N)

plt.hist(x, bins=100, range=(low, high))
plt.xticks(range(low,high+1))
plt.xlim(low, high)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```

## betavariate()

```python
import random as rnd

def databetavar(alpha, beta, N):
    xx = []
    for ii in range(N):
        rr = rnd.betavariate(alpha, beta)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = databetavar(4, 5, 16)

for i in x:
    print(i)
```

```python
import matplotlib.pyplot as plt

alpha = 4; beta = 5; N = 16
x = databetavar(alpha, beta, N)

plt.hist(x, bins=100)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```
