# random distribution

## uniform

```python
import random as rnd

xmin = 20
xmax = 30
N = 16

for i in range(N):
    r = rnd.uniform(xmin, xmax)
    r = round(r, 4)
    print(r)
```

```python
import random as rnd

xmin = 20
xmax = 30
x = []
N = 16

for i in range(N):
    r = rnd.uniform(xmin, xmax)
    r = round(r, 4)
    x.append(r)

for i in x:
    print(i)
```

```python
import random as rnd

def datauniform(xmin, xmax, N):
    xx = []
    for ii in range(N):
        rr = rnd.uniform(xmin, xmax)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = datauniform(20, 30, 16)

for i in x:
    print(i)
```

```python
import matplotlib.pyplot as plt

xmin = 20; xmax = 30; N = 16
x = datauniform(xmin, xmax, N)

plt.hist(x, bins=10, range=(xmin, xmax))
plt.xticks(range(xmin,xmax+1))
plt.xlim(xmin, xmax)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```
