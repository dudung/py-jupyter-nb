## marker customization

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(0, N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(0, N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

plt.scatter(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N)]
y = [(i-4)*(i-10)*(i-16) for i in x]

c = ['r', 'b'] * 10

plt.scatter(x, y, color=c)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

c = ['r', 'g', 'b'] * 7

plt.scatter(x, y, color=c)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

c = ['r', 'g', 'b'] * 7

plt.scatter(x, y, color=c, s=150)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
