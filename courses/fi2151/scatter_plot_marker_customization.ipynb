## marker customization

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(0, N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

plt.scatter(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(0, N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

plt.scatter(x, y)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N)]
y = [(i-4)*(i-10)*(i-16) for i in x]

c = ['r', 'b'] * 10

plt.scatter(x, y, color=c)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

c = ['r', 'g', 'b'] * 7

plt.scatter(x, y, color=c)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

c = ['r', 'g', 'b'] * 7

plt.scatter(x, y, c=c, s=150)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N)]
y = [(i-4)*(i-10)*(i-16) for i in x]

c = ['r', 'b'] * 10
s = [50, 200] * 10

plt.scatter(x, y, s=s, c=c)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N+1)]
y = [(i-4)*(i-10)*(i-16) for i in x]

s = [50, 100, 200] * 7
c = ['r', 'g', 'b'] * 7

plt.scatter(x, y, s=s, c=c)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N)]
y = [(i-4)*(i-10)*(i-16) for i in x]

s = [50, 100, 150, 200, 250] * 4
c = ['#000', '#888',
     '#aaa', '#ccc', '#eee'] * 4

plt.scatter(x, y, s=s, c=c)
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import matplotlib.pyplot as plt

N = 20
x = [*range(N)]
y = [(i-4)*(i-10)*(i-16) for i in x]

s = [50, 100, 150, 200, 250] * 4
c = ['#f00', '#f88',
     '#faa', '#fcc', '#fee'] * 4

plt.scatter(x, y, s=s, c=c,
           marker='s')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
