```python
import random as rnd

def randwalk4(n):
    x = [0]
    y = [0]
    for i in range(n):
        j = rnd.randint(1, 4)
        match j:
            case 1:
                x.append(x[i] + 0)
                y.append(y[i] + 1)
            case 2:
                x.append(x[i] + 1)
                y.append(y[i] + 0)
            case 3:
                x.append(x[i] + 0)
                y.append(y[i] - 1)
            case 4:
                x.append(x[i] - 1)
                y.append(y[i] + 0)
    return(x, y)
```

```python
import matplotlib.pyplot as plt

x, y = randwalk4(1000)
plt.plot(x, y, '-')
plt.plot([x[0], x[-1]],
         [y[0], y[-1]], '--r')
dx = (x[-1] - x[0])
dy = (y[-1] - y[0])
dr = (dx*dx + dy*dy)**0.5
print(dr)
```

```python
sdr = 0
n = 400
m = 10000
for _ in range(m):
    x, y = randwalk4(n)
    dx = (x[-1] - x[0])
    dy = (y[-1] - y[0])
    dr = (dx*dx + dy*dy)**0.5
    sdr += dr
avg = sdr / m
print(avg)
```
