# kinematics_xva_graphs
Plot kinematics graphs using Python `matplotlib` module.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

## position
Suppose there is a function of position

<a name='eqn1'></a>
\begin{equation}\tag{1}
x(t) = x^2 - 2x + 1,
\end{equation}

where $t$ is time.

## velocity
Function of velocity is as follow

<a name='eqn2'></a>
\begin{equation}\tag{2}
v(t) = 2x - 2,
\end{equation}

which is obtaned by deriving Eqn ([1](#eqn1)) with respect to time $t$.

## acceleration
Function of acceleration is as follow

<a name='eqn3'></a>
\begin{equation}\tag{3}
a(t) = 1,
\end{equation}

which is obtaned by deriving Eqn ([2](#eqn2)) with respect to time $t$.

```python
# define a function of x
def f(x):
    y = 2 * np.sin(x)
    return y
```

```python
# create data
xbeg = 0
xend = 2 * np.pi
N = 100
dx = (xend - xbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    x = xbeg + i * dx
    y = f(x)
    
    xx.append(x)
    yy.append(y)
```

```python
# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((xbeg, xend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()
```
