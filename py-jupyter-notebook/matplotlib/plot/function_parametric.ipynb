# function_parametric
Plot a function using Python `matplotlib` module.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

## function
Suppose there is a function

<a name='eqn1'></a>
\begin{equation}\tag{1}
f_1(t) = 2 \sin t
\end{equation}

<a name='eqn2'></a>
\begin{equation}\tag{2}
f_2(t) = 2 \sin 2t
\end{equation}

that will be drawn for $x \in [x_{\rm beg}, x_{\rm end}]$.

```python
# define 1st function of t
def f1(t):
    y = 2 * np.sin(t)
    return y

# define 2nd function of t
def f2(t):
    y = 2 * np.sin(2 * t)
    return y
```

```python
# create data
tbeg = 0
tend = 2 * np.pi
N = 100
dt = (tend - tbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    t = tbeg + i * dt
    x = f1(t)
    y = f2(t)
    
    xx.append(x)
    yy.append(y)
```

```python
# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((-2, 2), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)
line = plt.Line2D((0, 0), (-2, 2), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()
```
