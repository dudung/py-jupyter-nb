## uniform

```python
import random as rnd

xmin = 20
xmax = 30
N = 16

for i in range(N):
    r = rnd.uniform(xmin, xmax)
    r = round(r, 4)
    print(r)
```

```python
import random as rnd

xmin = 20
xmax = 30
x = []
N = 16

for i in range(N):
    r = rnd.uniform(xmin, xmax)
    r = round(r, 4)
    x.append(r)

for i in x:
    print(i)
```

```python
import random as rnd

def datauniform(xmin, xmax, N):
    xx = []
    for ii in range(N):
        rr = rnd.uniform(xmin, xmax)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = datauniform(20, 30, 16)

for i in x:
    print(i)
```

```python
import matplotlib.pyplot as plt

xmin = 20; xmax = 30; N = 16
x = datauniform(xmin, xmax, N)

plt.hist(x, bins=10, range=(xmin, xmax))
plt.xticks(range(xmin,xmax+1))
plt.xlim(xmin, xmax)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```

## triangular()

```python
import random as rnd

def datatriangular(low, high, mode, N):
    xx = []
    for ii in range(N):
        rr = rnd.triangular(low, high, mode)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = datatriangular(10, 20, 15, 16)

for i in x:
    print(i)
```

```python
import matplotlib.pyplot as plt

low = 10; high = 20; mode = 15; N = 16
x = datatriangular(low, high, mode, N)

plt.hist(x, bins=100, range=(low, high))
plt.xticks(range(low,high+1))
plt.xlim(low, high)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```

## expovariate()

```python
import random as rnd

def dataexpovar(lam, N):
    xx = []
    for ii in range(N):
        rr = rnd.expovariate(lam)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = dataexpovar(2, 80)

for i in x:
    print(i, end=', ')
```

```python
import matplotlib.pyplot as plt

lam = -100; N = 100000
x = dataexpovar(lam, N)

plt.hist(x, bins=100)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

lam = -5; N = 100000
x = dataexpovar(lam, N)

plt.hist(x, bins=100)
plt.xlabel('x')
plt.ylabel('n')
plt.grid()
plt.show()
```

## gauss()

```python
import random as rnd

def datagauss(mu, sig, N):
    xx = []
    for ii in range(N):
        rr = rnd.gauss(mu, sig)
        rr = round(rr, 4)
        xx.append(rr)
    return xx

x = datagauss(1, 5, 80)

for i in x:
    print(i, end=', ')
```

```python
import matplotlib.pyplot as plt

mu = 100; sig = 1; N = 100000
x = datagauss(mu, sig, N)

plt.hist(x, bins=100)
plt.xlabel('x')
plt.ylabel('n')
plt.xlim(90, 110)
plt.ylim(0, 4000)
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

mu = 100; sig = 3; N = 100000
x = datagauss(mu, sig, N)

plt.hist(x, bins=100)
plt.xlabel('x')
plt.ylabel('n')
plt.xlim(90, 110)
plt.ylim(0, 4000)
plt.grid()
plt.show()
```
