## lissajous figure

```python
import matplotlib.pyplot as plt
import math

def f(t, omega, phi, A):
    y = A * math.sin(omega * t + phi)
    return y

def d2r(x):
    y = x * 2 * math.pi / 180
    return y
```

```python
delta = 0
a = 1
b = 1

t = [d2r(i*3) for i in range(0, 120)]
x = [f(i, a, delta, 1) for i in t]
y = [f(i, b, 0, 1) for i in t]

plt.figure(figsize=(4, 4))
plt.plot(x, y)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.grid()
plt.show()
```
