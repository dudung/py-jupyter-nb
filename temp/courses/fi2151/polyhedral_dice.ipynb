```python
import matplotlib.pyplot as plt
import numpy as np
from itertools import product

# Define the dice
die1_sides = 10  # D6
die2_sides = 4 # D10

# Generate all possible sums
outcomes = [sum(pair) for pair in product(range(1, die1_sides + 1), range(1, die2_sides + 1))]

# Determine the range of possible sums
min_sum = 2  # Minimum sum of the dice
max_sum = die1_sides + die2_sides  # Maximum sum of the dice
bins = np.arange(min_sum, max_sum + 2)  # Bin edges for integers

# Create the histogram
plt.hist(outcomes, bins=bins, align='left', rwidth=0.8, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Histogram of Sums for a D6 and D10')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Frequency')
plt.xticks(np.arange(min_sum, max_sum + 1))  # Show all possible sums on the x-axis

# Show the plot
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, shapiro, skew, kurtosis

# Example data
data = np.random.normal(50, 10, 1000)  # Generate normal distribution data

# Histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Fit a normal curve
mean, stddev = np.mean(data), np.std(data)
x = np.linspace(min(data), max(data), 1000)
plt.plot(x, norm.pdf(x, mean, stddev), color='red', label='Normal Fit')

# Add labels
plt.title(f"Mean: {mean:.2f}, Std Dev: {stddev:.2f}")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()

# Normality checks
print(f"Skewness: {skew(data):.2f} (close to 0 for normality)")
print(f"Kurtosis: {kurtosis(data) + 3:.2f} (close to 3 for normality)")
stat, p_value = shapiro(data)
print(f"Shapiro-Wilk test p-value: {p_value:.3f}")
if p_value > 0.05:
    print("Data appears to be normally distributed (fail to reject H0).")
else:
    print("Data does not appear to be normally distributed (reject H0).")
```

```python
import random as rnd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, shapiro, skew, kurtosis

dice = [2, 3, 4, 6, 6]
N = 1000

y = []
for _ in range(N):
    sx = 0
    for d in dice:
        x = rnd.randint(1, d)
        sx += x
    y.append(sx)

min_sum = len(dice)  # Minimum sum of the dice
max_sum = sum(dice)  # Maximum sum of the dice
bins = np.arange(min_sum, max_sum + 2)  # Bin edges for integers

mean = np.mean(y)
stddev = np.std(y)

# Create the histogram
plt.hist(y, bins=bins, align='left', rwidth=0.8, color='skyblue', edgecolor='black')
plt.xticks(np.arange(min_sum, max_sum + 1))  # Show all possible sums on the x-axis

xx = np.linspace(min(y), max(y), 1000)
plt.plot(xx, N*norm.pdf(xx, mean, stddev), color='red', label='Normal Fit')

plt.show()

print('mean =', mean)
print('stdev =', stddev)
print(f"Skewness: {skew(y):.2f} (close to 0 for normality)")
print(f"Kurtosis: {kurtosis(y) + 3:.2f} (close to 3 for normality)")
stat, p_value = shapiro(y)
print(f"Shapiro-Wilk test p-value: {p_value:.3f}")
if p_value > 0.05:
    print("Data appears to be normally distributed (fail to reject H0).")
else:
    print("Data does not appear to be normally distributed (reject H0).")
```
