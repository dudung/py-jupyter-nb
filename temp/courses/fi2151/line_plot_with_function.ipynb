## function

```python
def linear(x, a, b):
    y = []
    for i in x:
        y.append(a + b * i)
    return y

x = [*range(0, 11)]
y = linear(x, 2, 1)
print('x:', x)
print('y:', y)
```

```python
import matplotlib.pyplot as plt

x = [*range(0, 11)]
y = linear(x, 4, -1)

plt.plot(x, y)
plt.show()
```

```python
def quadrat(x, a, b, c):
    y = []
    for i in x:
        y.append(a * i**2 + b * i + c)
    return y

x = [*range(0, 11)]
y = quadrat(x, 1, -10, 21)
print('x:', x)
print('y:', y)
```

```python
import matplotlib.pyplot as plt

x = [*range(0, 11)]
y = quadrat(x, 1, -13, 40)

plt.plot(x, y)
plt.show()
```

```python
def poly(x, c):
    y = []
    for xx in x:
        yy = 0
        for n, cc in enumerate(c):
            yy += cc * xx**n
        y.append(yy)
    return y

x = [*range(0, 11)]
y = poly(x, [0, 0, 1])
print('x:', x)
print('y:', y)
```

```python
import matplotlib.pyplot as plt

x = [*range(0, 11)]
y1 = quadrat(x, -1, 13, -40)
y2 = poly(x, [-40, 13, -1])

plt.plot(x, y1)
plt.plot(x, y2, 'o')
plt.legend(['quadrat', 'poly'])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [*range(-12, 5)]
y = poly(x, [1, -10, 10, 1])

plt.plot(x, y, 'sc-')
plt.legend(['cubic'])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [*range(-10, 11)]
y = poly(x, [1, 1, 1, 1, 1, 1, 1])

plt.plot(x, y, 'oy-')
plt.legend(['cubic'])
plt.show()
```
