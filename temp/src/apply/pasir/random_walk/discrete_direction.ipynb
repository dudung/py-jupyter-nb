# discrete direction

## integer direction

```python
import random as rnd
import math as m

angle4 = [0, 90, 180, 270]
dr = 1; x = 0; y = 0; N = 18

print('dir\tangle\tdx\tdy\tx\ty')
for i in range(N):
    d = rnd.randint(0, 3)
    a = angle4[d]
    dx = dr * m.cos(m.radians(a))
    dx = round(dx, 3)
    dy = dr * m.sin(m.radians(a))
    dy = round(dy, 3)
    print(d, a, dx, dy,
          x, y, sep='\t')
    x += dx
    y += dy
```

```python
import random as rnd
import math as m

angle4 = [0, 90, 180, 270]
dr = 1; x = [0]; y = [0]; N = 60

for i in range(N):
    d = rnd.randint(0, 3)
    a = angle4[d]
    dx = dr * m.cos(m.radians(a))
    dx = round(dx, 3)
    dy = dr * m.sin(m.radians(a))
    dy = round(dy, 3)
    x.append(x[i] + dx)
    y.append(y[i] + dy)

print('x =', x, '\n\ny =', y)
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, '-r')
plt.grid()
plt.xlabel('x')
plt.ylabel('x')
plt.plot(x[0], y[0], 'ob')
plt.show()
```

```python
def randomwalk4(xo, yo, N):
    angle4 = [0, 90, 180, 270]
    dr = 1; x = [xo]; y = [yo]
    
    for i in range(N):
        d = rnd.randint(0, 3)
        a = angle4[d]
        dx = dr * m.cos(m.radians(a))
        dx = round(dx, 3)
        dy = dr * m.sin(m.radians(a))
        dy = round(dy, 3)
        x.append(x[i] + dx)
        y.append(y[i] + dy)

    return(x, y)
```

```python
import matplotlib.pyplot as plt

(x, y) = randomwalk4(0, 0, 10)

plt.figure(figsize = (4, 4))
plt.plot(x, y, '-r')
plt.grid()
plt.xlabel('x')
plt.ylabel('x')
plt.plot(x[0], y[0], 'ob')
plt.show()
```

```python
def randomwalk8(xo, yo, N):
    angle4 = [i*45 for i in range(8)]
    dr = 1; x = [xo]; y = [yo]
    
    for i in range(N):
        d = rnd.randint(0, 7)
        a = angle4[d]
        dx = dr * m.cos(m.radians(a))
        dx = round(dx, 3)
        dy = dr * m.sin(m.radians(a))
        dy = round(dy, 3)
        x.append(x[i] + dx)
        y.append(y[i] + dy)

    return(x, y)
```
