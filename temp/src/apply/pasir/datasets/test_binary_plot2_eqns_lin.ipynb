## binary and plot2 -- linear

```python
from pasir.datasets.clasdata import binary
from pasir.datasets.dataviz import plot2
import matplotlib.pyplot as plt

r1 = [-1, 1.1, 0.1]
r2 = [-1, 1.1, 0.1]
coeffs = [[0], [1, 0]]
x, y, z = binary(coeffs, r1=r1, r2=r2)

plot2(x, y, z)
plt.yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```
