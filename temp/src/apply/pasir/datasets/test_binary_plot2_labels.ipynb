## binary and plot2

```python
from pasir.datasets.clasdata import binary
from pasir.datasets.dataviz import plot2

r1 = [0, 1.05, 0.05]
r2 = [0, 1.05, 0.05]
labels = [0, 1]
coeffs = [[0], [-1, 1]]
x, y, z = binary(coeffs,
                r1=r1, r2=r2,
                labels=labels)

print(z[:10])
print()
print(z[-10:])
```
