## binary and plot2

```python
from pasir.datasets.clasdata import binary
from pasir.datasets.dataviz import plot2

r1 = [0, 2.05, 0.1]
r2 = [0, 1.05, 0.05]
coeffs = [[1], [-4, -1], [4, 0, 0]]
x, y, z = binary(coeffs, r1=r1, r2=r2)

plot2(x, y, z)
```
