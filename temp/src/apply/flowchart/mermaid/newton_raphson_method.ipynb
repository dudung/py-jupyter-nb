```python
%%mermaidjs --img
graph TD
    B --> I --> P1 --> o1 --> P2 --> P3
    P3 --> P4 --> C
    C --"Y"--> O --> E
    C --"N"--> P5 --> o1
    B(["Begin"])
    I[/"x<sub>0</sub>, f(x), <br>f'(x),\
      &epsilon;<sub>max</sub>"/]
    P1["i = 1"]
    P2["&Delta;x = f(x<sub>i-1</sub>)<br>\
      &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;/\
      f'(x<sub>i-1</sub>)"]
    P3["x<sub>i</sub> = x<sub>i-1</sub>\
      &mdash; &Delta;x"]
    P4["&epsilon; = |f(x<sub>i</sub>)|"]
    C{"&epsilon; &le; &epsilon;<sub>max</sub>"}
    P5["i = i + 1"]
    E(["End"])
    o1((" "))
    O[/"x<sub>i</sub>"/]
```
