```python
import mermaid as md
from mermaid.graph import Graph
```

```python
graph = Graph('simple flowchart',"""
graph LR
    A --> B --"Y"--> C
    B --"N"--> D
    C & D --> E
    A(["Begin"])
    B{"Condition?"}
    C["Process 1"]
    D["Process 2"]
    E(["End"])
""")
```

```python
from IPython.display import Image,display
display(Image(md.Mermaid(graph).img_response.content))
```
