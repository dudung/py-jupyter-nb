```python
import matplotlib.pyplot as plt
from mpl_flow import Flow

fig, axes = plt.subplots(1, 1, figsize=(4, 3))

ax1 = axes
f = Flow(ax=ax1)
f.node(label="Label\nA", bbox=dict(boxstyle='circle'))
f.node(label="Label\nB", bbox=dict(boxstyle='square'))
f.node(label="Label\nC", bbox=dict(boxstyle='ellipse'))
f.node(label="Label\nD", bbox=dict(boxstyle='round'), travel='s')
f.node(label="Label\nE", bbox=dict(boxstyle='round4'), travel='w')
f.node(label="Label\nF", bbox=dict(boxstyle='roundtooth'), travel='w')
f.node(label="Label\nG", bbox=dict(boxstyle='sawtooth'), travel='s')
f.node(label="Label\nH", bbox=dict(boxstyle='darrow'))
f.node(label="Label\nI", bbox=dict(boxstyle='rarrow'))
f.node(label="Label\nJ", bbox=dict(boxstyle='larrow'))
plt.show()
```
