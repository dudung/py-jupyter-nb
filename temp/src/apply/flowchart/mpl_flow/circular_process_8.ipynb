```python
import random
import matplotlib.pyplot as plt
from mpl_flow import Flow

def node(id, label, travel):
    bs = dict(boxstyle='round4')
    f.node(id, label=label, bbox=bs, connect=True, travel=travel)

fig, axes = plt.subplots(1, 1, figsize=(5, 4))
ax1 = axes
f = Flow(ax=ax1)

dirs = ['n', 'ne', 'e', 'se', 's', 'sw', 'w', 'nw']
labs = [
    "Concept\n1", "Concept\n2", "Concept\n3", "Concept\n4",
    "Concept\n5", "Concept\n6", "Concept\n7", "Concept\n8"
]

N = len(labs)
for i in range(N):
    id = f"{i+1:02d}"
    node(id, label=labs[i], travel=dirs[i])
f.edge('08', '01')
plt.show()
```
