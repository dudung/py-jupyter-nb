# read_wav_file
Read wav file (see full feature at [nbviewer](https://nbviewer.org/github/dudung/py-jupyter-nb/blob/main/src/fft/read_wav_file.ipynb))

```python
# import necessary libraries
from scipy.io import wavfile
from IPython.display import Audio, display
import os
```

```python
# define filename
fname = 'data/s1_signal_500hz.wav'

# get file existence
file_exist = os.path.exists(fname)
```

```python
# read file
if file_exist:
    sample_rate_2, data = wavfile.read(fname)
    print(sample_rate_2)
    print(data)
```

```python
# show player
if file_exist:
    display(Audio(fname, autoplay=False))
```
