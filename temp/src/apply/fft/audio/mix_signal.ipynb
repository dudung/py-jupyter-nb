# mix_signal
Mix sinewaves and plot result

```python
# import necessary libraries
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
```

```python
# create 1st signal -- information
freq = 50 # Hz
sample_rate = 44100 # Hz
duration = 5 # s
x1, y1 = gen.sine_wave(freq, sample_rate, duration)
```

```python
# plot 1st signal
plt.grid()
plt.xlabel("t (s)")
plt.ylabel("y")
plt.xlim(0, 1/10)
plt.plot(x1, y1, 'g-')
plt.legend(['50 Hz'])
plt.show()
```

```python
# create 2nd signal -- noise
freq = 500 # Hz
sample_rate = 44100 # Hz
duration = 5 # s
x2, y2 = gen.sine_wave(freq, sample_rate, duration)
```

```python
# plot 2nd signal
plt.grid()
plt.xlabel("t (s)")
plt.ylabel("y")
plt.xlim(0, 1/10)
plt.plot(x2, y2, 'r-')
plt.legend(['500 Hz'])
plt.show()
```

```python
# mix signals -- information + noise
x3 = x2.copy()
y3 = y1 + 0.3 * y2
```

```python
# plot mixed signal
plt.grid()
plt.xlabel("t (s)")
plt.ylabel("y")
plt.xlim(0, 1/10)
plt.plot(x3, y3, 'b-')
plt.legend(['50 Hz + 0.3 500 Hz', '500 Hz'])
plt.show()
```
