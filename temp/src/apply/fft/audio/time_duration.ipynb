# time_duration
Try to understand sample rate parameter.

## terms

Number of samples

$$\tag{1}
N_s = f_s \  \Delta t_s
$$

depends on sampling rate or sampling frequency $f_s$ and duration of sampling $\Delta t_s$.

Maximum frequency observed from FFT

$$\tag{2}
f_{\rm max} = \frac12 N_s
$$

is related to number of samples $N_s$.

## warning
Sampling rate $f_s$ should be larger than observed frequency $f_{\rm obs}$

$$\tag{3}
N_s = 2 f_{\rm max} > f_{\rm obs},
$$

which can not be accomodated by increasing sampling duration

$$\tag{4}
\Delta t_s = \frac{N_s}{f_s} = \frac{2 f_{\rm max}}{f_s} > f_{\rm obs},
$$

due to insufficient information to perform FFT.

## libraries

```python
# import necessary libraries
import numpy as np
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

## 2 Hz sine -- duration 2 s

```python
# define sample rate
duration = 2 # s
sample_rate = 10 # Hz
Nsample = duration * sample_rate
print("Nsample =", Nsample)
print("duration =", duration)

# create signal with sine
freq = 2 # Hz
x, y = gen.sine_wave(freq, sample_rate, duration)

# plot signal in time domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.plot(x, y, '.')
plt.show()
```

```python
# perform fft for positive frequency only
yf = rfft(y)
xf = rfftfreq(Nsample, 1.0/sample_rate)
fmax = len(xf) - 1
print("fmax =", fmax)

# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf, np.abs(yf), 'r-')
plt.show()
```
