# plot_wave_curve
Plot sine and cosine wave

```python
# import necessary libraries
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
```

```python
# create data from sine wave
freq = 2 # Hz
sample_rate = 44100 # Hz
duration = 5 # s
x1, y1 = gen.sine_wave(freq, sample_rate, duration)
```

```python
# plot sine wave
plt.grid()
plt.xlabel("t (s)")
plt.ylabel("y")
plt.plot(x1, y1, 'r-')
plt.legend(['sine 2 Hz'])
plt.show()
```

```python
# create data from cosine wave
freq = 2 # Hz
sample_rate = 44100 # Hz
duration = 5 # s
x2, y2 = gen.cosine_wave(freq, sample_rate, duration)
```

```python
# plot cosine wave
plt.grid()
plt.xlabel("t (s)")
plt.ylabel("y")
plt.plot(x2, y2, 'b-')
plt.legend(['cosine 2 Hz'])
plt.show()
```

```python
# plot both waves
plt.grid()
plt.xlabel("t (s)")
plt.ylabel("y")
plt.plot(x1, y1, 'r-', x2, y2, 'b-')
plt.legend(['sine 2 Hz', 'cosine 2 Hz'])
plt.show()
```
