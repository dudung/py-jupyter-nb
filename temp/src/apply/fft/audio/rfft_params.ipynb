# rfft_params
Try various parameters for rfft and rfftfreq

## libraries

```python
# import necessary libraries
import numpy as np
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

## constants

```python
# define normalize factors
INT16 = 32767 # 16-bit int [-32768, 32767] for audio
INT8 = 255 # 8-bit int [0, 255] for image

# define sample rate
AUDIO = 44100 # Hz
```

## parameters

```python
# define sample rate
sample_rate = 10000 # Hz
duration = 2 # s
Nsample = sample_rate * duration
```

## 100 Hz

```python
# create signal
freq = 100 # Hz
x, y = gen.sine_wave(freq, sample_rate, duration)

# perform fft for positive frequency only
yf = rfft(y)
xf = rfftfreq(Nsample, 1/ sample_rate)
```

```python
# for M wave to plot
M = 1
Nshow = int(M * (sample_rate / freq))

# plot signal in time domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.plot(x[:Nshow], y[:Nshow], '.')
plt.show()
```

```python
# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf, np.abs(yf), 'r-')
plt.show()

print("Nsample in t domain =", Nsample)
print("Number of x points =", len(xf))
print("Number of y points =", len(yf))
```

```python
# for rfft fmax = len(xf) - 1 or Nsample / 2
fmax = len(xf) - 1
print("fmax =", fmax)
print("Nsample / 2 =", int(Nsample / 2))
```

```python
# get Nend for fend
fend = 1000
Nend = int(fend / fmax * (Nsample / 2))
print("Nend =", Nend)

# or get it using points per freq
points_per_freq = len(xf) / (sample_rate / 2)
points_id = int(points_per_freq * fend)
print("points_id =", points_id)

# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf[:Nend], np.abs(yf[:Nend]), 'r-')
plt.show()
```

```python
# get Nend for fend
fend = 500
Nend = int(fend / fmax * (Nsample / 2))
print("Nend =", Nend)

# or get it using points per freq
points_per_freq = len(xf) / (sample_rate / 2)
points_id = int(points_per_freq * fend)
print("points_id =", points_id)

# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf[:Nend], np.abs(yf[:Nend]), 'r-')
plt.show()
```

```python
# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.ylim([0, 1])
plt.plot(xf[:Nend], np.abs(yf[:Nend]), 'r-')
plt.show()
```
