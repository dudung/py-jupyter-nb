# multi_frequency
Try to understand sample rate parameter.

## libraries

```python
# import necessary libraries
import numpy as np
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

## 2 Hz sine -- duration 2 s

```python
# define sample rate
duration = 2 # s
sample_rate = 10 # Hz
Nsample = duration * sample_rate
print("Nsample =", Nsample)
print("duration =", duration)

# create signal with sine
freq = 10 # Hz
x, y = gen.sine_wave(freq, sample_rate, duration)
```

```python
# perform fft for positive frequency only
yf = rfft(y)
xf = rfftfreq(Nsample, 1.0/sample_rate)
fmax = len(xf) - 1
print("fmax =", fmax)

# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf, np.abs(yf), 'r-')
plt.show()
```

## 2 Hz sine -- duration 5 s

```python
# define sample rate
duration = 5 # s
sample_rate = 10 # Hz
Nsample = duration * sample_rate
print("Nsample =", Nsample)
print("duration =", duration)

# create signal with sine
freq = 10 # Hz
x, y = gen.sine_wave(freq, sample_rate, duration)

# plot signal in time domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.plot(x, y, '.')
plt.show()
```

```python
# perform fft for positive frequency only
yf = rfft(y)
xf = rfftfreq(Nsample, 1.0/sample_rate)
fmax = len(xf) - 1
print("fmax =", fmax)

# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf, np.abs(yf), 'r-')
plt.show()
```
