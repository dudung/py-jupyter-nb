# gen_wave_rfft
Generate sine wave then perform rfft and rfftfreq

```python
# import necessary libraries
import numpy as np
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

```python
# define default parameters
sample_rate = 50000 # Hz -- audio should 44100 Hz
int16max = 32767 # 16-bit int -32768 -- 32767 -- to normalize wave for audio
duration = 2 # s
N = sample_rate * duration # number of samples
```

## 100 Hz

```python
# create 1st signal -- information
freq = 100 # Hz
x, y = gen.sine_wave(freq, sample_rate, duration)

# perform fft for positive frequency only
yf = rfft(y)
xf = rfftfreq(N, 1/ sample_rate)
```

```python
# for M wave to plot
M = 1
Ndata = round(M * N / (duration * freq))

# plot signal in time domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.plot(x[:Ndata], y[:Ndata], 'b.', x[:Ndata], y[:Ndata], '#ccf')
plt.show()
```

```python
# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf[:2000], np.abs(yf[:2000]), 'r-')
plt.show()
```
