# sample_rate
Try to understand sample rate parameter.

## terms

Number of samples

$$\tag{1}
N_s = f_s \  \Delta t_s
$$

depends on sampling rate or sampling frequency $f_s$ and duration of sampling $\Delta t_s$.

Maximum frequency observed from FFT

$$\tag{2}
f_{\rm max} = \frac12 N_s
$$

is related to number of samples $N_s$.

## libraries

```python
# import necessary libraries
import numpy as np
import ipynb.fs.full.generator as gen
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
```

## 1 Hz sine

```python
# define sample rate
sample_rate = 10 # Hz
duration = 1 # s
Nsample = sample_rate * duration
print("Nsample =", Nsample)

# create signal with sine
freq = 1 # Hz
x, y = gen.sine_wave(freq, sample_rate, duration)

# plot signal in time domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.plot(x, y, '.')
plt.show()
```

```python
# perform fft for both positive and negative frequency only
yf = rfft(y)
xf = rfftfreq(Nsample, 1/ sample_rate)
fmax = len(xf) - 1
print("fmax =", fmax)

# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf, np.abs(yf), 'r-')
plt.show()
```
