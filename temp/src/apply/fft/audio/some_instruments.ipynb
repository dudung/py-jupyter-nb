+ Examples are available at [Physical Audio Signal Processing](https://ccrma.stanford.edu/~jos/pasp/Sound_Examples.html)

```python
# import necessary libraries
from scipy.io import wavfile
from IPython.display import Audio, display
import os
from scipy.fft import rfft, rfftfreq
import matplotlib.pyplot as plt
import numpy as np
```

## instrument_1

```python
# define filename
fname = 'data/instrument_1.wav'

# get file existence
file_exist = os.path.exists(fname)
```

```python
# read file
if file_exist:
    sample_rate, data = wavfile.read(fname)
    print(sample_rate)
    print(data)
```

```python
# show player
if file_exist:
    display(Audio(fname, autoplay=False))
```

```python
Nsample = len(data)
print("Nsample =", Nsample)
```

```python
yf = rfft(data)
xf = rfftfreq(Nsample, 1/ sample_rate)
```

```python
# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('f')
plt.ylabel('y')
plt.plot(xf[:9200], np.abs(yf[:9200]), 'r-')
plt.show()
```

```python
# plot signal in freqency domain
plt.figure(figsize=(10, 4))
plt.grid()
plt.xlabel('bin')
plt.ylabel('y')
plt.plot(data)
plt.show()
```
