# laplace_heat_square_plate_diff_iter
Solve Laplace equation with FDM for square plate with different number of iteration. See [laplace_heat_square_plate.ipynb](laplace_heat_square_plate.ipynb) for previous step.

```python
import copy
import matplotlib.pyplot as plt
```

```python
ROWS = 100
COLS = 100
u = [[0 for j in range(COLS)] for i in range(ROWS)]

i = ROWS - 1
for j in range(COLS):
    u[i][j] = 1

TIME = 10
for t in range(0, TIME + 1):
    v = copy.deepcopy(u)
    for i in range(1, ROWS - 1):
        for j in range(1, COLS - 1):
            u[i][j] = 0.25 * (v[i-1][j] + v[i+1][j] + v[i][j-1] + v[i][j+1])

# https://stackoverflow.com/a/58512992/9475509
plt.set_cmap('coolwarm')

# https://stackoverflow.com/a/16492880/9475509
fig = plt.figure(figsize=(6, 3.2))
ax = fig.add_subplot(111)
ax.set_title('T(x, y)')
ax.set_aspect('equal')
plt.imshow(u)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(orientation='vertical')
plt.show()
```
