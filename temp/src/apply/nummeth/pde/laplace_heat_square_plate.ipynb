# laplace_heat_square_plate
Solve Laplace equation with FDM for square plate.

## laplace equation

$$\tag{1}
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0,
$$

$$\tag{2}
u_{xx} + u_{yy} = 0.
$$

## finite difference method
$$\tag{3}
f''(x) \approx \frac{f(x + \Delta x) - 2f(x) + f(x - \Delta x)}{\Delta x^2},
$$

$$\tag{4}
u_{xx} = \frac{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}{h^2},
$$

$$\tag{5}
u_{yy} = \frac{u_{i,j+1} - 2u_{i,j} + u_{i,j-1}}{h^2}.
$$

## (4) & (5) &rightarrow; (2)
$$\tag{6}
u_{i,j} = \frac{1}{4h^2} \left( u_{i+1,j} + u_{i,j+1} + u_{i-1,j} + u_{i,j-1} \right).
$$

```python
import copy
import matplotlib.pyplot as plt
```

```python
ROWS = 100
COLS = 100
u = [[0 for j in range(COLS)] for i in range(ROWS)]

i = ROWS - 1
for j in range(COLS):
    u[i][j] = 1

TIME = 10000
for t in range(0, TIME + 1):
    v = copy.deepcopy(u)
    for i in range(1, ROWS - 1):
        for j in range(1, COLS - 1):
            u[i][j] = 0.25 * (v[i-1][j] + v[i+1][j] + v[i][j-1] + v[i][j+1])

# https://stackoverflow.com/a/58512992/9475509
plt.set_cmap('coolwarm')

# https://stackoverflow.com/a/16492880/9475509
fig = plt.figure(figsize=(6, 3.2))
ax = fig.add_subplot(111)
ax.set_title('T(x, y)')
ax.set_aspect('equal')
plt.imshow(u)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(orientation='vertical')
plt.show()
```

```python
ROWS = 100
COLS = 100
u = [[0 for j in range(COLS)] for i in range(ROWS)]

imin = 40
imax = 60
jmin = 40
jmax = 60
for i in range(imin, imax):
    for j in range(jmin, jmax):
        u[i][j] = 1

TIME = 10000
for t in range(0, TIME + 1):
    v = copy.deepcopy(u)
    for i in range(1, ROWS - 1):
        for j in range(1, COLS - 1):
            if not((imin < i and i < imax) and (jmin < j and j < jmax)):
                u[i][j] = 0.25 * (v[i-1][j] + v[i+1][j] + v[i][j-1] + v[i][j+1])

# https://stackoverflow.com/a/58512992/9475509
plt.set_cmap('coolwarm')

# https://stackoverflow.com/a/16492880/9475509
fig = plt.figure(figsize=(6, 3.2))
ax = fig.add_subplot(111)
ax.set_title('T(x, y)')
ax.set_aspect('equal')
plt.imshow(u)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(orientation='vertical')
plt.show()
```

```python
ROWS = 100
COLS = 100
u = [[0 for j in range(COLS)] for i in range(ROWS)]

imin1 = 20
imax1 = 40
jmin1 = 20
jmax1 = 40
for i in range(imin1, imax1):
    for j in range(jmin1, jmax1):
        u[i][j] = 1

imin2 = 60
imax2 = 80
jmin2 = 60
jmax2 = 80
for i in range(imin2, imax2):
    for j in range(jmin2, jmax2):
        u[i][j] = 1

TIME = 10000
for t in range(0, TIME + 1):
    v = copy.deepcopy(u)
    for i in range(1, ROWS - 1):
        for j in range(1, COLS - 1):
            s1 = (imin1 < i and i < imax1) and (jmin1 < j and j < jmax1)
            s2 = (imin2 < i and i < imax2) and (jmin2 < j and j < jmax2)
            if not(s1 or s2):
                u[i][j] = 0.25 * (v[i-1][j] + v[i+1][j] + v[i][j-1] + v[i][j+1])

# https://stackoverflow.com/a/58512992/9475509
plt.set_cmap('coolwarm')

# https://stackoverflow.com/a/16492880/9475509
fig = plt.figure(figsize=(6, 3.2))
ax = fig.add_subplot(111)
ax.set_title('T(x, y)')
ax.set_aspect('equal')
plt.imshow(u)
plt.xlabel("x")
plt.ylabel("y")
plt.colorbar(orientation='vertical')
plt.show()
```
