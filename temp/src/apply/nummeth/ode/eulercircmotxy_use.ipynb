
# eulercircmotxy_use
Use of module for circular motion in xy coordinate with Euler algorithm.

## magnetic force on moving charge

```python
# import necessary libarry
import math
import matplotlib.pyplot as plt
import ipynb.fs.full.eulercircmotxy as ecmxy
```

```python
# set default parameters
m = 1
q = 1
v = [0, 1, 0]
B = [0, 0, -1]
tbeg = 0
tend = 2 * math.pi
```

```python
N = 10
t, x, y, z = ecmxy.position(m, q, v, B, tbeg, tend, N)
xmax = max(abs(min(x)), max(x))
ymax = max(abs(min(y)), max(y))
xymax = max(xmax, ymax)
plt.figure(figsize=(6, 6))
plt.xlim([-xymax, xymax])
plt.ylim([-xymax, xymax])
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x, y, 'ro-', mfc='#f88')
plt.show()
```
