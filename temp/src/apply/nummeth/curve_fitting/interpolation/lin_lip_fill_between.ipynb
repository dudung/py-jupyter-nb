# lin_lip_fill_between
Use Lagrange interpolating polynomial and linear interpolation implementations with different data

```python
# import libraries
import ipynb.fs.full.linearint as lin
import ipynb.fs.full.lagipoly as lip
import matplotlib.pyplot as plt
import numpy as np
```

### data 1

```python
x1 = [0, 1, 2, 3, 4, 5]
y1 = [0, 1, 2, 3, 4, 5]

y1a = [y + 1 for y in y1]
y1b = [y - 1 for y in y1]

x2 = np.linspace(0, 5, 100).tolist()
y2 = lin.linear_interpolation(x2, x1, y1)
y3a = lip.lagrange_polynomial(x2, x1, y1a)
y3b = lip.lagrange_polynomial(x2, x1, y1b)

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x1, y1, 'ro', x2, y2, 'b--', x2, y3a, 'g-', x2, y3b, 'g-')
plt.fill_between(x2, y3a, y3b, color='green', alpha=0.2)
plt.show()
```
