# newton_raphson_method
Find a root in an interval using newton-raphson method

$$\tag{1}
f(x) = \sin(x - 0.26\pi) - 0.41
$$

```python
import math

def f(x):
    y = math.sin(x - 0.26 * math.pi) - 0.41
    return y

x = [i * 0.03  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

def df(x):
    dy = math.cos(x - 0.26 * math.pi)
    return dy
```

$$\tag{2}
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

```python
xroot = []
yroot = []

x0 = 1
y0 = f(x0)

xroot.append(x0)
yroot.append(y0)

i = 1
eps = 1E-6
dx = 10
print("iter root         f(root)")
while dx > eps and i < 20:
    x0 = xroot[i-1]
    if i == 1:
        print(f"{i-1:04d}", f"{x0:.6E}", f"{f(x0):+.6E}")
    
    x1 = x0 - f(x0) / df(x0)
    
    xroot.append(x1)
    dx = abs(f(x1))
    
    print(f"{i:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```
