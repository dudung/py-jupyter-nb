# bisection_method
Find a root in an interval using bisection method

$$\tag{1}
f(x) = \sin(x - 0.26\pi) - 0.41
$$

```python
import math

def f(x):
    y = math.sin(x - 0.26 * math.pi) - 0.41
    return y

x = [i * 0.03  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

xroot = [1, 2]
yroot = [f(i) for i in xroot]

print("Root existance between", xroot[0], "and", xroot[1], ":", end=' ')
if yroot[0] * yroot[1] < 0:
    print(True)
else:
    print(False)
```

```python
x1 = xroot[0]
x2 = xroot[1]
x3 = 0.5 * (x1 + x2)
xroot.append(x3)

eps = 1E-6
dx = 10
i = 2
print("iter root         f(root)")
while dx > eps and i < 100:
    x1 = xroot[i-2]
    x2 = xroot[i-1]
    x3 = xroot[i]
    
    if i == 2:
        print(f"{i-2:04d}", f"{x1:.6E}", f"{f(x1):+.6E}")
        print(f"{i-1:04d}", f"{x2:.6E}", f"{f(x2):+.6E}")
        print(f"{i:04d}", f"{x3:.6E}", f"{f(x3):+.6E}")
   
    f13 = f(x1) * f(x3)
    f23 = f(x2) * f(x3)
     
    if f13 < 0:
        x4 = 0.5 * (x1 + x3)
        xroot[i-2] = x2
        xroot[i-1] = x1
    else:
        x4 = 0.5 * (x2 + x3)
    
    xroot.append(x4)
    dx = abs(f(x4))
    
    print(f"{i:04d}", f"{x4:.6E}", f"{f(x4):+.6E}")
    i += 1
```

```python
print("root =", xroot[-1])
print("f(root) =", f(xroot[-1]))
```

```python
yroot = [f(i) for i in xroot]
```

```python
import matplotlib.pyplot as plt
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.plot(x, y, '-', xroot, yroot, 'r.', lw=2)
plt.show()
```
