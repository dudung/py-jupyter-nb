# graphical_method
Root finding using graphical method

$$\tag{1}
f(x) = \sin(x - 0.26\pi) - 0.41
$$

```python
import math
import matplotlib.pyplot as plt

def f(x):
    y = math.sin(x - 0.26 * math.pi) - 0.41
    return y

x = [i * math.pi / 100  for i in range(0, 100 + 1)]
y = [f(i) for i in x]

ans = math.pi / 4  + math.pi / 6
xroot = [1.25]
yroot = [0]

plt.grid()
plt.plot(x, y, xroot, yroot, 'r+', ms=10, lw=2)
plt.show()
```

```python
xroot = [1.25]
yroot = [0]

plt.grid()
plt.xlim(1, 1.5)
plt.ylim(-0.25, 0.25)
plt.plot(x, y, xroot, yroot, 'r+', ms=10, lw=2)
plt.show()
```
