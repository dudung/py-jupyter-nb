```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# river flows from left to right
river_velocity = 40
river_width = 800

# boat sails from bottom to top
crossing_velocity = 50

# crossing time
t_cross = river_width / crossing_velocity
t = np.arange(0, t_cross + 1, 1)

# dataset for boat relative to river
y0 = 0
y = y0 + crossing_velocity * t

# dataset for river relative to ground
x0 = 0
x = x0 + river_velocity * t
```

```python
plt.scatter(t, x, label='water from ground', marker='o', color='tab:blue')
plt.xlabel('t (s)');
plt.ylabel('x (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(t, y, label='boad from water', marker='^', color='tab:red')
plt.xlabel('t (s)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.scatter(x, y, label='boat from ground', marker='s', color='tab:green')
plt.xlabel('x (m)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 6))

plt.subplot(2,2,1)
plt.scatter(t, y, label='boad from water', marker='^', color='tab:red')
plt.xlabel('t (s)');
plt.ylabel('y (m)');
plt.grid()
plt.legend()

plt.subplot(2,2,4)
plt.scatter(x, t, label='water from ground', marker='o', color='tab:blue')
plt.ylabel('t (s)');
plt.xlabel('x (m)');
plt.grid()
plt.legend()

plt.subplot(2,2,2)
plt.scatter(x, y, label='boat from ground', marker='s', color='tab:green')
plt.grid()
plt.legend()

plt.show()
```
