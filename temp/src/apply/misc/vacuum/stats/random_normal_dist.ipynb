```python
import matplotlib.pyplot as plt
import numpy as np

mu = 80
si  = 10
N = 100000
v2 = np.random.normal(mu, si, N)
  
plt.hist(v2, 100)
plt.axvline(v2.mean(), color='r')
plt.xlim([20,160])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

v3 = np.concatenate((v2, v1))
plt.hist(v3, 100)
plt.axvline(v2.mean(), color='r')
plt.axvline(v1.mean(), color='r')
plt.xlim([20,160])
plt.show()
```

```python
si  = 10
N = 100000
w1 = np.random.normal(80, si, N)
w2 = np.random.normal(100, si, N)
w3 = np.random.normal(120, si, N)
w = np.concatenate((w1, w2, w3))
plt.hist(w, 100)
plt.axvline(w1.mean(), color='r')
plt.axvline(w2.mean(), color='r')
plt.axvline(w3.mean(), color='r')
plt.xlim([20,160])
plt.show()
```

```python
import random as rnd
N = range(0, 1000)
x = [rnd.randint(1, 6) for i in N]
y = [rnd.randint(1, 6) for i in N]
z = [x[i]+y[i] for i in N]
plt.hist(z, 100)
plt.hist(x, 100)
plt.hist(y, 100)
plt.show()
```

```python
x = np.random.uniform(40, 120, 10000)
y = np.random.normal(80, 10, 10000)
zp = x + y
zm = x - y

plt.subplot(2, 2, 1)
plt.hist(x, bins=100)
plt.subplot(2, 2, 2)
plt.hist(y, bins=100)
plt.subplot(2, 2, 3)
plt.hist(zp, bins=100)
plt.subplot(2, 2, 4)
plt.hist(zm, bins=100)
plt.show()
```

```python
x = np.random.uniform(40, 120, 10000)
y = np.random.normal(80, 10, 10000)
zp = x + y
zm = x - y

b = 100
plt.subplot(2, 2, 1); plt.hist(x, bins=b)
plt.subplot(2, 2, 2); plt.hist(y, bins=b)
plt.subplot(2, 2, 3); plt.hist(zp, bins=b)
plt.subplot(2, 2, 4); plt.hist(zm, bins=b)
plt.show()
```

```python
cc = np.corrcoef((x, y, zp, zm))
cc3 = np.round(cc, 3)
print(cc3)
```
