```python
import random as rnd
import matplotlib.pyplot as plt
```

```python
N = 100
a = 2
b = 8
x = []
for i in range(N):
    j = rnd.randint(a, b)
    x.append(j)
    print(j, end=' ')
print('\n\n')

print(len(x))
```

```python
bins = (b - a) + 1

plt.figure(figsize=(4, 3))
plt.hist(x, range=[a, b + 1], bins=bins)
plt.xlabel('x')
plt.ylabel('N')
plt.xticks(range(a, b + 2))
plt.yticks(range(0, 26, 2))
plt.grid()
plt.show()
```

```python
def f(x):
    y = 3*x**2 + 2*x + 1
    return y

xx = [*range(a, b + 1)]
y = [f(i) for i in xx]

plt.figure(figsize=(4, 3))
plt.plot(xx, y, 'o-')
plt.xticks(range(a, b + 2))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
rx = rnd.randint(a, b)
ry = f(rx)
xr = [a, b]
yr = [ry, ry]
c = '#cfc'

plt.figure(figsize=(4, 3))
plt.plot(xx, y)
plt.plot(rx, ry, 'o')
plt.plot(xr, yr, '-')
plt.fill_between(xr, yr, y2=0, color=c)
plt.xticks(range(a, b + 2))
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
c = '#cfc'

plt.figure(figsize=(5, 5))
for i in range(6):
    rx = rnd.randint(a, b)
    ry = f(rx)
    xr = [a, b]
    yr = [ry, ry]
    
    plt.subplot(2, 3, i+1)
    plt.plot(xx, y)
    plt.plot(rx, ry, 'o')
    plt.plot(xr, yr, '-')
    plt.fill_between(xr, yr, y2=0, color=c)
    plt.xticks(range(a, b + 2))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()
plt.show()
```

$$
\begin{array}{rcl}
\displaystyle \int_2^8 \left[ 3x^2 + 2x + 1 \right] \ \ dx & = & \displaystyle \left[ x^3 + x^2 + x \right]_{x = 3}^{8} \newline
& = & \displaystyle (8^3 + 8^2 + 8) - (2^3 + 2^2 + 2) \newline
& = & (512 + 64 + 8) - (8 + 4 + 2) \newline
& = & 584 - 14 \newline
& = & 570.
\end{array}
$$

```python
N2 = 1000
n = []
A = []
AA = 0
for i in range(1, N2 + 1):
    xi = rnd.randint(a, b)
    fi = f(xi)
    Ai = (b - a) * fi
    
    n.append(i)
    if i == 1:
        AA = Ai
    else:
        AA = (A[i-2] * (i-1) + Ai) / i
    A.append(AA)

plt.plot([0, N2], [570, 570], 'r-')
plt.plot(n, A, '-')
plt.xlabel('n')
plt.ylabel('A')
plt.grid()
plt.show()
```
