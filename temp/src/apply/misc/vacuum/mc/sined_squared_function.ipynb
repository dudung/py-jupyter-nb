```python
import math
import random as rnd
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 0
b = 2 * math.pi
N = 200

def f(x):
    y = math.sin(2*x) * math.sin(2*x)
    return y

x = np.linspace(a, b, N, endpoint=True)
y = [f(i) for i in x]
rng = np.linspace(a, b, 5, endpoint=True)

plt.figure(figsize=(4, 3))
plt.plot(x, y, '.-')
plt.xticks(rng)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
rx = (1.0/N) * rnd.randint(0, N)
rx *= (b - a)
rx += a
ry = f(rx)
xr = [a, b]
yr = [ry, ry]
c = '#cfc'
y2 = [0, 0]

plt.figure(figsize=(4, 3))
plt.plot(x, y)
plt.plot(rx, ry, 'o')
plt.plot(xr, yr, '-')
if yr[0] > 0:
    plt.fill_between(xr, yr, y2, color=c)
else:
    plt.fill_between(xr, y2, yr, color=c)
plt.xticks(rng)
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
c = '#cfc'
rng2 = np.linspace(a, b, 3, endpoint=True)

plt.figure(figsize=(6, 5))
for i in range(6):
    rx = (1.0/N) * rnd.randint(0, N)
    rx *= (b - a)
    rx += a
    ry = f(rx)
    xr = [a, b]
    yr = [ry, ry]
    
    plt.subplot(2, 3, i+1)
    plt.plot(x, y)
    plt.plot(rx, ry, 'o')
    plt.plot(xr, yr, '-')

    if yr[0] > 0:
        plt.fill_between(xr, yr, y2, color=c)
    else:
        plt.fill_between(xr, y2, yr, color=c)
    
    plt.xticks(rng2)
    plt.grid()
plt.show()
```

$$
\begin{array}{rcl}
\displaystyle \int_0^{2\pi} \sin^2 x \ \ dx & = & \displaystyle \int_0^{2\pi} (\tfrac12 - \tfrac12 \cos 2x) \ \ dx \newline
& = & \displaystyle \tfrac12 \int_{0}^{2\pi} dx - \tfrac12 \int_{0}^{2\pi} \cos 2x \ \ dx \newline
& = & \displaystyle \tfrac12 \left[ x \right]_{x=0}^{2\pi} - \tfrac14 \left[ \cos 2x \right]_{x=0}^{2\pi} \newline
& = & \tfrac12 (2\pi - 0) - \tfrac14 (1 - 1) \newline
& = & \pi.
\end{array}
$$

```python
N2 = 10000
n = []
A = []
AA = 0
for i in range(1, N2 + 1):
    xi = (1.0/N2) * rnd.randint(0, N2)
    xi *= (b - a)
    xi += a
    fi = f(xi)
    Ai = (b - a) * fi
    
    n.append(i)
    if i == 1:
        AA = Ai
    else:
        AA = (A[i-2] * (i-1) + Ai) / i
    A.append(AA)

plt.plot([0, N2], [math.pi, math.pi], 'r-')
plt.plot(n, A, '-')
plt.xlabel('n')
plt.ylabel('A')
plt.grid()
plt.show()
```
