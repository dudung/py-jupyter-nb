```python
from matplotlib import pyplot as plt 
import numpy as np 
from matplotlib.animation import FuncAnimation 

fig = plt.figure(figsize=(4, 4))
axis = plt.axes(xlim =(-2, 2), ylim =(-2, 2)) 
plt.grid()

scat1, = axis.plot([], [], ls='none', ms=10, marker='o', mfc='none')
scat2, = axis.plot([], [], ls='none', ms=5, marker='o', mfc='r')

def init(): 
 scat1.set_data([], []) 
 scat2.set_data([], []) 
 return _

def animate(i): 
 x = np.linspace(-2, 2, 10) 
 y = np.sin(2 * np.pi * (x - 0.01 * i)) 
 scat1.set_data(x, y)
 scat2.set_data(y, x)
 return _

anim = FuncAnimation(fig, animate, init_func = init, 
     frames = 160, interval = 20, blit = True) 
anim.save('test.mp4', writer = 'ffmpeg', fps = 30) 
```
