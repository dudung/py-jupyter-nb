# chatterbot_venv_ipykernel_py3.8.0
+ Create virtual environment (and install ipykernel in it), create jupyter kernel, relate the environment with the kernel for `chatterbot`.
+ **Status**: $\color{\red}{\rm Fail}$.

## install python and virtualenv
1. Visit `https://www.python.org/downloads/release/python-370/`.
2. Download Python 3.8.0 https://www.python.org/ftp/python/3.8.0/python-3.8.0-amd64.exe.
3. Execute `python-3.8.0-amd64.exe`.
4. Set folder to `C:\Python\Python38`.
5. Optional features: `pip`.
6. Install `virtualenv` as suggeste in [venv vs virtualenv](https://stackoverflow.com/a/64586483/9475509).
