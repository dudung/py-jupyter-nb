# chatterbot_venv_ipykernel_py3.7.0
+ Create virtual environment (and install ipykernel in it), create jupyter kernel, relate the environment with the kernel for `chatterbot`.
+ **Note**: See previous try in [chatterbot_venv_ipykernel_py3.8.0](chatterbot_venv_ipykernel_py3.8.0.ipynb).
+ **Status**: $\color{\red}{\rm Fail}$.

## install python and virtualenv
1. Visit `https://www.python.org/downloads/release/python-370/`.
2. Download `https://www.python.org/ftp/python/3.7.0/python-3.7.0-amd64.exe`.
3. Execute `python-3.7.0-amd64.exe`.
4. Uncheck `Install launcher for all user (recommended)`.
5. Do not check `Add Python 3.7 to PATH`.
6. Choose `Customize installation`.
7. Uncheck `Documentation`, `tcl/tk and IDLE`, `Python test suite`, `py launcher`.
8. Uncheck `Create shortcuts for installed applications`.
7. Set folder to `C:\Python\Python37`.
8. Optional features: `pip`.
