+ TensorFlow does not support Python 3.11 as in [here](https://github.com/tensorflow/tensorflow/issues/58032)
+ It support only Python 3.7 - 3.10.4 as in [here](https://stackoverflow.com/a/75132976/9475509).
+ Solution: Wait until it is supported, since it is issued since 10 Oct 2022.
+ Alternative solution: Install `tf-nightly` which could contain bug as in [here](https://stackoverflow.com/a/59502169/9475509).
