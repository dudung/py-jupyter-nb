# math_function_dedic
Plot a mathematical function using Python `matplotlib` module.

```python
import math
import matplotlib.pyplot as plt
```

## function
Suppose there is a function

<a name='eqn1'></a>
\begin{equation}\tag{1}
f(x) = (x-0.5)(x-1)(x-1.5)(x-8.5)(x-9)(x-9.5)(x-3)(x-4.5)(x-5.5)(x-7)
\end{equation}

that will be drawn for $x \in [x_{\rm beg}, x_{\rm end}]$.

```python
# define a function of x
def f(x):
    y = (x-0.5)*(x-1)*(x-1.5) *\
    (x-8.5)*(x-9)*(x-9.5) *\
    (x-3)*(x-4.5)*(x-5.5)*(x-7)
    return y
```

```python
# create data
xbeg = 0
xend = 10
N = 100
dx = (xend - xbeg) / N

xx = []
yy = []

for i in range(0, N + 1):
    x = xbeg + i * dx
    y = f(x)
    
    xx.append(x)
    yy.append(y)
```

```python
# plot data
plt.ion()
fig, ax = plt.subplots()
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")

plt.ylim(-20000, 20000)

plt.plot(xx, yy)

# draw horizontal line
line = plt.Line2D((xbeg, xend), (0, 0), color='#aaa', lw=1, ls='dashed')
plt.gca().add_line(line)

# show plot result
plt.show()
```
