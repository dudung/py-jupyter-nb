# draw_shapes
Draw shapes using Python `matplotlib` module.

```python
import matplotlib.pyplot as plt
```

## circle

```python
xc = 1
yc = 0
r = 0.5
circle = plt.Circle(
    (xc, yc), radius=r,
    fc='#ccf',
    lw=2, ls='dashed', color='k'
)
plt.gca().add_patch(circle)
plt.axis('scaled')
plt.show()
```

## rectangle

```python
x = 20
y = 20
w = 80
h = 90
rectangle = plt.Rectangle(
    (x, y), w, h,
    fc='y',
    color='r', ls='dotted', lw=3
)
plt.gca().add_patch(rectangle)
plt.axis('scaled')
plt.show()
```

## line

```python
x1 = 10
y1 = 20
x2 = 50
y2 = 60
line = plt.Line2D((x1, x2), (y1, y2), color='#4c4', lw=1, ls='solid')
plt.gca().add_line(line)
plt.axis('scaled')
plt.show()
```
