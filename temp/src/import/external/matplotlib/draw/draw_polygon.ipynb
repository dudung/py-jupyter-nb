# draw_polygon
Draw polygon using Python `matplotlib` module.

```python
import math
import matplotlib.pyplot as plt
```

## polygon

```python
R = 10
N = 6
dq = 2 * math.pi / N
q0 = 0

points = []

for i in range(N):
    q = i * dq + q0
    x = R * math.cos(q)
    y = R * math.sin(q)

    p = [x, y]
    points.append(p)
    
line = plt.Polygon(
    points, closed=True,
    fill=None,
    edgecolor='r', lw=2
)
plt.gca().add_patch(line)
plt.axis('scaled')
plt.show()
```
