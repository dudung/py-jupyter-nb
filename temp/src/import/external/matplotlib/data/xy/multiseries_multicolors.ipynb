# multiseries_multicolors
Plot xy using Python `matplotlib` module for multiseries and multicolors

```python
import matplotlib.pyplot as plt
x1 = [*range(0, 10 + 1)]
x2 = [*range(10, 20 + 1)]
x3 = [*range(20, 30 + 1)]
x4 = [*range(30, 40 + 1)]

y1 = [(2*x)                for x in x1]
y2 = [(x**2 - 18*x + 100)  for x in x2]
y3 = [(22*x - 300)         for x in x3]
y4 = [(-(x - 30)**2 + 360) for x in x4]

plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.plot(
    x1, y1, '#f44',
    x2, y2, '#4f8',
    x3, y3, '#fa4',
    x4, y4, '#44f',
    linewidth = 3
)
plt.show()
```
