```python
import math
from anastruct import SystemElements

ss = SystemElements(EA=2000)
ss.add_truss_element(location=[[0, 0], [0, 2]])
ss.add_truss_element(location=[[0, 2], [2, 2]])
ss.add_truss_element(location=[[2, 2], [4, 2]])
ss.add_truss_element(location=[[4, 2], [4, 0]])
ss.add_truss_element(location=[[0, 0], [2, 2]], EA=2000*math.sqrt(2))
ss.add_truss_element(location=[[2, 2], [4, 0]], EA=2000*math.sqrt(2))


ss.add_support_hinged(node_id=1)
ss.add_support_hinged(node_id=5)

ss.point_load(Fx=20, node_id=3)
```

```python
ss.solve()
```

```python
ss.show_structure()
```

```python
ss.show_reaction_force()
```

```python
ss.show_axial_force()
```

```python
ss.show_displacement(factor=10)
```
