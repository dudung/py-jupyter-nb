```python
from anastruct import SystemElements

ss = SystemElements(EA=100)
ss.add_truss_element(location=[[0, 0], [2, 0]])
ss.add_truss_element(location=[[2, 0], [4, 0]])

ss.add_support_hinged(node_id=1)
ss.add_support_hinged(node_id=3)

ss.point_load(Fy=-10, node_id=2)
```

```python
ss.solve()
```

```python
ss.show_structure()
```

```python
ss.show_reaction_force()
```

```python
ss.show_axial_force()
```

```python
ss.show_displacement(factor=10)
```
