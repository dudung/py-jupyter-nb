# bar
Read excel file using pandas and plot bar chart using matplotlib

## read dataframe

```python
import pandas as pd
df = pd.read_excel('data/bar.xlsx', sheet_name='PYPL2022')
lang = df['Language'].tolist()
print(lang)
```

## plot for january 2022

```python
jan = [i*100 for i in df['Jan'].tolist()]

import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.rc('font', size=20)
plt.xticks(rotation='vertical')
plt.xlabel("Programming languages")
plt.ylabel("Popularity (%)")
#plt.ylim([0, 100])
plt.bar(lang, jan, color='#aaaaff', width=0.9)
plt.legend(['Jan 2022'])
plt.show()
```

## plot for february 2022

```python
feb = [i*100 for i in df['Feb'].tolist()]

import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.rc('font', size=20)
plt.xticks(rotation='vertical')
plt.xlabel("Programming languages")
plt.ylabel("Popularity (%)")
#plt.ylim([0, 100])
plt.bar(lang, jan, color='tab:blue', width=0.9)
plt.legend(['Feb 2022'])
plt.show()
```

## multiple bars chart (different width approach)

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.rc('font', size=20)
plt.xticks(rotation='vertical')
plt.xlabel("Programming languages")
plt.ylabel("Popularity (%)")
plt.bar(lang, jan, color='#aaaaff', width=0.9)
plt.bar(lang, feb, color='tab:blue', width=0.4)
plt.legend(['Jan 2022', 'Feb 2022'])
plt.show()
```

## multiple bars chart (different position and xticks approach)

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.rc('font', size=20)
plt.xticks(rotation='vertical')
plt.xlabel("Programming languages")
plt.ylabel("Popularity (%)")

w = 0.4
x1 = [*range(len(lang))]
x2 = [i+w for i in x1]
xmid = [i+w/2 for i in x1]

plt.bar(x1, jan, color='#aaaaff', width=w)
plt.bar(x2, feb, color='tab:blue', width=w)

plt.xticks(xmid, lang)
plt.legend(['Jan 2022', 'Feb 2022'])
plt.show()
```
