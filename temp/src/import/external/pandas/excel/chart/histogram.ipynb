# histogram
Read excel file using pandas and plot histogram chart using matplotlib

## read dataframe and draw histogram (sheet G)

```python
import pandas as pd
df = pd.read_excel('data/histogram.xlsx', sheet_name='G')
x = df['x'].tolist()
print(x)
```

```python
import matplotlib.pyplot as plt
plt.hist(x, 10)
plt.show()
```
