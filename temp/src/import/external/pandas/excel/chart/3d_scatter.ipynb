# 3d_scatter
Read excel file using pandas and plot 3d scatter chart using matplotlib

## read dataframe from file

```python
import pandas as pd
df = pd.read_excel('data/3d_scatter.xlsx')
print(df)
```

## get x, y, z data

```python
x = df['x'].tolist()
y = df['y'].tolist()
z = df['z'].tolist()
```

## plot data

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt

plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

ax.scatter3D(x, y, z, s=30, color='red')
ax.plot3D(x, y, z, 'tab:blue')

plt.title("A parametric curve in 3d")
plt.show()
```
