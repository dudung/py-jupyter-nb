# pie
Read excel file using pandas and plot pie chart using matplotlib

## read dataframe from file

```python
import pandas as pd
df = pd.read_excel('data/pie.xlsx')
print(df)
```

## get available columns header

```python
headers = df.columns.tolist()
print(headers)
```

## get fruit kind as list

```python
fruits = df['Item'].tolist()
print(fruits)
```

## get january data

```python
jan = df['Jan'].tolist()
print(jan)
```

## plot pie chart for january

```python
import matplotlib.pyplot as plt
plt.pie(jan, labels= fruits)
plt.show()
```

## get february data

```python
feb = df['Feb'].tolist()
print(feb)
```

## plot pie chart for february

```python
import matplotlib.pyplot as plt
plt.pie(feb, labels= fruits, explode = [0.2, 0.0, 0, 0, 0, 0, 0])
plt.show()
```

## get march data

```python
mar = df['Mar'].tolist()
print(mar)
```

## plot pie chart for march

```python
import matplotlib.pyplot as plt
plt.pie(mar, labels= fruits, explode=[0, 0.0, 0.2, 0, 0, 0, 0], shadow=True)
plt.show()
```

## get april data

```python
apr = df['Apr'].tolist()
print(apr)
```

## plot pie chart for april

```python
import matplotlib.pyplot as plt
exp=[0, 0, 0.2, 0, 0, 0, 0]
col=[
    '#0000ff',
    '#2222ff',
    '#4444ff',
    '#6666ff',
    '#8888ff',
    '#aaaaff',
    '#ccccff'
]
plt.pie(apr, labels= fruits, explode=exp, colors=col)
plt.show()
```

## get may data

```python
may = df['May'].tolist()
print(may)
```

## plot pie chart for may

```python
import matplotlib.pyplot as plt
plt.pie(feb, labels= fruits, explode = [0, 0, 0, 0, 0, 0, 0.2])
plt.legend(title='Fruits (8):')
plt.show()
```
