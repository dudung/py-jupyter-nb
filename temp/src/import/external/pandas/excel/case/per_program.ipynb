# per_program
Using pandas read excel file with various header lines

```python
def separate(x, y):
    x1 = []
    y1 = []
    x2 = []
    y2 = []
    x3 = []
    y3 = []
    x4 = []
    y4 = []

    for i in range(len(x)):
        xx = x[i]
        yy = y[i]

        if (xx < 77) and (yy < 475):
            x1.append(xx)
            y1.append(yy)
        elif (xx >= 77) and (yy < 475):
            x2.append(xx)
            y2.append(yy)
        elif (xx < 77) and (yy >= 475):
            x3.append(xx)
            y3.append(yy)
        else:
            x4.append(xx)
            y4.append(yy)

    
    return x1, y1, x2, y2, x3, y3, x4, y4
```

```python
def name(x, y, z):
    c1 = []
    c2 = []
    c3 = []
    c4 = []

    for i in range(len(x)):
        xx = x[i]
        yy = y[i]
        zz = z[i]

        if (xx < 77) and (yy < 475):
            c1.append(zz)
        elif (xx >= 77) and (yy < 475):
            c2.append(zz)
        elif (xx < 77) and (yy >= 475):
            c3.append(zz)
        else:
            c4.append(zz)
    
    return c1, c2, c3, c4
```

```python
import pandas
fn = 'data/2022_2.xlsx'
sn = 'Sheet1'
df = pandas.read_excel(fn, sheet_name=sn)
hd = list(df)
ps = df[hd[1]].unique()
print("number of ps = ", len(ps))
```

```python
import matplotlib.pyplot as plt
```

```python
psnow = ps[0]
rows = df.loc[df[hd[1]] == psnow, (hd[1], hd[2], hd[3], hd[4])]
hdnow = list(rows)
x = list(rows[hdnow[2]])
y = list(rows[hdnow[3]])
z = list(rows[hdnow[1]])
"""
c1, c2, c3, c4 = name(x, y, z)
print("ELPT < 77, UPDA < 475, n =", len(c1))
for i in c1:
    print("  ", i)
print("ELPT >= 77, UPDA < 475, n =", len(c2))
for i in c2:
    print("  ", i)
print("ELPT < 77, UPDA >= 475, n =", len(c3))
for i in c3:
    print("  ", i)
print("ELPT >= 77, UPDA >= 475, n =", len(c4))
for i in c4:
    print("  ", i)
"""
x1, y1, x2, y2, x3, y3, x4, y4 = separate(x, y)
l1 = f"ELPT < 77, UPDA < 475, n = {len(x1)}"
l2 = f"ELPT >= 77, UPDA < 475, n = {len(x2)}"
l3 = f"ELPT < 77, UPDA >= 475, n = {len(x3)}"
l4 = f"ELPT >= 77, UPDA >= 475, n = {len(x4)}"
ll = [l4, l3, l2, l1]

plt.figure(figsize=(4,5))
plt.grid()
plt.xlabel('ELPT')
plt.ylabel('UPDA')
plt.title(psnow)
plt.plot(x4, y4, 'g^', x3, y3, 'c<', x2, y2, 'y>', x1, y1, 'rv')
plt.legend(ll, loc="lower left", bbox_to_anchor=(1, 0))
plt.xlim([-25, 150])
plt.ylim([-100, 700])
plt.show()
```
