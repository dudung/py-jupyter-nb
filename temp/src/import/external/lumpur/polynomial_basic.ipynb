```python
from lumpur.num.polynomial import Polynomial

def plot_polynomial(
    p : Polynomial, x,
    grid=True, labels=['x', 'y']):

    y = p.evaluate(x)
    plt.plot(x, y)
    plt.xlabel(labels[0])
    plt.ylabel(labels[1])
    if grid:
        plt.grid()

import matplotlib.pyplot as plt
import numpy as np

p1 = Polynomial([-1, 1])
p2 = Polynomial([-3, 1])
p3 = Polynomial([-5, 1])
p4 = Polynomial([-7, 1])
p5 = Polynomial([-9, 1])
p = p1 * p2 * p3 * p4 * p5
x = list(np.arange(1, 9.1, 0.1))

plot_polynomial(p, x)
plt.show()
```

```python
from lumpur.num.polynomial import Polynomial
from lumpur.viz.plot.polynomial import plot_polynomial

p1 = Polynomial([-1, 1])
p2 = Polynomial([-3, 1])
p3 = Polynomial([-5, 1])
p4 = Polynomial([-7, 1])
p = p1 * p2 * p3 * p4
dp = p.differentiate()
d2p = dp.differentiate()
d3p = d2p.differentiate()
d4p = d3p.differentiate()
d5p = d4p.differentiate()
x = [0.1*i for i in range(10, 71)]

plot_polynomial(x, p, label='p')
plot_polynomial(x, dp, label='dpdx')
plot_polynomial(x, d2p, label='d2p/dx2')
plot_polynomial(x, d3p, label='d3p/dx3')
plot_polynomial(x, d4p, label='d4p/dx4')
```
