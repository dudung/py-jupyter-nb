# overview
Thermal energy stograge overview as suggested in https://doi.org/10.1007/s10973-019-08703-w, Fig 1.

```python
import matplotlib.pyplot as plt
from mpl_flow import Flow
```

```python
# It still from NF_3
s = [
    "Nano-\nfluid \n effective \n thermal \n conduc-\ntivity",
    "Nano-\nparticle",
    "Material \n thermal \n conduc-\ntivity",
    " Shape ",
    "  Size  ",
    "Concen-\ntration",
    "Basefluid \n type",
    "Additives",
    "Clustering",
    "Acidity \n (PH)",
    "Tempe-\nrature"
]

c = '#fff'
b1 = dict(boxstyle='round', fc='#000', ec='none')
b2 = dict(boxstyle='round', fc='#400', ec='none')
b3 = dict(boxstyle='round', fc='#040', ec='none')
b4 = dict(boxstyle='round', fc='#004', ec='none')
b5 = dict(boxstyle='round', fc='#840', ec='none')
b6 = dict(boxstyle='round', fc='#880', ec='none')
a = dict(arrowprops=dict(arrowstyle='-'))
```

```python
fig, axes = plt.subplots(1, 1, figsize=(4, 8))
ax1 = axes
f = Flow(ax=ax1, )
f.node('o', label=s[0], c=c, bbox=b1)
f.node('np', label=s[1], travel='n', color=c, bbox=b2, startpoint='o', edge_kwargs=a)
f.node(label=s[2], travel='nw', color=c, bbox=b2, startpoint='np', edge_kwargs=a)
f.node(label=s[3], travel='n', color=c, bbox=b2, startpoint='np', edge_kwargs=a)
f.node(label=s[4], travel='ne', color=c, bbox=b2, startpoint='np', edge_kwargs=a)
f.node(label=s[5], travel='e', color=c, bbox=b3, startpoint='np', edge_kwargs=a)
f.node(label=s[6], travel='e', color=c, bbox=b4, startpoint='o', edge_kwargs=a)
f.node(label=s[7], travel='se', color=c, bbox=b5, startpoint='o', edge_kwargs=a)
f.node(label=s[8], travel='w', color=c, bbox=b6, startpoint='np', edge_kwargs=a)
f.node(label=s[9], travel='sw', color=c, bbox=b1, startpoint='o', edge_kwargs=a)
f.node(label=s[10], travel='w', color=c, bbox=b1, startpoint='o', edge_kwargs=a)
plt.show()
```
