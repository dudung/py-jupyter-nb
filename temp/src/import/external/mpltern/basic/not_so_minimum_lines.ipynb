# not_so_minimum_lines
Minimum lines of code to use mpltern.

```python
import matplotlib.pyplot as plt
import mpltern

t = [0.2, 0.2, 0.6]
l = [0.6, 0.2, 0.2]
r = [0.2, 0.6, 0.2]

ax = plt.subplot(projection='ternary')
ax.grid()
ax.plot(t, l, r, color='green', linestyle='dashed', marker='o',
     markerfacecolor='none', markersize=12)
plt.show()
```
