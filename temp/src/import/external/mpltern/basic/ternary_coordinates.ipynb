# ternary_coordinates
Ternary coordinates to use in mpltern.

## 100% is in on the triangle vertices

```python
import matplotlib.pyplot as plt
import mpltern

ax = plt.subplot(projection='ternary')
ax.grid()

# top red point (1, 0, 0)
ax.plot(1, 0, 0, marker='o', ms=16, c='r', mfc='#f44', clip_on=False)

# left green point (0, 1, 0)
ax.plot(0, 1, 0, marker='o', ms=16, c='g', mfc='#4f4', clip_on=False)

# right blue point (0, 0, 1)
ax.plot(0, 0, 1, marker='o', ms=16, c='b', mfc='#44f', clip_on=False)

plt.show()
```

## exercise
url https://csmgeo.csm.jmu.edu/geollab/fichter/sedrx/readternary.html

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator
import mpltern

# point 5
l = '5'
p = [0.7, 0.2, 0.1] 

ax = plt.subplot(projection='ternary')
ax.taxis.set_minor_locator(MultipleLocator(0.1))
ax.grid(axis='t', which='minor')
ax.laxis.set_minor_locator(MultipleLocator(0.1))
ax.grid(axis='l', which='minor')
ax.raxis.set_minor_locator(MultipleLocator(0.1))
ax.grid(axis='r', which='minor')
ax.grid()
ax.text(p[0]-0.05, p[1]+0.025, p[2]+0.025, l, fontsize=12, ha='center', va='top')
ax.plot(p[0], p[1], p[2], marker='o', ms=10, c='r', mfc='#f44', clip_on=False)
plt.show()
```
