# random_radius_circular_motion
Draw circular motion with random radius

```python
import random, math
import matplotlib.pyplot as plt 
import matplotlib as mpl

# create points of data
N = 1000
r = []
for i in range(N):
    dr = (random.randint(0, 10) - 5) / 50
    r.append(1 + dr)

T = 200
x = []
y = []
for i in range(N):
    j = 2 * math.pi * i / T
    x.append(r[i] * math.cos(j))
    y.append(r[i] * math.sin(j))
    
# draw the points
mpl.rcParams['figure.figsize'] = (6, 6)
mpl.rcParams['axes.grid'] = True
plt.scatter(x, y, s=10) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("random radius of circular motion") 
plt.show()
```
