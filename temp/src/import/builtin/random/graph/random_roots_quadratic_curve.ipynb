# random_roots_quadratic_curve
Draw quadratic curve with random roots

```python
import random, math
import matplotlib.pyplot as plt 

# create roots
x1 = random.randint(1, 9)
x2 = random.randint(1, 9)

# define a quadratic function
def quadrat(x):
    y = (x - x1) * (x - x2)
    return y

xx = []
yy = []
N = 20
xbeg = 0
xend = 10
dx = (xend - xbeg) / N
for i in range(N):
    x = xbeg + i * dx
    y = quadrat(x)
    
    xx.append(x)
    yy.append(y)

# draw the points
plt.plot(xx, yy) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title('random roots')
plt.axhline(y=0.0, color='k', linestyle='-')
plt.show()
```
