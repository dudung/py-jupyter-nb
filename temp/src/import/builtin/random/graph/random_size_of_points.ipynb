# random_size_of_points
Draw points with random size

```python
import random, math
import matplotlib.pyplot as plt 
import matplotlib as mpl

# create points of data in regular position
Nx = 10
Ny = 10
x = []
y = []
d = []

for iy in range(Ny + 1):
    for ix in range(Nx + 1):
        d.append(random.randint(10, 100))
        x.append(ix * Nx)
        y.append(iy * Ny)    
    
# draw the points
mpl.rcParams['figure.figsize'] = (6, 6)
mpl.rcParams['axes.grid'] = True
plt.scatter(x, y, s=d) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("random radius of points") 
plt.show()
```
