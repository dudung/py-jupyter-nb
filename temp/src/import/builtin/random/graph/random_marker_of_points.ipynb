# random_marker_of_points
Draw points with random marker

```python
import random, math
import matplotlib.pyplot as plt 
import matplotlib as mpl

Nx = 10
Ny = 10
x = []
y = []
m = []

# create regular position and random marker
h = ".,ov^<>12348spP*hH+xXDd|_"
for iy in range(Ny + 1):
    for ix in range(Nx + 1):
        x.append(ix * Nx)
        y.append(iy * Ny)
        
        mark = h[random.randint(0, 24)]
        m.append(mark)
    
# draw the points
mpl.rcParams['figure.figsize'] = (6, 6)
mpl.rcParams['axes.grid'] = True
for x, y, m in zip(x, y, m):
    plt.scatter(x, y, s=300, c="#000", marker=m) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("random marker of points") 
plt.show()
```
