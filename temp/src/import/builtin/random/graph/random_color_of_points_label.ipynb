# random_color_of_points_label
Draw points with random color and label

```python
import random, math
import matplotlib.pyplot as plt 
import matplotlib as mpl

Nx = 10
Ny = 10
x = []
y = []
c = []
t = []

# create regular position and random color
h = "0123456789abcdef"
for iy in range(Ny + 1):
    for ix in range(Nx + 1):
        x.append(ix * Nx)
        y.append(iy * Ny)
        
        r = random.randint(9, 15)
        g = random.randint(9, 15)
        b = random.randint(9, 15)
        t.append(h[random.randint(0, 15)])
        color = "#" + h[r] + h[g] + h[b]
        c.append(color)
    
# draw the points
mpl.rcParams['figure.figsize'] = (6, 6)
mpl.rcParams['axes.grid'] = True
fig, ax = plt.subplots()
for i, txt in enumerate(x):
    ax.annotate(t[i], (x[i] - 1.5, y[i] - 1))
plt.scatter(x, y, s=400, c=c) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("random color of points") 
plt.show()
```
