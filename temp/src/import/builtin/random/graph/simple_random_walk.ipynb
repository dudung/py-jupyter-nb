# simple_random_walk
Draw simple random walk trajectory

```python
import random, math
import matplotlib.pyplot as plt 

# create points of data
x = [0]
y = [0]
N = 1000
for i in range(N):
    dx = random.randint(0, 2) - 1
    dy = random.randint(0, 2) - 1
    
    x.append(x[i] + dx)
    y.append(y[i] + dy)

# draw the points
plt.plot(x, y) 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title('random walk trajectory') 
plt.show()
```
