# random_points_with_label
Draw random points with label

```python
import random, math
import matplotlib.pyplot as plt
import matplotlib as mpl

# create points of data
x = []
y = []
N = 26

ids="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
for i in range(N):
    x.append(random.randint(0, 20))
    y.append(random.randint(0, 20))


# draw the points
mpl.rcParams['figure.figsize'] = (6, 6)
mpl.rcParams['axes.grid'] = True
fig, ax = plt.subplots()
for i, txt in enumerate(x):
    ax.annotate(ids[i], (x[i] - 0.1, y[i] - 0.1))
plt.scatter(x, y, s=400, c="#afa") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("random points with label") 
plt.show()
```
