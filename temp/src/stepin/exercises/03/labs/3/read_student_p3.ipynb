## import libraries, read data

```python
# import pandas and matplotlib libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read all data
df = pd.read_csv("student.csv")
```

## 1

```python
# get data from colum 'nilai_math' and convert it to a list
math = df['nilai_mat'].tolist()

fig = plt.figure()
fig.suptitle("Math scores", fontsize=10)

plt.hist(math)
plt.ylabel("Frequency", fontsize=10)
plt.xlabel("Score", fontsize=10)
plt.show()
```

## 2

```python
# get unique values of column 'fakultas'
faculty = df['fakultas'].unique()
print(faculty)

# calculate occurence of unique values in column 'fakultas'
N_faculty = []
for f in faculty:
    nf = len(df.groupby('fakultas').get_group(f))
    N_faculty.append(nf)

# draw horizontal bar chart
plt.ylabel("Faculty")
plt.xlabel("Student")
plt.title("Number of students")
plt.barh(faculty, N_faculty)
plt.show()
```

## 3

```python
plt.pie(N_faculty, labels = faculty)
plt.show()
```

## 4

```python
# get index of faculty with most number of students
idx = N_faculty.index(max(N_faculty))
fac = faculty[idx]
print("Faculty with most number of students =", fac)
print("Number of students =", N_faculty[idx])
```

It is easier to get faculty with maximum number of students using horizontal bar chart than pie chart.
