```python
from IPython.display import Audio
from scipy.io import wavfile
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
import numpy as np
```

```python
fns = []
frs = []
```

```python
fn = "20221305.wav"
sample_rate, data = wavfile.read(fn)
yf = rfft(data)
b = 0
f = []
for i in yf:
    v = round(np.abs(i), 0)
    if v > len(yf):
        f.append(b)
    b += 1
frs.append(f)
fns.append(fn)
```

```python
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq
from scipy.fft import fft, fftfreq
import numpy as np

fn = "20221305.wav"
sample_rate, data = wavfile.read(fn)
ryf = rfft(data)
yf = fft(data)

print(len(data), len(yf), len(ryf))
print(len(yf) / sample_rate)

b = 0
f = []
for i in yf:
    v = round(np.abs(i), 0)
    if v > len(yf):
        print(b)
    b += 1

print()
    
b = 0
f = []
for i in ryf:
    v = round(np.abs(i), 0)
    if v > len(ryf):
        print(b)
    b += 1
```

```python
def sine_wave(freq, sample_rate, duration):
    t = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    omega = 2 * np.pi * freq
    y = np.sin(omega * t)
    return t, y

_, y = sine_wave(500, 44100, 4)
yf = rfft(y)
xf = rfftfreq(44100 * 4, 1/44100)

print(len(y), len(yf), len(xf))
b = 0
f = []
for i in yf:
    v = round(np.abs(i), 0)
    if v > len(yf) / 4:
        print(b, xf[b])
    b += 1
```
