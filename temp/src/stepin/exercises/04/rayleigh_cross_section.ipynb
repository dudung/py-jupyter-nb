# rayleigh_cross_section
Plot Rayleigh scattering cross section

## equation

+ There is an equation in the form of

![](rayleigh-cross-section.png)

$$\tag{1}
C_S = \frac{8\pi}{3} \left( \frac{\varepsilon_1 - \varepsilon_2}{\varepsilon_1 + 2 \varepsilon_2} \right)^2 k^4 a^6.
$$

+ See [discussion](https://github.com/NazlaInnaya95/Diskusi/issues/1#issuecomment-1326961492) before previous equation.
+ The equation might be written in different form as in [here](https://en.wikipedia.org/wiki/Rayleigh_scattering#Small_size_parameter_approximation)

$$\tag{2}
\sigma_S = \frac{2 \pi^5}{3} \frac{d^6}{\lambda^4} \left( \frac{n^2 - 1}{n^2 + 2} \right)^2,
$$

    which may have different limitation and interpretation.
+ Only Eqn $(1)$ will be used here.
+ Proposed [example code](https://github.com/waannuullll/scat-single-sphere/blob/master/Main_Program.m) have not yet been read, since it is in MATLAB.

## parameters and function

```python
# air
eps2 = 1.0054

# list of materials
materials = {
    "PE": 2.25,
    "Paper": 2.1,
    "PTFE": 2.4,
    "PS": 4.7,
    "Glass": 7,
    "Rubber": 11.68,
    "Silicon": 30,
    "Methanol": 4.5
}

# sphere radius (in nm)
a = 40

# wavelength l is in nm
def wavenumber(l):
    k = 2 * math.pi / (l * 1E-9)
    return k

# calculate Rayleigh cross section
def rayleigh_cross_section(l, a, eps1, eps2):
    g = (eps1 - eps2) / ( eps1 + 2 * eps2)
    k = wavenumber(l)
    b = a * 1E-9
    cs = (8 * math.pi / 3) * (g**2) * (k**4) * (b**6)
    return cs
```

## polystyrene (PS)

```python
import math
import matplotlib.pyplot as plt

# material
eps1 = materials["PS"]

# wavelength
lmin = 300
lmax = 800
dl = 10
l = [i for i in range(lmin, lmax + dl, dl)]

# calculate Rayleigh cross section
c1 = []
for i in l:
    cs = rayleigh_cross_section(i, a, eps1, eps2)
    c1.append(cs)

# draw the points
plt.plot(l, c1, '-o') 
plt.xlabel('lambda') 
plt.ylabel('cross section') 
plt.title('Polystyerene') 
plt.show()
```
